## Основные теоретические положения

**Метод наименьших квадратов (МНК)** — статистический метод оценки параметров регрессионной модели, основанный на минимизации суммы квадратов отклонений наблюдаемых значений зависимой переменной от значений, предсказанных моделью:
$$
S(\beta_0, \beta_1, \dots, \beta_k) = \sum_{i=1}^{n} \varepsilon_i^2 = \sum_{i=1}^{n} \left[y_i - f(x_i; \beta_0, \beta_1, \dots, \beta_k)\right]^2 \to \min
$$
Необходимо найти такие оценки параметров $\hat{\beta}_j$, при которых сумма квадратов остатков минимальна.

**Статистические свойства МНК-оценок** (при выполнении условий Гаусса–Маркова):
- *Несмещённость*: $M(\hat{\beta}_j) = \beta_j$;
- *Эффективность*: среди всех линейных несмещённых оценок МНК-оценки имеют наименьшую дисперсию (BLUE — Best Linear Unbiased Estimator);
- *Состоятельность*: $\hat{\beta}_j \xrightarrow{P} \beta_j$ при $n \to \infty$;
- *Асимптотическая нормальность*: при больших $n$ распределение оценок приближается к нормальному.

**Функция регрессии** — зависимость условного математического ожидания одной случайной величины от значения другой:
$$
M(Y \mid x) = f(x), \quad M(X \mid y) = g(y)
$$
В случае линейной регрессии: $M(Y \mid x) = a x + b$.

**Выборочная прямая среднеквадратической регрессии $Y$ на $X$** — линейная функция, минимизирующая сумму квадратов вертикальных отклонений:
$$
\bar{y}_x = \bar{y} + r_{xy} \frac{s_y}{s_x}(x - \bar{x}) = a x + b,
$$
где $a = \rho_{xy} = r_{xy} \dfrac{s_y}{s_x} $, $b = \bar{y} - a \bar{x}$.

**Выборочная прямая среднеквадратической регрессии $X$ на $Y$** — линейная функция, минимизирующая сумму квадратов горизонтальных отклонений:
$$
\bar{x}_y = \bar{x} + r_{xy} \frac{s_x}{s_y}(x - \bar{y}) = a y + b,
$$
где $a = \rho_{yx} = r_{xy} \dfrac{s_x}{s_y} $, $b = \bar{x} - a \bar{y}$.

**Геометрическая интерпретация**: обе прямые проходят через точку $(\bar{x}, \bar{y})$; угол между ними характеризует тесноту связи: при $\lvert r_{xy} \rvert = 1$ прямые совпадают, при $r_{xy} = 0$ — перпендикулярны.

**Корреляционное отношение** — мера тесноты произвольной (в том числе нелинейной) статистической связи, основанная на разложении дисперсии:
$$
\eta_{yx}^2 = \frac{\sigma_{\text{межгруп}}^2}{\sigma_{\text{общ}}^2} = \frac{M\left[\left(M(Y \mid X) - M(Y)\right)^2\right]}{D(Y)}, \quad
\eta_{xy}^2 = \frac{M\left[\left(M(X \mid Y) - M(X)\right)^2\right]}{D(X)}
$$
Выборочная оценка:
$$
\bar{\eta}_{yx}^2 = \frac{\sum\limits_{j=1}^{k} n_j (\bar{y}_j - \bar{y})^2}{\sum\limits_{i=1}^{n} (y_i - \bar{y})^2},
$$
где $k$ — число групп по $X$, $n_j$ — объём $j$-й группы, $\bar{y}_j$ — среднее в группе.

**Свойства корреляционного отношения**:
- $0 \leq \eta \leq 1$;
- $\eta = 0$ — отсутствие статистической связи;
- $\eta = 1$ — функциональная зависимость;
- $\eta_{yx} \geq \lvert r_{xy} \rvert$, причём равенство достигается только при линейной связи.

**Корреляционная таблица** — инструмент для вычисления корреляционного отношения по сгруппированным данным. Содержит условные распределения $Y$ при фиксированных значениях (интервалах) $X$ и позволяет оценить условные средние $\bar{y}_j$, необходимые для расчёта $\eta_{yx}$.

**Показатели качества регрессии**:
- *Теоретический коэффициент детерминации*:
$$
R^2 = 1 - \frac{\sum\limits_{i=1}^{n} (y_i - \bar{y}_i)^2}{\sum\limits_{i=1}^{n} (y_i - \bar{y})^2} = \frac{\sum\limits_{i=1}^{n} (\bar{y}_i - \bar{y})^2}{\sum\limits_{i=1}^{n} (y_i - \bar{y})^2}
$$
Показывает долю дисперсии $Y$, объяснённую моделью; $0 \leq R^2 \leq 1$.

- *Средняя квадратическая ошибка*:
$$
S_\varepsilon = \sqrt{\frac{1}{n - p} \sum_{i=1}^{n} (y_i - \bar{y}_i)^2},
$$
где $p$ — число оцениваемых параметров; Характеризует точность прогноза в единицах $Y$


- *Средняя ошибка аппроксимации*:
$$
A = \frac{1}{n} \sum_{i=1}^{n} \left| \frac{y_i - \bar{y}_i}{y_i} \right| \cdot 100\%
$$
Позволяет оценить относительную точность модели; считается приемлемой при $A < 10\%$.

# Задания 1-2

Отобразить двумерную выборку на графике. Для заданной выборки построить уравнения средней квадратичной регрессии $x$ на $y$ и $y$ на $x$ соответственно. Построить полученные прямые на множестве выборки. Объяснить результаты.

```python
x = df['nu'].values
y = df['E'].values

# Основные статистики
x_mean = np.mean(x)
y_mean = np.mean(y)
x_std = np.std(x, ddof=1)
y_std = np.std(y, ddof=1)
r_xy = np.corrcoef(x, y)[0, 1]

print(f"Среднее X: {x_mean:.4f}")
print(f"Среднее Y: {y_mean:.4f}")
print(f"СКО X: {x_std:.4f}")
print(f"СКО Y: {y_std:.4f}")
print(f"Коэффициент корреляции: {r_xy:.4f}")

# Коэффициенты регрессии
b_yx = r_xy * (y_std / x_std)
b_xy = r_xy * (x_std / y_std)

a_yx = y_mean - b_yx * x_mean
a_xy = x_mean - b_xy * y_mean

print(f"\nРегрессия Y на X: y = {a_yx:.4f} + {b_yx:.4f} * x")
print(f"Регрессия X на Y: x = {a_xy:.4f} + {b_xy:.4f} * y")

# Построение графика
plt.figure(figsize=(10, 8))
plt.scatter(x, y, alpha=0.6, label='Выборочные данные', color='blue')

# Прямая регрессии Y на X
x_line = np.linspace(x.min(), x.max(), 100)
y_line_yx = a_yx + b_yx * x_line
plt.plot(x_line, y_line_yx, 'r-', linewidth=2, label='Y на X')

# Прямая регрессии X на Y (переводим в форму y = f(x) для построения)
y_line = np.linspace(y.min(), y.max(), 100)
x_line_xy = a_xy + b_xy * y_line
plt.plot(x_line_xy, y_line, 'g-', linewidth=2, label='X на Y')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Двумерная выборка и прямые регрессии')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

По выборочным оценкам матожидания, СКО и коэффициента корреляции построены прямые среднеквадратической регрессии $Y$ на $X$ и $X$ на $Y$. Графики прямых отображены совместно с диаграммой рассеяния выборки. Как и ожидалось, прямые пересекаются в точке $(\bar{x}, \bar{y})$.

# Задание 3

Составить корреляционную таблицу для нахождения выборочного корреляционного отношения. Убедиться, что неравенства $η_{xy}\geq|r_{xy}|$ и $η_{yx}\geq|r_{xy}|$ выполняются.

```python
# Расчет числа интервалов по формуле Стерджесса
k = round(1 + 3.322 * math.log10(N))

# Двумерный интервальный ряд
H, x_edges, y_edges = np.histogram2d(x, y, bins=k)

# Середины интервалов
x_mid = (x_edges[:-1] + x_edges[1:]) / 2
y_mid = (y_edges[:-1] + y_edges[1:]) / 2

# Транспонируем H
freq_table = H.T.astype(int)

# Вычисление частотных сумм
n_x = freq_table.sum(axis=0)
n_y = freq_table.sum(axis=1)
```

```python
# Выборочные средние
x_mean_v = np.sum(n_x * x_mid) / N
y_mean_v = np.sum(n_y * y_mid) / N

# Выборочные СКО
x_std_v = np.sqrt(np.sum(n_x * (x_mid - x_mean_v)**2) / N)
y_std_v = np.sqrt(np.sum(n_y * (y_mid - y_mean_v)**2) / N)

# Двойная сумма
double_sum = 0
for i in range(len(y_mid)):
    for j in range(len(x_mid)):
        double_sum += freq_table[i, j] * y_mid[i] * x_mid[j]

# Коэффициент корреляции
r_xy = (double_sum - N * x_mean_v * y_mean_v) / (N * x_std_v * y_std_v)
```

```python
# Групповые средние Y по X
y_bar_x = np.zeros(k)
for j in range(k):
    if n_x[j] > 0:
        y_bar_x[j] = np.sum(freq_table[:, j] * y_mid) / n_x[j]

# Групповые средние X по Y
x_bar_y = np.zeros(k)
for i in range(k):
    if n_y[i] > 0:
        x_bar_y[i] = np.sum(freq_table[i, :] * x_mid) / n_y[i]

# Корреляционная таблица с групповыми средними
x_labels = [f"{val:.2f}" for val in x_mid]
y_labels = [f"{val:.2f}" for val in y_mid]

corr_table = pd.DataFrame(freq_table, index=y_labels, columns=x_labels)
corr_table['n_y'] = n_y
corr_table['x_bar_y'] = np.round(x_bar_y, 2)
corr_table.loc['n_x'] = n_x.tolist() + [N] + ['']
corr_table.loc['y_bar_x'] = np.round(y_bar_x, 2).tolist() + [''] + ['']

print("Корреляционная таблица")
corr_table
```

```python
# Внутригрупповая дисперсия X по Y
D_x_within = 0
for i in range(k):
    if n_y[i] > 1:
        # Дисперсия X в i-й группе по Y
        D_x_group_i = np.sum(freq_table[i, :] * (x_mid - x_bar_y[i])**2) / (n_y[i] - 1)
        D_x_within += n_y[i] * D_x_group_i

D_x_within /= N

# Межгрупповая дисперсия X по Y
D_x_between = np.sum(n_y * (x_bar_y - x_mean)**2) / (N - 1)

# Общая дисперсия X
D_x_total = np.sum(n_x * (x_mid - x_mean)**2) / (N - 1)

# Внутригрупповая дисперсия Y по X
D_y_within = 0
for j in range(k):
    if n_x[j] > 1:
        # Дисперсия Y в j-й группе по X
        D_y_group_j = np.sum(freq_table[:, j] * (y_mid - y_bar_x[j])**2) / (n_x[j] - 1)
        D_y_within += n_x[j] * D_y_group_j
D_y_within /= N

# Межгрупповая дисперсия Y по X
D_y_between = np.sum(n_x * (y_bar_x - y_mean)**2) / (N - 1)

# Общая дисперсия Y
D_y_total = np.sum(n_y * (y_mid - y_mean)**2) / (N - 1)

print("Дисперсии для X:")
print(f"Общая: {D_x_total:.4f}")
print(f"Межгрупповая: {D_x_between:.4f}")
print(f"Внутригрупповая: {D_x_within:.4f}")
print(f"Сумма межгрупповой и внутригрупповой: {D_x_between + D_x_within:.4f}")
print()

print("Дисперсии для Y:")
print(f"Общая: {D_y_total:.4f}")
print(f"Межгрупповая: {D_y_between:.4f}")
print(f"Внутригрупповая: {D_y_within:.4f}")
print(f"Сумма межгрупповой и внутригрупповой: {D_y_between + D_y_within:.4f}")
print()

# Корреляционные отношения
eta_xy = np.sqrt(D_x_between / D_x_total)
eta_yx = np.sqrt(D_y_between / D_y_total)

print(f"Корреляционное отношение eta_xy = {eta_xy:.4f}")
print(f"Корреляционное отношение eta_yx = {eta_yx:.4f}")
print(f"Коэффициент корреляции |r_xy| = {abs(r_xy):.4f}")
```

Неравенства $\eta_{xy} \geq |r_{xy}|$ и $\eta_{yx} \geq |r_{xy}|$ выполняются:
$0.8668 \geq 0.8625$ и $0.8712 \geq 0.8625$
что подтверждает корректность вычислений. Поскольку корреляционные отношения лишь незначительно превышают модуль коэффициента корреляции (разница $< 0.01$), связь между переменными близка к линейной. Оба показателя указывают на тесную статистическую зависимость ($\eta > 0.7$).

# Задание 4

Для заданной выборки построить корреляционную кривую параболического вида $y=β_0+β_1x+β_2x^2$, сделать выводы.

```python
# Подготовка данных для полиномиальной регрессии 2-й степени
X_poly = np.vstack([x**2, x, np.ones(N)]).T

# Решение системы нормальных уравнений МНК
coeffs = np.linalg.lstsq(X_poly, y, rcond=None)[0]
a, b, c = coeffs

print("Параболическая регрессия:")
print(f"y = {a:.6f}*x^2 + {b:.6f}*x + {c:.6f}")

# Предсказанные значения
y_pred_poly = a * x**2 + b * x + c

# Показатели качества
SS_res = np.sum((y - y_pred_poly)**2)
SS_tot = np.sum((y - y_mean)**2)

R2_poly = 1 - (SS_res / SS_tot)
p = 3
S_epsilon_poly = np.sqrt(SS_res / (N - p))
A_poly = np.mean(np.abs((y - y_pred_poly) / y)) * 100

print("Показатели качества:")
print(f"R^2 = {R2_poly:.4f}")
print(f"S_epsilon = {S_epsilon_poly:.4f}")
print(f"A = {A_poly:.2f}%")

# График
x_smooth = np.linspace(x.min(), x.max(), 200)
y_smooth = a * x_smooth**2 + b * x_smooth + c

plt.figure(figsize=(8, 5))
plt.scatter(x, y, alpha=0.6, label='Выборочные данные', color='blue', s=30)
plt.plot(x_smooth, y_smooth, 'r-', linewidth=2.5, label=f'y = {a:.4f}x^2 + {b:.4f}x + {c:.4f}')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Параболическая регрессия')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

Построена корреляционная кривая по уравнению $y=β_0+β_1x+β_2x^2$. Коэффициент при второй степени близок к нулю, это говорит о преобладании линейной зависимости над нелинейной. Показатели качества регрессии высокие, значит данная модель может быть использована для оценки зависимости.

# Задание 5

Для заданной выборки построить корреляционную кривую в зависимости от номера варианта и сделать выводы.  
Вариант 19  
19 mod 6 = 1 => Вариант 1  
Cтепенная функция: $y=β_0x^{β_1}$

```python
# Линеаризация: ln(y) = ln(beta_0) + beta_1 * ln(x)
mask = (x > 0) & (y > 0)
x_log = np.log(x[mask])
y_log = np.log(y[mask])

# Построение линейной регрессии для линеаризованных данных
X_log = np.vstack([x_log, np.ones(len(x_log))]).T
coeffs_log = np.linalg.lstsq(X_log, y_log, rcond=None)[0]
beta_1 = coeffs_log[0]
beta_0_log = coeffs_log[1]
beta_0 = np.exp(beta_0_log)

print("Степенная регрессия:")
print(f"y = {beta_0:.6f} * x^{beta_1:.6f}")

# Предсказанные значения в исходной шкале
y_pred_power = beta_0 * (x[mask] ** beta_1)

# Показатели качества
residuals = y[mask] - y_pred_power
SS_res = np.sum(residuals**2)
SS_tot = np.sum((y[mask] - np.mean(y[mask]))**2)

R2_power = 1 - (SS_res / SS_tot)
p = 2
n_power = len(y_pred_power)
S_epsilon_power = np.sqrt(SS_res / (n_power - p))
A_power = np.mean(np.abs(residuals / y[mask])) * 100

print("Показатели качества:")
print(f"R^2 = {R2_power:.4f}")
print(f"S_epsilon = {S_epsilon_power:.4f}")
print(f"A = {A_power:.2f}%")

# График
x_smooth = np.linspace(x[mask].min(), x[mask].max(), 200)
y_smooth = beta_0 * (x_smooth ** beta_1)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(x[mask], y[mask], alpha=0.6, label='Выборочные данные', color='blue', s=30)
plt.plot(x_smooth, y_smooth, 'g-', linewidth=2.5, label=f'y = {beta_0:.2f} * x^{beta_1:.4f}')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Степенная регрессия')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

Степенная регрессия также имеет высокие оценки качества, сравнимые с параболической регрессией. Однако коэффициент в показателе функции близок к единице, поэтому кривая зависимости близка к прямой. Это еще один аргумент в пользу того, что выборка согласуется с предположением о линейной зависимости.

Построение альтернативных регрессионных моделей (параболической и степенной) подтвердило этот вывод. В параболической регрессии коэффициент при квадратичном члене оказался близок к нулю, а в степенной функции показатель степени составил $1.23$, что близко к единице. Все модели продемонстрировали высокие показатели качества: коэффициент детерминации $R^2$ близок к $0.75$, средняя ошибка аппроксимации находится в приемлемых пределах. Таким образом, для описания зависимости между исследуемыми параметрами достаточно использовать простую линейную регрессию.
