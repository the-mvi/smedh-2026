## Основные теоретические положения

**Интервальное оценивание**

 Доверительным интервалом называется интервал, который с вероятностью $\gamma$ накрывает неизвестное истинное значение параметра распределения $\theta$:
$$ P(\theta - \delta < \theta < \theta + \delta) = \gamma $$

**Доверительный интервал для математического ожидания ($a$) при неизвестном СКО**

Если случайная величина распределена нормально и среднеквадратичное отклонение генеральной совокупности неизвестно, доверительный интервал вычисляется через распределение Стьюдента:
$$ \bar{x}_{\textit{в}} - t_{\gamma} \cdot \frac{s}{\sqrt{n}} < a < \bar{x}_{\textit{в}} + t_{\gamma} \cdot \frac{s}{\sqrt{n}} $$
где $\bar{x}_{\textit{в}}$ — выборочное среднее, $s$ — исправленное выборочное СКО, $n$ — объем выборки, $t_{\gamma}$ — критическое значение $t$-распределения Стьюдента для доверительной вероятности $\gamma$.

**Доверительный интервал для среднеквадратического отклонения ($\sigma$)**

Доверительный интервал для СКО строится с использованием распределения хи-квадрат ($\chi^2$). Для надежности $\gamma$ вероятность выхода за границы интервала равна $\alpha = 1 - \gamma$. Для построения двустороннего интервала применяют вероятности $\alpha / 2$ и $1 - \alpha / 2$.
$$ s(1-q) < \sigma < s(1+q) $$
где q, определяющий ширину интервала, вычисляется по формуле:
$$\mathbb{P}\left(\frac{\sqrt{n-1}}{1+q} < \chi < \frac{\sqrt{n-1}}{1-q}\right) = \gamma$$

**Проверка гипотезы о нормальном распределении (критерий согласия Пирсона $\chi^2$)**

Статистическая гипотеза $H_0$ — исследуемая случайная величина подчиняется нормальному закону распределения. Для её проверки используется критерий согласия Пирсона.
Наблюдаемое значение критерия вычисляется по сгруппированным данным (интервальный ряд):
$$ \chi^2_{\textit{набл}} = \sum_{i=1}^k \frac{(n_i - n'_i)^2}{n'_i} $$
где $k$ — число интервалов, $n_i$ — эмпирические частоты, $n'_i = n \cdot p_i$ — теоретические частоты ($p_i$ — вероятность попадания в $i$-й интервал при нормальном распределении).

Вычисление теоретических вероятностей:
$$ p_i = \Phi\left(\frac{x_{i+1} - \bar{x}_{\textit{в}}}{s}\right) - \Phi\left(\frac{x_i - \bar{x}_{\textit{в}}}{s}\right) $$
где $\Phi(x)$ — функция Лапласа, $x_i, x_{i+1}$ — границы $i$-го интервала (а для крайних интервалов берутся бесконечные границы $-\infty$ и $+\infty$, чтобы $\sum p_i = 1$).

Правило принятия решения:
Гипотеза отвергается, если наблюдаемое значение превышает критическое $\chi^2_{\textit{набл}} \ge \chi^2_{\textit{крит}}(\alpha, df)$.
Число степеней свободы вычисляется как $df = k - 3$ (так как оценивается два параметра: $a$ и $\sigma$, плюс условие равенства суммы частот $n$).

## Выполнение работы

```python
from math import ceil, log10

import numpy as np
import pandas as pd
from scipy import stats
from IPython.display import display_markdown
```

```python

df = pd.read_csv('../samples.csv')
length_ranged = df.SepalLengthCm.sort_values().to_numpy()
width_ranged = df.SepalWidthCm.sort_values().to_numpy()
```

  - Вычислить точность и доверительный интервал для математического ожидания при неизвестном среднеквадратичном отклонении при заданном объёме выборки для доверительной точности $ \gamma \in \{0.95, 0.99\} $. Сделать выводы.


```python
def calc_mean_ci(data: np.ndarray, gammas: list[float]) -> pd.DataFrame:
    """
    Вычисляет доверительный интервал для математического ожидания
    при неизвестном среднеквадратичном отклонении.
    """
    n = len(data)
    mean = np.mean(data)
    s = np.std(data, ddof=1)
    
    results = []
    for gamma in gammas:
        alpha = 1 - gamma
        t_val: float = stats.t.ppf(1 - alpha / 2, df=n - 1)
        margin = t_val * s / np.sqrt(n)
        ci_lower = mean - margin
        ci_upper = mean + margin
        
        results.append({
            r"$\gamma$": gamma,
            r"Доверительный интервал": f"({round(ci_lower, 4)}; {round(ci_upper, 4)})"
        })
        
    return pd.DataFrame(results)

display_markdown("### Длина чашелистика (SepalLengthCm)", raw=True)
display_markdown(calc_mean_ci(length_ranged, [0.95, 0.99]).to_markdown(index=False), raw=True)

display_markdown("### Ширина чашелистика (SepalWidthCm)", raw=True)
display_markdown(calc_mean_ci(width_ranged, [0.95, 0.99]).to_markdown(index=False), raw=True)
```

С увеличением уровня надежности $\gamma$ (от 0.95 до 0.99) увеличивается значение квантиля распределения Стьюдента $t_{\gamma}$, что приводит к увеличению точности оценки. То есть, для достижения большей уверенности в том, что интервал накроет истинное значение математического ожидания, сам интервал должен быть расширен.

  - Для вычисления границ доверительного интервала для среднеквадратичного отклонения определить значение $ q $ при заданных $ \gamma $ и $ n $. Построить доверительные интервалы, сделать выводы.


```python
def calc_std_ci(data: np.ndarray, gammas: list[float]) -> pd.DataFrame:
    """
    Вычисляет доверительный интервал для среднеквадратичного отклонения.
    Для расчета применяются квантили распределения хи-квадрат, так как они
    позволяют получить точные (несимметричные) границы.
    """
    n = len(data)
    s = np.std(data, ddof=1)
    
    results = []
    for gamma in gammas:
        alpha = 1 - gamma
        
        # Квантили распределения хи-квадрат
        chi2_left: float = stats.chi2.ppf(alpha / 2, df=n - 1)
        chi2_right: float = stats.chi2.ppf(1 - alpha / 2, df=n - 1)
        
        # Границы для СКО
        ci_lower = s * np.sqrt((n - 1) / chi2_right)
        ci_upper = s * np.sqrt((n - 1) / chi2_left)
        
        q_approx = (ci_upper - ci_lower) / (2 * s) # Примерное симметричное 'q' для оценки
        
        results.append({
            r"$\gamma$": gamma,
            r"$q_{approx}$": round(q_approx, 4),
            r"Доверительный интервал": f"({round(ci_lower, 4)}; {round(ci_upper, 4)})"
        })
        
    return pd.DataFrame(results)

display_markdown("### Длина чашелистика (SepalLengthCm)", raw=True)
display_markdown(calc_std_ci(length_ranged, [0.95, 0.99]).to_markdown(index=False), raw=True)

display_markdown("### Ширина чашелистика (SepalWidthCm)", raw=True)
display_markdown(calc_std_ci(width_ranged, [0.95, 0.99]).to_markdown(index=False), raw=True)
```

Так же, как и с математическим ожиданием, с ростом доверительной вероятности $\gamma$ от 0.95 до 0.99 доверительный интервал для среднеквадратичного отклонения расширяется. В силу асимметрии распределения $\chi^2$, сам интервал относительно выборочного СКО ($s$) также является несимметричным.

  - Проверить гипотезу о нормальности заданного распределения с помощью критерия $ \chi^2 $ (Пирсона). Для этого необходимо найти теоретические частоты и вычислить наблюдаемое значение критерия. Для удобства вычисления необходимо заполнить табл. 1.


```python
def create_freq_table(data: np.ndarray) -> tuple[np.ndarray, np.ndarray]:
    n = len(data)
    intervals_number = ceil(1 + 3.322 * log10(n))
    data_min, data_max = data[0], data[-1]
    
    interval_size = (data_max - data_min) / intervals_number
    bins = data_min + np.array([i * interval_size for i in range(intervals_number + 1)])
    bins[-1] += 1e-10
    
    cuts = pd.cut(data, bins=bins, right=False)
    interval_counts = cuts.value_counts().sort_index().to_numpy()
    return bins, interval_counts

def calc_chi2_pearson(data: np.ndarray) -> tuple[pd.DataFrame, float, float]:
    n = len(data)
    mean = np.mean(data)
    s = np.std(data, ddof=1)
    
    bins, n_i = create_freq_table(data)
    intervals_count = len(n_i)
    
    # Чтобы сумма вероятностей была 1, расширяем крайние интервалы в бесконечность
    prob_bins = bins.copy()
    prob_bins[0] = -np.inf
    prob_bins[-1] = np.inf
    
    p_i = np.zeros(intervals_count)
    for i in range(intervals_count):
        p_i[i] = stats.norm.cdf(prob_bins[i+1], loc=mean, scale=s) - stats.norm.cdf(prob_bins[i], loc=mean, scale=s)
        
    n_prime_i = n * p_i
    
    diff_sq = (n_i - n_prime_i)**2
    diff_sq_norm = diff_sq / n_prime_i
    n_i_sq = n_i**2
    n_i_sq_norm = n_i_sq / n_prime_i
    
    df_chi2 = pd.DataFrame({
        "$i$": np.arange(1, intervals_count + 1),
        "$[x_i, x_{i+1})$": [f"[{round(bins[i], 2)}, {round(bins[i+1], 2)})" for i in range(intervals_count)],
        "$n_i$": n_i,
        "$p_i$": np.round(p_i, 4),
        "$n'_i$": np.round(n_prime_i, 4),
        "$(n_i - n'_i)^2$": np.round(diff_sq, 4),
        r"$\frac{(n_i - n'_i)^2}{n'_i}$": np.round(diff_sq_norm, 4),
        "$n_i^2$": n_i_sq,
        r"$\frac{n_i^2}{n'_i}$": np.round(n_i_sq_norm, 4),
    })
    
    sum_row = pd.Series({
        "$i$": r"$\Sigma$",
        "$[x_i, x_{i+1})$": "-",
        "$n_i$": n_i.sum(),
        "$p_i$": np.round(p_i.sum(), 4),
        "$n'_i$": np.round(n_prime_i.sum(), 4),
        "$(n_i - n'_i)^2$": "-",
        r"$\frac{(n_i - n'_i)^2}{n'_i}$": np.round(diff_sq_norm.sum(), 4),
        "$n_i^2$": "-",
        r"$\frac{n_i^2}{n'_i}$": np.round(n_i_sq_norm.sum(), 4)
    })
    
    df_chi2 = pd.concat([df_chi2, sum_row.to_frame().T], ignore_index=True)
    
    chi2_obs1 = diff_sq_norm.sum()
    chi2_obs2 = n_i_sq_norm.sum() - n
    
    return df_chi2, chi2_obs1, chi2_obs2

display_markdown("### Таблица Пирсона: Длина чашелистика (SepalLengthCm)", raw=True)
df_chi2_len, obs1_len, obs2_len = calc_chi2_pearson(length_ranged)
display_markdown(df_chi2_len.to_markdown(index=False), raw=True)

display_markdown("### Таблица Пирсона: Ширина чашелистика (SepalWidthCm)", raw=True)
df_chi2_width, obs1_width, obs2_width = calc_chi2_pearson(width_ranged)
display_markdown(df_chi2_width.to_markdown(index=False), raw=True)
```

  - Доказать, что $$ \chi^2_{\textit{набл}} = \sum_{i = 1}^k\frac{n^2_i}{n'_i} - n. $$ Проконтролировать корректность вычисления $ \chi^2_{\textit{набл}} $. 

**Доказательство:**

$$ \chi^2_{\textit{набл}} = \sum_{i=1}^k \frac{(n_i - n'_i)^2}{n'_i} $$

$$ \chi^2_{\textit{набл}} = \sum_{i=1}^k \frac{n_i^2 - 2n_i n'_i + (n'_i)^2}{n'_i} = \sum_{i=1}^k \left( \frac{n_i^2}{n'_i} - \frac{2n_i n'_i}{n'_i} + \frac{(n'_i)^2}{n'_i} \right) = \sum_{i=1}^k \left( \frac{n_i^2}{n'_i} - 2n_i + n'_i \right) $$

$$ \chi^2_{\textit{набл}} = \sum_{i=1}^k \frac{n_i^2}{n'_i} - 2 \sum_{i=1}^k n_i + \sum_{i=1}^k n'_i $$

Известно, что сумма всех эмпирических частот равна объёму выборки: $ \sum_{i=1}^k n_i = n $. 
Аналогично, сумма теоретических частот: $ \sum_{i=1}^k n'_i = \sum_{i=1}^k (n \cdot p_i) = n \sum_{i=1}^k p_i = n \cdot 1 = n $.

$$ \chi^2_{\textit{набл}} = \sum_{i=1}^k \frac{n_i^2}{n'_i} - 2n + n = \sum_{i=1}^k \frac{n_i^2}{n'_i} - n $$

Что и требовалось доказать.

```python
print("Контроль вычислений для длины чашелистика (SepalLengthCm):")
print(f"Сумма (n_i - n'_i)^2 / n'_i: {obs1_len:.4f}")
print(f"Сумма (n_i^2 / n'_i) - n : {obs2_len:.4f}")
print(f"Совпадение: {np.isclose(obs1_len, obs2_len)}\n")

print("Контроль вычислений для ширины чашелистика (SepalWidthCm):")
print(f"Сумма (n_i - n'_i)^2 / n'_i: {obs1_width:.4f}")
print(f"Сумма (n_i^2 / n'_i) - n : {obs2_width:.4f}")
print(f"Совпадение: {np.isclose(obs1_width, obs2_width)}")
```

  - По заданному уровню значимости $ \alpha = 0.05 $ и числу степеней свободы $ df $ найти критическую точку $ \chi^2_{\textit{крит}} $ и сравнить с наблюдаемым значением. Сделать выводы.

```python
def check_hypothesis(data_name: str, chi2_obs: float, k: int, alpha: float = 0.05):
    # k - количество интервалов. Оцениваются 2 параметра (ср. арифметическое и СКО), поэтому df = k - 2 - 1
    df_chi = k - 3
    chi2_crit = stats.chi2.ppf(1 - alpha, df_chi)
    
    print(f"Анализ для признака: {data_name}")
    print(f"Количество интервалов (k): {k}")
    print(f"Число степеней свободы (df = k - 3): {df_chi}")
    print(f"Уровень значимости (alpha): {alpha}")
    print(f"Наблюдаемое значение хи-квадрат: {chi2_obs:.4f}")
    print(f"Критическое значение хи-квадрат: {chi2_crit:.4f}")
    
    if chi2_obs < chi2_crit:
        print("Вывод: Наблюдаемое значение МЕНЬШЕ критического. Нет оснований отвергать гипотезу о нормальном распределении.\n")
    else:
        print("Вывод: Наблюдаемое значение БОЛЬШЕ или равно критическому. Гипотеза о нормальном распределении отвергается.\n")

k_len = len(df_chi2_len) - 1 # Т.к. последняя строка это сумма
check_hypothesis("Длина чашелистика (SepalLengthCm)", obs1_len, k_len)

k_width = len(df_chi2_width) - 1
check_hypothesis("Ширина чашелистика (SepalWidthCm)", obs1_width, k_width)
```
