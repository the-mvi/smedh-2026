## Основные теоретические положения

Корреляционный момент определяется как:
$$\mu_{xy}=M\{[x - M(X)] \cdot [y - M(Y)]\}$$

Коэффициент корреляции определяется как:
$$r_{xy}=\frac{\mu_{xy}}{\sigma_x\sigma_y}$$

Для коэффициента корреляции справедливо соотношение:
$$|r_{xy}| \leq 1$$

Случайные величины $X$ и $Y$ называют коррелированными, если их корреляционный момент или их коэффициент корреляции отличен от нуля. В противном случае эти величины некоррелированы. \
Если случайные величины $X$ и $Y$ коррелированы, то они зависимы.

Значение $\bar r_{xy}$ - статистической оценки $r_{xy}$ - коэффициента корреляции можно вычислить по формуле:
$$\bar r_{xy} = \frac{\sum\limits_{i=1}^{K_y}\sum\limits_{i=1}^{K_x} n_{ij} y_i x_j}{NS_xS_y}$$

При вычислении выборочного коэффициента корреляции удобно использовать условные варианты:
$$\bar r_{xy} = \frac{\sum\limits_{i=1}^{K_y}\sum\limits_{i=1}^{K_x} n_{ij} u_i v_j - N \bar u_{\text в} \bar v_{\text в} }{N \bar\sigma_u \bar\sigma_v}$$

С помощью преобразования Фишера можно перейти к случайно величине z:
$$\bar z = 0.5 \cdot ln \frac{1 + \bar r_{xy}}{1 - \bar r_{xy}}, (4.1)$$

Распределение $z$ при неограниченном возрастании объема выборки асимпототически нормальное со значением СКВО, равным:
$$\bar\sigma_z = \frac{1}{\sqrt{N-3}}, (4.2)$$

Доверительный интервал генеральное совокупности с доверительной вероятностью $\gamma$ определяют по следующей схеме:
1. По формуле (4.1) вычисляют выборочное значение $\bar z$;
2. По формуле (4.2) вычисляют значение $\bar\sigma_z$;
3. Доверительный интервал для генерального значения представляется в виде:
$$(\bar z - \lambda (\gamma) \bar\sigma_z, \bar z + \lambda (\gamma) \bar\sigma_z),$$
где значение $\lambda (\gamma)$ должно удовлетворять условию:
$$\Phi[\lambda (\gamma)] = \frac{\gamma}{2}$$
4. Для пересчета интервала в доверительный интервал для коэффициента корреляции с тем же значением $\gamma$ необходимо воспользоваться обратным преобразованием Фишера:
$$r = th(z)$$

Так как $\bar r_{xy}$ является случайной величиной, то это еще не значит что $r_{xy}$ - коэффициент корреляции для генеральной совокупности тоже отличен от нуля. \
Возникает необходимость проверить гипотезу $H_0: r_{xy} = 0$. \
Альтернативной будет гипотеза $H_1: r_{xy} \neq 0$. \
Если основная гипотеза $H_0$ отвергается, то это означает, что выборочный коэффициент корреляции $\bar r_{xy}$ значимо отличается от нуля (значим). В противном случае - $\bar r_{xy}$ незначим.

В качестве критерия проверки статистической гипотезы о значимости выборочного коэффициента корреляции можно принять случайную величину:
$$T = \frac{\bar r_{xy} \sqrt{N-2}}{\sqrt{1 - \bar r_{xy}^2}}$$

```python
freq = table.iloc[:-1, :-1].to_numpy(dtype=float)

x = (X[:-1] + X[1:]) / 2   # середины интервалов E
y = (Y[:-1] + Y[1:]) / 2   # середины интервалов nu

n = freq.sum()

# частоты по X и Y
n_x = freq.sum(axis=0)
n_y = freq.sum(axis=1)

# выборочные средние
x_bar = (n_x * x).sum() / n
y_bar = (n_y * y).sum() / n

# вторые моменты
x2_bar = (n_x * x**2).sum() / n
y2_bar = (n_y * y**2).sum() / n

# смешанный момент
xy_bar = (freq * np.outer(y, x)).sum() / n

# дисперсии и ковариация
D_x = x2_bar - x_bar**2
D_y = y2_bar - y_bar**2
cov_xy = xy_bar - x_bar * y_bar

r_standard = cov_xy / np.sqrt(D_x * D_y)

print("Способ 1: стандартная формула")
display(Math(rf"\bar x = {round(x_bar, 6)}"))
display(Math(rf"\bar y = {round(y_bar, 6)}"))
display(Math(rf"D_x = {round(D_x, 6)}"))
display(Math(rf"D_y = {round(D_y, 6)}"))
display(Math(rf"cov_{{xy}} = {round(cov_xy, 6)}"))
display(Math(rf"r = {round(r_standard, 6)}"))


# Коэффициент корреляции через условные варианты

h_x = x[1] - x[0]
h_y = y[1] - y[0]

# условные начала отсчёта
A_x = x[len(x)//2]
A_y = y[len(y)//2]

u = (x - A_x) / h_x
v = (y - A_y) / h_y

u_bar = (n_x * u).sum() / n
v_bar = (n_y * v).sum() / n

u2_bar = (n_x * u**2).sum() / n
v2_bar = (n_y * v**2).sum() / n

uv_bar = (freq * np.outer(v, u)).sum() / n

r_cond = (uv_bar - u_bar * v_bar) / np.sqrt((u2_bar - u_bar**2) * (v2_bar - v_bar**2))

print("\nСпособ 2: условные варианты")
display(Math(rf"A_x = {A_x} \quad h_x = {round(h_x, 6)}"))
display(Math(rf"A_y = {A_y} \quad h_y = {round(h_y, 6)}"))
display(Math(rf"\bar u = {round(u_bar, 6)}"))
display(Math(rf"\bar v = {round(v_bar, 6)}"))
display(Math(rf"r = {round(r_cond, 6)}"))

print("\nПроверка:")
print("Разница |r1 - r2| =", abs(r_standard - r_cond))
```

Результаты, найденные двумя способами, совпадают, следовательно, расчёты выполнены верно. Значение коэффициента корреляции $r \approx 0.877$ свидетельствует о сильной положительной линейной связи между признаками
$E$ и $nu$. Это означает, что при увеличении одного признака другой также имеет выраженную тенденцию к увеличению.

```python
from scipy.stats import norm, t

r = r_standard

gammas = [0.95, 0.99]
rows = []

z = np.arctanh(r)
se = 1 / np.sqrt(N - 3)

for gamma in gammas:
    u = norm.ppf((1 + gamma) / 2)

    z_low = z - u * se
    z_high = z + u * se

    rho_low = np.tanh(z_low)
    rho_high = np.tanh(z_high)

    rows.append({
        "γ": gamma,
        "u": round(u, 4),
        "Нижняя граница": rho_low,
        "Верхняя граница": rho_high
    })

ci_df = pd.DataFrame(rows)
ci_df["Нижняя граница"] = ci_df["Нижняя граница"].round(4)
ci_df["Верхняя граница"] = ci_df["Верхняя граница"].round(4)

display(ci_df)
```

В обоих случаях доверительные интервалы полностью лежат в области положительных значений и не содержат ноль. Это подтверждает наличие устойчивой положительной связи между признаками. При увеличении доверительной вероятности интервал становится шире, что соответствует теории.

```python
alpha = 0.05

t_obs = r * np.sqrt(n - 2) / np.sqrt(1 - r**2)
t_crit = t.ppf(1 - alpha/2, df=n-2)

display(Math(rf"r = {round(r, 6)}"))
display(Math(rf"t_{{obs}} = {round(t_obs, 6)}"))
display(Math(rf"t_{{crit}} = {round(t_crit, 6)}"))

if abs(t_obs) > t_crit:
    print("Гипотеза H0: r = 0 отвергается.")
    print("Коэффициент корреляции статистически значим.")
else:
    print("Нет оснований отвергнуть гипотезу H0")
    print("Статистически значимая корреляция не подтверждена.")
```

Нулевая гипотеза о равенстве коэффициента корреляции нулю отвергается. Следовательно, коэффициент корреляции статистически значим и линейная связь между признаками $E$ и $nu$ действительно существует. Причём, с учётом ранее полученного значения $r \approx 0.877$, эта связь является сильной и положительной.
