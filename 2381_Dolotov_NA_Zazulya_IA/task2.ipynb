## Основные теоретические положения

Условными называют варианты, определяемые равенством
$$u_j = \frac{x_j - C}{h}$$
где C - ложный ноль (новое начало отсчета), значение которого выбирается равным значению варианты интервального ряда, являющейся средней или близкой к средней по значению в этом ряду, h - ширина интервала.

Начальным эмпирическим моментом k-го порядка называется среднее значение k-х степеней элементов вариационного или интервального ряда:
$$\bar M_k = \frac{1}{N} \sum n_j x_j^k$$

В частности выборочное среднее:
$$\bar x_{\text{в}} = \bar M_1 = \frac{1}{N} \sum n_j x_j$$

Центральным эмпирическим моментом k-го порядка называется среднее значение k-х степеней разностей $(x_j - \bar x)$ для вариационного или интервального ряда:
$$\bar m_k = \frac{1}{N} \sum n_j (x_j - \bar x_{\text{в}})^k$$

В частности, дисперсия:
$$D_{\text{в}} = \bar m_2 = \frac{1}{N} \sum n_j (x_j - \bar x_{\text{в}})^2$$

Условные моменты k-го порядка:
$$\bar M_k^*=\frac{1}{N} \sum n_j \left(\frac{x_j - C}{h} \right)^k = \frac{1}{N} \sum n_j \cdot u_j^k$$

Исправленная оценка дисперсии определяется как:
$$s^2 = \frac{N}{N-1} D_{\text{в}}$$

Выборочное СКО определятеся как:
$$\sigma_{\text{в}} = \sqrt{D_{\text{в}}}$$

Исправленное СКО определяется как:
$$s = \sqrt{s^2}$$

Статистические оценки асимиетрии и эксцесса вычисляются по формулам:
$$\bar A_s = \frac{\bar m_3}{s^3}; \bar E = \frac{\bar m_4}{s^4} - 3$$

Модой $M_o^{*}$ вариационного ряда называется варианта, которой соответствует наибольшая частота. Для интервального ряда мода находится внутри интервала $(x_{i-1}, x_i)$, которому соответствует наибольшая частота и ее приближенное значение может быть вычислено по формуле:
$$M_o^{*} = x_{i-1} + h \frac{\tilde{m}_i - \tilde{m}_{i-1}}{(\tilde{m}_i - \tilde{m}_{i-1}) + (\tilde{m}_i - \tilde{m}_{i+1})}$$

Медианой $M_e^{*}$ называется варианта, соответствующая середине ранжированного ряда. Для интервального ряда выборочное значение медианы может быть вычислено по формуле:
$$M_e^{*} = x_{i-1} + \frac{h}{\tilde{m}_i}\left(0.5 - \sum_{j=1}^{i-1} \tilde{m}_j\right)$$

Выборочный коэффициент вариации $V^{*}$ характеризует разброс значений вариант относительно выборочного среднего:
$$V^{*} = \frac{\sigma_{\text{в}}}{|\bar x_{\text{в}}|} \cdot 100 \%$$

### Задание 1

```python
C = 466.06
h = 34.88
N = 104

x_mids = np.array([361.44, 396.31, 431.19, 466.06, 500.94, 535.81, 570.69, 605.56], dtype=float)
n = np.array([3, 17, 29, 22, 20, 8, 3, 2], dtype=float)
u = np.array([(x_mid - C) / h for x_mid in x_mids], dtype=float)


rows = []

for i in range(len(u)):
    rows.append({
        "x_i": x_mids[i],
        "n_i": n[i],
        "u_i": round(u[i]),
        "n_i*u_i": n[i] * round(u[i]),
        "n_i*u_i^2": n[i] * round(u[i])**2,
        "n_i*u_i^3": n[i] * round(u[i])**3,
        "n_i*u_i^4": n[i] * round(u[i])**4,
        "n_i*(u_i+1)^4": n[i] * (round(u[i]) + 1)**4,
    })

df = pd.DataFrame(rows)
df.index = range(1, len(df) + 1)
df.loc["Σ"] = {
    "x_i": "-",
    "n_i": df["n_i"].sum(),
    "u_i": "-",
    "n_i*u_i": df["n_i*u_i"].sum(),
    "n_i*u_i^2": df["n_i*u_i^2"].sum(),
    "n_i*u_i^3": df["n_i*u_i^3"].sum(),
    "n_i*u_i^4": df["n_i*u_i^4"].sum(),
    "n_i*(u_i+1)^4": df["n_i*(u_i+1)^4"].sum(),
}

df
```

Проведем контроль вычислений по формуле: \
$N + 4\sum n_iu_i + 6\sum n_iu_i^2 + 4\sum n_iu_i^3 + \sum n_iu_i^4 = \sum n_i(u_i + 1)^4$

```python
sum_n = df.loc["Σ", 'n_i']
sum_nu = df.loc["Σ", 'n_i*u_i']
sum_nu2 = df.loc["Σ", 'n_i*u_i^2']
sum_nu3 = df.loc["Σ", 'n_i*u_i^3']
sum_nu4 = df.loc["Σ", 'n_i*u_i^4']
sum_nu_1_4 = df.loc["Σ", 'n_i*(u_i+1)^4']

final = sum_n + 4*sum_nu + 6*sum_nu2 + 4*sum_nu3 + sum_nu4
print(f"{sum_n} + 4*{sum_nu} + 6*{sum_nu2} + 4*{sum_nu3} + {sum_nu4} = {final}")
```

### Задание 3

Выведем формулы через условные варианты:

$x_j = u_j \cdot h + C$ \
$\bar x_{\text{в}} = \frac{1}{N} \sum n_j x_j = \frac{1}{N} \sum n_j (u_j \cdot h + C) = C + \frac{h}{N} \sum n_j u_j$

$D_{\text{в}} = \frac{1}{N} \sum n_j (x_j - \bar x_{\text{в}})^2 = \frac{1}{N} \sum n_j \left( h (u_j - \bar u)  \right)^2 = h^2 \left( \frac{1}{N} \sum n_j (u_j - \bar u)^2 \right)$


```python
xv_std = 1 / N * np.sum(n * x_mids)
u_sum = 1 / N * np.sum(n * u)
xv_cnd = C + h * u_sum

D_std = np.sum(n * (x_mids - xv_std)**2) / N
D_cnd = h**2 / N * np.sum(n * (u - u_sum)**2)

display(Math(rf"\bar{{x}}_в = {xv_std}"))
display(Math(rf"\bar{{x}}_в = {xv_cnd}"))
display(Math(rf"D_в = {D_std}"))
display(Math(rf"D_в = {D_cnd}"))

sigma_v = np.sqrt(D_std)
display(Math(rf"\sigma_в = {sigma_v}"))
```

### Задание 4

```python
s_2 = N / (N - 1) * D_std
s = np.sqrt(s_2)

display(Math(rf"s^2 = {s_2}"))
display(Math(rf"s = {s}"))
```

Исправленная выборочная дисперсия получилась немного больше смещённой, потому что при расчёте смещённой дисперсии деление выполняется на $N$, из-за чего оценка разброса слегка занижается, тогда как деление на $N-1$ вводит поправку и компенсирует это занижение, делая оценку ближе к разбросу в генеральной совокупности; аналогично, смещённое стандартное отклонение чуть меньше исправленного, но разница между ними совсем мала, поскольку СКО получается извлечением корня из дисперсии и корень дополнительно сглаживает различие.
