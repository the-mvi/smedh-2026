## Основные теоретические положения

Начальный эмпирический момент k-го порядка:

$\bar{M}_k = \frac{1}{N}\sum n_j x_j^k$

Центральный эмпирический момент k-го порядка:

$\bar{m}_k = \frac{1}{N}\sum n_j (x_j - \bar{x}_в)^k,$
где $\bar{x}_в$ = $\bar{M}_1$.

Выборочная дисперсия:

$D_в = \bar{m}_2 = \frac{1}{N}\sum n_j(x_j - \bar{x}_в)^2$

Исправленная дисперсия:

$s^2 = \frac{N}{N-1}D_в$

Выборочное СКО:

$\sigma_в = \sqrt{D_в}$

Исправленное СКО:

$s = \sqrt{s^2}$

Статистическая оценка асимметрии:

$\bar{A}_s = \frac{\bar{m}_3}{s^3}$

Статистическая оценка эксцесса:

$\bar{E} = \frac{\bar{m}_4}{s^4} - 3$

Условная варианта:

$u_j = \frac{x_j - C}{h},$
где $C$ − условный ноль, значение которого выбирается равным значению варианты интервального ряда, являющейся средней или близкой к средней по значению в этом ряду.

Условный момент k-го порядка:

$\bar{M}_k^* = \frac{1}{N}\sum n_j\left( \frac{x_j-C}{h} \right)^k = \frac{1}{N}\sum n_j u_j^k$

Связь между условными и центральными моментами:

$\bar{m}_1 = 0$

$\bar{m}_2 = h^2(\bar{M}_2^* - (\bar{M}_1^*)^2)$

$\bar{m}_3 = h^3(\bar{M}_3^* - 3\bar{M}_2^*\bar{M}_1^* + 2(\bar{M}_1^*)^3)$

$\bar{m}_4 = h^4(\bar{M}_4^* - 4\bar{M}_3^*\bar{M}_1^* + 6\bar{M}_2^*(\bar{M}_1^*)^2 - 3(\bar{M}_1^*)^4)$

Выборочное среднее через условные варианты:

$\bar{x}_в = C + h \cdot \bar{M}_1^*$

Выборочная дисперсия через условные варианты:

$D_в = h^2(\bar{M}_2^* - (\bar{M}_1^*)^2)$

Модой $M_o^*$ вариационного ряда называется варианта, которой соответствует наибольшая частота.

Формула для интервального ряда:

$M_o^* = x_{i-1} + h \cdot \frac{\tilde{m}_i - \tilde{m}_{i-1}}{(\tilde{m}_i - \tilde{m}_{i-1})+(\tilde{m}_i - \tilde{m}_{i+1})},$
где $h$ − длина интервалов; $\tilde{m}_i$ − относительная частота i-го интервала.

Медианой $M_e^*$ называется варианта, соответствующая середине ранжированного ряда.

Формула для интервального ряда:

$M_e^* = x_{i-1} + h \cdot \frac{0.5 - \sum_{j=1}^{i-1} \tilde{m}_j}{\tilde{m}_i}$

Выборочный коэффициент вариации $V^*$ характеризует разброс значений вариант относительно выборочного среднего:

$V^* = \frac{\sigma_в}{\bar{x}_в} \cdot 100\%$

## Задание 1

Для заданного интервального ряда необходимо составить таблицу, содержащую условные варианты и их произведения с частотами. Условный ноль $C$ выбирается равным середине интервала с наибольшей частотой. В данном случае это интервал с серединой $x_i = 9.78$ и частотой $n_i = 32$.

Условная варианта вычисляется по формуле:
$u_i = \frac{x_i - C}{h}$

где $h = 0.55$ — ширина интервала.

Для контроля вычислений используется тождество:
$\sum n_j (u_j + 1)^4 = \sum n_j u_j^4 + 4\sum n_j u_j^3 + 6\sum n_j u_j^2 + 4\sum n_j u_j + \sum n_j$

```python
import pandas as pd
import numpy as np

# Данные из условия
x_i = np.array([8.68, 9.23, 9.78, 10.32, 10.88, 11.42, 11.98, 12.52])
n_i = np.array([1, 22, 32, 18, 10, 9, 4, 6])
h = 0.55
C = 9.78   # ложный нуль (середина интервала с наибольшей частотой)
n = np.sum(n_i)

# Вычисление условных вариант u_i
u_i = (x_i - C) / h

# Заполнение столбцов таблицы 1
n_u = n_i * u_i
n_u2 = n_i * u_i**2
n_u3 = n_i * u_i**3
n_u4 = n_i * u_i**4
n_u_plus1_4 = n_i * (u_i + 1)**4

# Суммы по столбцам
sum_n = np.sum(n_i)
sum_n_u = np.sum(n_u)
sum_n_u2 = np.sum(n_u2)
sum_n_u3 = np.sum(n_u3)
sum_n_u4 = np.sum(n_u4)
sum_n_u_plus1_4 = np.sum(n_u_plus1_4)

# Создаём DataFrame для таблицы 1
table1 = pd.DataFrame({
    'i': range(1, len(x_i)+1),
    'x_i': x_i,
    'n_i': n_i,
    'u_i': u_i,
    'n_i u_i': n_u,
    'n_i u_i^2': n_u2,
    'n_i u_i^3': n_u3,
    'n_i u_i^4': n_u4,
    'n_i (u_i+1)^4': n_u_plus1_4
})

# Добавляем строку с суммами
sum_row = pd.DataFrame({
    'i': ['Σ'],
    'x_i': ['-'],
    'n_i': [sum_n],
    'u_i': ['-'],
    'n_i u_i': [sum_n_u],
    'n_i u_i^2': [sum_n_u2],
    'n_i u_i^3': [sum_n_u3],
    'n_i u_i^4': [sum_n_u4],
    'n_i (u_i+1)^4': [sum_n_u_plus1_4]
})
table1 = pd.concat([table1, sum_row], ignore_index=True)

print("Таблица №1")
print(table1.to_string(index=False))

# Контроль вычислений: проверяем тождество Σ n_i (u_i+1)^4 = Σ n_i u_i^4 + 4Σ n_i u_i^3 + 6Σ n_i u_i^2 + 4Σ n_i u_i + Σ n_i
right_side = sum_n_u4 + 4*sum_n_u3 + 6*sum_n_u2 + 4*sum_n_u + sum_n
print(f"\nКонтроль: Σ n_i (u_i+1)^4 = {sum_n_u_plus1_4:.6f}")
print(f"Правая часть тождества = {right_side:.6f}")
print(f"Разность: {abs(sum_n_u_plus1_4 - right_side):.2e}")
if np.isclose(sum_n_u_plus1_4, right_side):
    print("Контроль пройден: равенство выполняется.\n")
else:
    print("Контроль не пройден! Ошибка в вычислениях.\n")
```

Контрольная сумма, полученная методом произведений, равна результату по формуле: $\sum n_j (u_j + 1)^4 = \sum n_j u_j^4 + 4 \sum n_j u_j^3 + 6 \sum n_j u_j^2 + 4 \sum n_j u_j + \sum n_j$. Равенство выполняется, что подтверждает корректность вычислений.

## Задание 2

На основе вычисленных условных моментов определяются центральные эмпирические моменты. Центральный момент первого порядка всегда равен нулю по определению.

Вывод формулы для центрального эмпирического момента 1-го порядка:
$\bar m_1 = \frac{1}{N} \sum n_j (x_j - \bar x_в) = \frac{1}{N} \sum n_j x_j - \frac{1}{N} \sum n_j \bar x_в = \bar x_в - \bar x_в \cdot \frac{1}{N} \sum n_j = \bar x_в - \bar x_в \cdot 1 = 0$

Центральные моменты высших порядков вычисляются через условные моменты с использованием формул преобразования.

```python
M1_star = sum_n_u / n
M2_star = sum_n_u2 / n
M3_star = sum_n_u3 / n
M4_star = sum_n_u4 / n

# Центральные эмпирические моменты m_k (через условные моменты)
m1 = 0.0   # центральный момент первого порядка всегда равен нулю
m2 = h**2 * (M2_star - M1_star**2)
m3 = h**3 * (M3_star - 3*M2_star*M1_star + 2*M1_star**3)
m4 = h**4 * (M4_star - 4*M3_star*M1_star + 6*M2_star*M1_star**2 - 3*M1_star**4)

# Таблица 2
table2 = pd.DataFrame({
    'k': [1, 2, 3, 4],
    'M_k*': [M1_star, M2_star, M3_star, M4_star],
    'm_k': [m1, m2, m3, m4]
})

print("Таблица №2")
print(table2.to_string(index=False))
print()
```

На основе условных моментов были вычислены центральные эмпирические моменты различных порядков. Центральный момент первого порядка равен нулю, что подтверждает корректность вычислений. Моменты высших порядков используются для характеристики формы распределения.

## Задание 3

Вычисление выборочных среднего и дисперсии двумя способами: стандартным методом и через условные варианты.

Вывод формул выборочных среднего и дисперсии через условные варианты:

$\bar x_в = \bar M_1^* h + C = h \cdot \frac{1}{N} \sum n_j u_j + C$

$ D_в = \frac{1}{N} \sum n_j (x_j - \bar x_в)^2 = \frac{1}{N} \sum n_j \left( (C + h u_j) - \left( C + h \cdot \frac{1}{N} \sum n_j u_j \right) \right)^2 = \frac{1}{N} \sum n_j \left( h u_j - h \cdot \frac{1}{N} \sum n_j u_j \right)^2 = h^2 \cdot \frac{1}{N} \sum n_j \left( u_j - \frac{1}{N} \sum n_j u_j \right)^2 = h^2 \cdot \frac{1}{N} \sum n_j \left( u_j^2 - 2u_j \frac{1}{N} \sum n_j u_j + \left( \frac{1}{N} \sum n_j u_j \right)^2 \right) = h^2 \left( \frac{1}{N} \sum n_j u_j^2 - 2 \left( \frac{1}{N} \sum n_j u_j \right)^2 + \left( \frac{1}{N} \sum n_j u_j \right)^2 \right) = h^2 \left( \frac{1}{N} \sum n_j u_j^2 - \left( \frac{1}{N} \sum n_j u_j \right)^2 \right)$

```python
x_bar_std = np.sum(n_i * x_i) / n
D_std = np.sum(n_i * (x_i - x_bar_std)**2) / n
sigma_std = np.sqrt(D_std)

# Вычисление через условные варианты
x_bar_cond = C + h * M1_star
D_cond = h**2 * (M2_star - M1_star**2)
sigma_cond = np.sqrt(D_cond)

print("Пункт 3. Сравнение стандартного метода и метода условных вариантов:")
print(f"Среднее (стандартное):          {x_bar_std:.6f}")
print(f"Среднее (условные варианты):    {x_bar_cond:.6f}")
print(f"Дисперсия (стандартная):         {D_std:.6f}")
print(f"Дисперсия (условные варианты):   {D_cond:.6f}")
print(f"СКО (стандартное):               {sigma_std:.6f}")
print(f"СКО (условные варианты):         {sigma_cond:.6f}")
print(f"Разница средних: {abs(x_bar_std - x_bar_cond):.2e}")
print(f"Разница дисперсий: {abs(D_std - D_cond):.2e}")
print("Результаты совпадают в пределах погрешности вычислений.\n")
```

Значения выборочных среднего и дисперсии, вычисленные с помощью стандартных формул и условных вариант, практически равны. Небольшие различия связаны с ошибками округления при записи промежуточных результатов.

## Задание 4

Исправленная выборочная дисперсия используется для компенсации систематической ошибки, возникающей при использовании смещённой оценки дисперсии для малых выборок.

```python
s2 = n / (n - 1) * D_cond   # исправленная дисперсия
s = np.sqrt(s2)

print("Пункт 4. Исправленные оценки:")
print(f"Исправленная дисперсия s^2 = {s2:.6f}")
print(f"Исправленное СКО s = {s:.6f}")
print(f"Смещённая дисперсия D = {D_cond:.6f}")
print(f"Смещённое СКО σ = {sigma_cond:.6f}")
print(f"Отношение s^2 / D = {s2/D_cond:.6f} (должно быть n/(n-1) = {n/(n-1):.6f})")
print("Вывод: исправленные оценки немного больше смещённых, теоретически верно. Объема выборки недостаточно\n")
```

Исправленная выборочная дисперсия получилась немного больше смещённой. Это связано с тем, что при вычислении смещённой дисперсии деление выполняется на $N$, следовательно, разброс данных немного уменьшается. Деление на $N-1$ позволяет компенсировать возникающее занижение оценки, поэтому исправленная дисперсия получается немного больше и точнее отражает разброс данных в генеральной совокупности.

Аналогичная ситуация наблюдается и для стандартных отклонений: смещённое СКО немного меньше исправленного. Однако различие между ними очень мало, так как отклонение вычисляется на основе дисперсии и корень смягчает эту разницу.

## Задание 5

Коэффициент вариации характеризует относительную изменчивость признака. Асимметрия и эксцесс описывают форму распределения относительно нормального распределения.

```python
# Блок 5: Коэффициент вариации, асимметрия и эксцесс

V = sigma_cond / abs(x_bar_cond) * 100   # в процентах
As = m3 / sigma_cond**3
Ek = m4 / sigma_cond**4 - 3

print("Пункт 5. Статистические оценки формы распределения:")
print(f"Коэффициент вариации V = {V:.2f}%")
print(f"Асимметрия A_s = {As:.6f}")
print(f"Эксцесс E_k = {Ek:.6f}")
print("Выводы:")
if V < 10:
    print("- Коэффициент вариации менее 10%, совокупность считается однородной.")
elif V < 33:
    print("- Коэффициент вариации в пределах 10-33%, совокупность умеренно неоднородна.")
else:
    print("- Коэффициент вариации более 33%, совокупность неоднородна.")
if As > 0:
    print("- Положительная асимметрия указывает на правостороннюю скошенность (хвост справа).")
elif As < 0:
    print("- Отрицательная асимметрия указывает на левостороннюю скошенность (хвост слева).")
else:
    print("- Асимметрия близка к нулю, распределение симметрично.")
if Ek > 0:
    print("- Положительный эксцесс говорит о более островершинном распределении по сравнению с нормальным.")
elif Ek < 0:
    print("- Отрицательный эксцесс говорит о более плосковершинном распределении.")
else:
    print("- Эксцесс близок к нулю, распределение по форме близко к нормальному.\n")
```

Положительное значение коэффициента асимметрии указывает на правую асимметрию распределения, то есть распределение слегка вытянуто вправо, а основная масса данных смещена немного левее центра. При этом форма распределения близка к симметричной, так как значение коэффициента близко к 0.

Отрицательное значение коэффициента эксцесса указывает на плосковершинность распределения по сравнению с нормальным. Его вершина менее выражена, а данные более равномерно распределены вокруг среднего, без ярко выраженного пика.

## Задание 6

Мода и медиана являются дополнительными характеристиками центра распределения. Мода показывает наиболее часто встречающееся значение, а медиана — значение, делящее выборку пополам.

```python
# Блок 6: Мода, медиана и коэффициент вариации (интервальный ряд)

# Мода
# Модальный интервал [9.5; 10.05) с частотой 32
x0_mo = 9.5          # нижняя граница модального интервала
f_mo = 32            # частота модального интервала
f_prev = 22          # частота предыдущего интервала
f_next = 18          # частота следующего интервала
Mo = x0_mo + h * (f_mo - f_prev) / ((f_mo - f_prev) + (f_mo - f_next))

# Медиана
# Медианный интервал [9.5; 10.05) (накопленная частота до него 23, после 55 > 51)
x0_me = 9.5
F_prev = 23          # накопленная частота до медианного интервала
f_me = 32            # частота медианного интервала
n_half = n / 2
Me = x0_me + h * (n_half - F_prev) / f_me

print("Пункт 6. Мода и медиана интервального ряда:")
print(f"Мода Mo = {Mo:.6f}")
print(f"Медиана Me = {Me:.6f}")
print(f"Коэффициент вариации (ранее вычислен) V = {V:.2f}%")
print("\nСравнение со средним арифметическим:")
print(f"Среднее x̄ = {x_bar_cond:.6f}")
print(f"Медиана Me = {Me:.6f}")
print(f"Мода Mo = {Mo:.6f}")
if Me < x_bar_cond:
    print("Медиана меньше среднего, что согласуется с положительной асимметрией.")
elif Me > x_bar_cond:
    print("Медиана больше среднего, что согласуется с отрицательной асимметрией.")
else:
    print("Медиана близка к среднему, распределение симметрично.")
print("Значение моды близко к среднему и медиане, что также характерно для умеренно асимметричных распределений.")
```

Значение моды меньше значения медианы, что говорит о том, что распределение имеет умеренный правосторонний уклон: наиболее часто встречающееся значение расположено чуть левее центра распределения.

Коэффициент вариации характеризует относительную изменчивость признака. Его полученное значение указывает на значительную степень рассеивания данных. Значение коэффициента вариации не превышает 33%, следовательно, совокупность является однородной.

Полученные значения позволили охарактеризовать форму распределения: выявлено наличие правосторонней асимметрии и плосковершинность по сравнению с нормальным распределением. Также были определены мода и медиана. Сопоставление этих параметров подтвердило наличие правосторонней асимметрии. Рассчитанный коэффициент вариации показал, что совокупность является однородной, а вариация признака — умеренной.
