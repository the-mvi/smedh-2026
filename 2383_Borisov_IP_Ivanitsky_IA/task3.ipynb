## Основные теоретические положения

Интервал $(\theta^* - \delta;\ \theta^* + \delta)$ называется доверительным интервалом и является интервальной оценкой параметра $\theta$.

Доверительный интервал для математического ожидания при неизвестном СКО:

$\left( \bar{x}_в - \frac{t_{\gamma} S}{\sqrt{N}}, \bar{x}_в + \frac{t_{\gamma} S}{\sqrt{N}} \right)$, где $ t_\gamma $ — коэффициент Стьюдента.

Доверительный интервал для математического ожидания при известном СКО:

$(S(1 - q), S(1 + q))$, где $ q $ — табличный коэффициент, зависящий от объема выборки $ N $ и надежности $ \gamma $.

Статистической гипотезой называют гипотезу о виде неизвестного распледеления исследуемой случайной величиный или о значении параметров известных распределений.

Выдвинутую гипотезу называют нулевой или основной гипотезой и называют $H_0$.

Проверку статистической гипотезы проводят для принятого уровня значимости $\alpha$

При проверке статистических гипотез на основе выборочных данных вычисляются значения этих случайных величин (наблюденные значения), которые сопоставляются с некоторыми, так называемыми, критическими (зависящими от заданного уровня значимости $\alpha$) значениями этих случайных величин.

Критерий Пирсона:

$\chi^2_{набл} = \sum_{i=1}^{K} \frac{(n_i - n_i')^2}{n_i'}$

Здесь $n_i'$ — теоретические частоты — частоты, с которыми нормально распределенная величина с параметрами $\bar{x}_в$ и $S$ попадала бы в $i$-й интервал интервального ряда при проведении $N$ испытаний.

Формула для вычисления теоретических частот:

$n_i' = N \cdot p_i$, где $i = 1, 2, 3, ..., K$

Формула для вычисления вероятности попадания случайной величины в каждый $i$-й интервал интервального ряда:

$p_i = \Phi (z_{i+1}) - \Phi (z_{i})$, где $i = 1, 2, 3, ..., K$

Формула для вычисления значений границ интервалов интервального ряда:

$z_i = \frac{x_i - \bar{x}_в}{S}$, где $i = 2, 3, ..., K$

Функция Лапласа:

$\Phi (z) = \frac{1}{\sqrt{2\pi}} \int_0^z e^{-\frac{t^2}{2}} dt$

Число степеней свободы:

$k = K - 3$

Областью принятия нулевой гипотезы называют совокупность значения критерия, при которых нулевую гипотезу принимают.

Критической областью называют совокупность значения критерия, при которых нулевую гипотезу отвергают.

Гипотеза о нормальном распределении принимается, если выполняется:

$ \chi^2_{набл} \leq \chi^2_{крит} $

## Задание 1
### Доверительный интервал для математического ожидания

```python
# Данные
N = 102
x_v = 10.245882
s = 0.952782

gamma_values = [0.95, 0.99]

# коэффициенты Стьюдента (df = 101)
t_values = [
    stats.t.ppf((1+0.95)/2, 101),
    stats.t.ppf((1+0.99)/2, 101)
]

delta_values = [round(t * s / m.sqrt(N), 6) for t in t_values]
border_values = [(round(x_v - d, 6), round(x_v + d, 6)) for d in delta_values]

for i in range(len(gamma_values)):
    print(f"γ = {gamma_values[i]}")
    print(f"δ = {delta_values[i]}")
    print(f"Доверительный интервал: ({border_values[i][0]}, {border_values[i][1]})\n")
```

В данном блоке производится построение доверительных интервалов
для математического ожидания генеральной совокупности.

Используется распределение Стьюдента, поскольку истинное значение
среднеквадратического отклонения генеральной совокупности неизвестно
и заменяется выборочной оценкой.

Полученные интервалы показывают диапазон значений, в котором
с заданной доверительной вероятностью может находиться
математическое ожидание исследуемой случайной величины.

При увеличении доверительной вероятности γ возрастает значение
критического коэффициента распределения Стьюдента,
что приводит к увеличению предельной ошибки δ
и расширению доверительного интервала.

## Задание 3
### Проверка гипотезы о нормальном распределении

```python
# Функция Лапласа
def Phi(z):
    return stats.norm.cdf(z) - 0.5


intervals = [
    {"left": 8.4, "right": 8.95, "n_i": 1},
    {"left": 8.95, "right": 9.5, "n_i": 22},
    {"left": 9.5, "right": 10.05, "n_i": 32},
    {"left": 10.05, "right": 10.6, "n_i": 18},
    {"left": 10.6, "right": 11.15, "n_i": 10},
    {"left": 11.15, "right": 11.7, "n_i": 9},
    {"left": 11.7, "right": 12.25, "n_i": 4},
    {"left": 12.25, "right": 12.8, "n_i": 6}
]


rows = []
table_data = []

for i, inter in enumerate(intervals):

    left = inter["left"]
    right = inter["right"]
    n_i = inter["n_i"]

    if i == 0:
        z_l = -np.inf
        phi_l = -0.5
    else:
        z_l = (left - x_v) / s
        phi_l = Phi(z_l)

    if i == len(intervals) - 1:
        z_r = np.inf
        phi_r = 0.5
    else:
        z_r = (right - x_v) / s
        phi_r = Phi(z_r)

    p_i = round(phi_r - phi_l, 4)
    n_i_theor = round(N * p_i, 2)

    rows.append({
        "interval": f"[{left}, {right})",
        "n_i": n_i,
        "p_i": p_i,
        "n_i_theor": n_i_theor
    })

    table_data.append([
        i+1,
        left,
        right,
        n_i,
        "-∞" if z_l == -np.inf else f"{z_l:.3f}",
        "+∞" if z_r == np.inf else f"{z_r:.3f}",
        f"{phi_l:.4f}",
        f"{phi_r:.4f}",
        p_i,
        n_i_theor
    ])


headers = ["i", "x_i", "x_i+1", "n_i", "z_i", "z_i+1", "Φ(z_i)", "Φ(z_i+1)", "p_i", "n_i'"]

print(tabulate(table_data, headers=headers, tablefmt="grid"))
```

Так как в 7 и 8 интервалах теоретическая частота меньше 5, следовательно, нужно объединить 7 и 8 интервалы.

```python
merged = rows[:-2]  # все интервалы кроме последних двух
# объединяем последние два
last1 = rows[-2].copy()
last2 = rows[-1].copy()
last1["interval"] = last1["interval"] + " + " + last2["interval"]
last1["n_i"] += last2["n_i"]
last1["p_i"] += last2["p_i"]
last1["n_i_theor"] += last2["n_i_theor"]
merged.append(last1)

# === Вывод итоговой таблицы ===
table_data = []
for i, row in enumerate(merged):
    table_data.append([
        i+1,
        row["interval"],
        row["n_i"],
        round(row["p_i"], 4),
        round(row["n_i_theor"], 2)
    ])

headers = ["i", "Интервал", "n_i (факт)", "p_i (теор.)", "n_i' (теор.)"]
print(tabulate(table_data, headers=headers, tablefmt="grid"))
```

После вычисления теоретических частот было проверено условие
применимости критерия Пирсона.

Если ожидаемая частота некоторого интервала меньше 5,
то такой интервал объединяется с соседним.

В результате объединения формируется новый набор интервалов,
для которых выполняется условие применимости критерия.

Полученные объединённые интервалы используются далее
для вычисления статистики критерия χ².

## Задание 4

Докажем, что

$\chi^2_{набл} = \sum_{i=1}^{k} \frac{n_i^2}{n'_i} - n$

Доказательство:

$\chi^2_{набл} = \sum_{i=1}^{k} \frac{(n_i - n'_i)^2}{n'_i} = \sum_{i=1}^{k} \frac{n_i^2 - 2n_i n'_i + (n'_i)^2}{n'_i} = \sum_{i=1}^{k} \frac{n_i^2}{n'_i} - 2\sum_{i=1}^{k} n_i + \sum_{i=1}^{k} n'_i = \sum_{i=1}^{k} \frac{n_i^2}{n'_i} - 2n + n = \sum_{i=1}^{k} \frac{n_i^2}{n'_i} - n$

```python
chi_table = []

sum_chi = 0
sum_n_sq_div = 0

for i, row in enumerate(merged, 1):

    n_i = row["n_i"]
    p_i = row["p_i"]
    n_i_theor = row["n_i_theor"]

    diff_sq = round((n_i - n_i_theor) ** 2, 4)
    chi = round(diff_sq / n_i_theor, 4)
    n_sq_div = round((n_i ** 2) / n_i_theor, 4)

    sum_chi += chi
    sum_n_sq_div += n_sq_div

    chi_table.append([
        i,
        row["interval"],
        n_i,
        p_i,
        n_i_theor,
        diff_sq,
        chi,
        n_i ** 2,
        n_sq_div
    ])


chi_table.append([
    "",
    "Σ",
    sum(r[2] for r in chi_table),
    sum(r[3] for r in chi_table),
    sum(r[4] for r in chi_table),
    "-",
    sum_chi,
    "-",
    sum_n_sq_div
])

headers = [
    "i",
    "[x_i, x_i+1)",
    "n_i",
    "p_i",
    "n_i'",
    "(n_i-n_i')^2",
    "(n_i-n_i')^2/n_i'",
    "n_i^2",
    "n_i^2/n_i'"
]

print(tabulate(chi_table, headers=headers, tablefmt="grid"))

print("\nНаблюдаемое значение χ² =", round(sum_chi,4))

chi_obs = round(sum_n_sq_div - N, 4)

print("Наблюдаемое значение по формуле χ² =", chi_obs)
```

Оба вычисленных наблюдаемых значения сошлись.

## Задание 5

```python
K = len(merged)

df = K - 3
alpha = 0.05

chi_crit = round(stats.chi2.ppf(1 - alpha, df), 4)

print("Число степеней свободы =", df)
print("Критическое значение χ² =", chi_crit)

if sum_chi <= chi_crit:
    print("Нулевая гипотеза принимается")
else:
    print("Нулевая гипотеза отвергается")
```

Так как наблюдаемое значение критерия $\chi^2_{\text{набл}}$ оказалось **больше критического** $\chi^2_{\text{кр}}$, нулевая гипотеза $H_0$ о нормальном распределении **отвергается** на уровне значимости $\alpha = 0{,}05$.

Это означает, что различия между эмпирическими и теоретическими частотами являются **статистически значимыми**. Следовательно, выборочные данные **не согласуются** с гипотезой о нормальном распределении исследуемой случайной величины.
