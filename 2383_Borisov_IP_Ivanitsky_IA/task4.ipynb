## Основные теоретические понятия

**Генеральная совокупность и выборка.**
Генеральной совокупностью называют множество всех возможных значений исследуемых признаков. Выборка — конечный набор наблюдений, полученный из генеральной совокупности.

**Одномерные статистические характеристики.**

- Выборочное среднее:
$\displaystyle \bar y = \frac{1}{N}\sum_{j=1}^k n_j y_j$

- Выборочная (смещённая) дисперсия:
$\displaystyle D_y = \frac{1}{N}\sum_{j=1}^k n_j (y_j - \bar y)^2$

- Исправленная дисперсия:
$\displaystyle s_y^2 = \frac{N}{N-1} D_y$

- Выборочное среднеквадратическое отклонение:
$\displaystyle \sigma_y = \sqrt{D_y}$

- Исправленное среднеквадратическое отклонение:
$\displaystyle s_y = \sqrt{s_y^2}$

- Коэффициент вариации:
$\displaystyle V_y = \frac{\sigma_y}{\bar y} \cdot 100\%$

- Коэффициент асимметрии:
$\displaystyle A_y = \frac{m_{3,y}}{s_y^3}$

- Коэффициент эксцесса:
$\displaystyle E_y = \frac{m_{4,y}}{s_y^4} - 3$

где $m_{k,y}$ — центральные эмпирические моменты:
$\displaystyle m_{k,y} = \frac{1}{N}\sum_{j=1}^k n_j (y_j - \bar y)^k$.

**Условные варианты.**

Для уменьшения вычислительных ошибок вводятся условные варианты:
$\displaystyle v_j = \frac{y_j - C_y}{h_y}$,
где $C_y$ — условный ноль (значение, близкое к центру ряда), $h_y$ — ширина интервалов.

Условные моменты:
$\displaystyle M_{k,y}^* = \frac{1}{N}\sum_{j=1}^k n_j v_j^k$.

Связь между условными и центральными моментами аналогична использованной в работе №2.

**Двумерный вариационный ряд и корреляция.**

Пусть $X$ — первый признак, $Y$ — второй признак. Для них строится двумерный интервальный вариационный ряд (корреляционная таблица) с частотами $n_{ij}$ — числом наблюдений, попавших в $i$-й интервал по $X$ и $j$-й интервал по $Y$.

Выборочные средние:
$\displaystyle \bar x = \frac{1}{N}\sum_{i,j} n_{ij} x_i$,\quad $\displaystyle \bar y = \frac{1}{N}\sum_{i,j} n_{ij} y_j$.

Выборочная ковариация:
$\displaystyle S_{xy} = \frac{1}{N}\sum_{i,j} n_{ij} (x_i - \bar x)(y_j - \bar y)$.

Выборочная дисперсия признаков:
$\displaystyle S_{xx} = \frac{1}{N}\sum_{i,j} n_{ij} (x_i - \bar x)^2$,\quad
$\displaystyle S_{yy} = \frac{1}{N}\sum_{i,j} n_{ij} (y_j - \bar y)^2$.

**Выборочный коэффициент корреляции:**
$\displaystyle r = \frac{S_{xy}}{\sqrt{S_{xx} S_{yy}}}$.

**Оценка коэффициента корреляции через условные варианты.**

При использовании условных вариантов $u_i$ для $X$ и $v_j$ для $Y$ ковариация и дисперсии выражаются через условные моменты, что уменьшает вычислительные погрешности (аналогично работе №2).

**Доверительный интервал для коэффициента корреляции.**

Для построения доверительного интервала используют преобразование Фишера:

$\displaystyle z = \frac{1}{2}\ln\frac{1+r}{1-r}$,

где $z$ приблизительно нормально распределён с дисперсией $\frac{1}{N-3}$.

Доверительный интервал уровня доверия $\gamma$ для истинного коэффициента корреляции $\rho$:

$z_{\min} = z - z_{\gamma} \cdot \frac{1}{\sqrt{N-3}}, \quad z_{\max} = z + z_{\gamma} \cdot \frac{1}{\sqrt{N-3}}$,

после чего значения $z_{\min}$ и $z_{\max}$ преобразуются обратно:

$\rho_{\min,\max} = \tanh(z_{\min,\max}).$

**Проверка гипотезы $H_0\!: \rho = 0$.**

При справедливости $H_0$ статистика

$\displaystyle t = r \sqrt{\frac{N-2}{1-r^2}}$

имеет распределение Стьюдента с $N-2$ степенями свободы. Если $|t| > t_{1-\alpha/2,\,N-2}$, то гипотеза $H_0$ отвергается, иначе оснований отвергать её нет.

Во всех последующих пунктах в качестве исходной двумерной генеральной совокупности используется файл `data.csv`, содержащий значения двух признаков: содержание алкоголя (`alcohol`) и плотность (`density`). В этой работе:

Из генеральной совокупности методом простого случайного бесповторного отбора формируется выборка фиксированного объёма (как в работах №1 и №2). Для воспроизводимости устанавливается параметр `random_state`.

Ниже выполняется загрузка данных и построение выборки.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from math import sqrt, log, tanh

# Загрузка исходных данных
data = pd.read_csv("./data.csv", sep=",")

# Формирование выборки заданного объёма (как в ПР1)
N = 102
sample = data.sample(n=N, random_state=26).reset_index(drop=True)

# Определяем признаки X и Y
X = sample["alcohol"].values
Y = sample["density"].values

print(f"Размер исходной таблицы: {len(data)} наблюдений")
print(f"Объём выборки: N = {N}")
print("Первые строки выборки:")
print(sample.head())
```

## Задание 2. Одномерный анализ второго признака (Y = density)

На этом шаге рассматривается распределение только второго признака $Y$ (плотность):
- строится ранжированный и вариационный ряды;
- формируется интервальный ряд с использованием формулы Стерджесса;
- по интервальному ряду вычисляются основные числовые характеристики; 
- результаты оформляются в виде таблиц и кратко интерпретируются.

```python
# Ранжированный и вариационный ряды для Y

# Ранжированный ряд
Y_sorted = np.sort(Y)
print("Ранжированный ряд (первые 20 значений):")
print(Y_sorted[:20])

# Вариационный ряд (уникальные значения и их частоты)
variation_Y = (
    pd.Series(Y)
    .value_counts()
    .sort_index()
    .reset_index()
)
variation_Y.columns = ["Значение Y", "Частота"]
print("\nВариационный ряд (фрагмент):")
print(variation_Y.head(10))

# Число интервалов по формуле Стерджесса
kY = round(1 + 3.31 * np.log10(N))
y_min, y_max = Y.min(), Y.max()
hY = (y_max - y_min) / kY

print(f"\nЧисло интервалов для Y: k = {kY}")
print(f"Минимум Y: {y_min:.5f}, максимум Y: {y_max:.5f}")
print(f"Шаг интервала h_Y ≈ {hY:.6f}")

# Построение интервального ряда
freqY, edgesY = np.histogram(Y, bins=kY)
midY = (edgesY[:-1] + edgesY[1:]) / 2
relY = freqY / N
cumY = np.cumsum(freqY)
relCumY = cumY / N

interval_table_Y = pd.DataFrame({
    "Интервал": [
        ("[%.5f; %.5f)" % (edgesY[i], edgesY[i+1])) if i < kY-1 else
        ("[%.5f; %.5f]" % (edgesY[i], edgesY[i+1])) for i in range(kY)
    ],
    "Середина": midY.round(5),
    "m_j": freqY,
    "w_j": relY.round(4),
    "m_нак": cumY,
    "w_нак": relCumY.round(4)
})

sum_row_Y = pd.DataFrame({
    "Интервал": ["Σ"],
    "Середина": ["-"],
    "m_j": [freqY.sum()],
    "w_j": [relY.sum().round(4)],
    "m_нак": ["-"],
    "w_нак": ["-"]
})

interval_table_Y = pd.concat([interval_table_Y, sum_row_Y], ignore_index=True)
print("\nИнтервальный ряд для Y:")
print(interval_table_Y)
```

```python
# Графическое представление распределения Y

plt.figure(figsize=(7, 4))
plt.plot(midY, freqY, marker="o")
plt.title("Полигон абсолютных частот для Y (density)")
plt.xlabel("Середины интервалов")
plt.ylabel("Частота")
plt.grid(True)
plt.show()

plt.figure(figsize=(7, 4))
plt.hist(Y, bins=kY)
plt.title("Гистограмма распределения Y")
plt.xlabel("Плотность")
plt.ylabel("Частота")
plt.grid(True)
plt.show()

sorted_Y = np.sort(Y)
Fn_Y = np.arange(1, N+1) / N
plt.figure(figsize=(7, 4))
plt.step(sorted_Y, Fn_Y, where="post")
plt.title("Эмпирическая функция распределения Y")
plt.xlabel("Плотность")
plt.ylabel("F*_Y(y)")
plt.grid(True)
plt.show()
```

```python
# Числовые характеристики Y (стандартные формулы)

# Выборочное среднее и дисперсия
mean_Y = np.mean(Y)
var_Y = np.mean((Y - mean_Y)**2)          # смещённая дисперсия
std_Y = np.sqrt(var_Y)

# Исправленные оценки
var_Y_unbiased = N / (N - 1) * var_Y
std_Y_unbiased = np.sqrt(var_Y_unbiased)

# Центральные моменты 3-го и 4-го порядков
m3_Y = np.mean((Y - mean_Y)**3)
m4_Y = np.mean((Y - mean_Y)**4)

# Асимметрия и эксцесс
A_Y = m3_Y / (std_Y_unbiased**3)
E_Y = m4_Y / (std_Y_unbiased**4) - 3

# Коэффициент вариации
V_Y = std_Y / mean_Y * 100

print("Основные числовые характеристики признака Y (плотность):")
print(f"Выборочное среднее:      ȳ = {mean_Y:.6f}")
print(f"Дисперсия (смещённая):   D_y = {var_Y:.8f}")
print(f"Дисперсия (исправленная): s_y^2 = {var_Y_unbiased:.8f}")
print(f"СКО (смещённое):         σ_y = {std_Y:.8f}")
print(f"СКО (исправленное):      s_y = {std_Y_unbiased:.8f}")
print(f"Асимметрия:              A_y = {A_Y:.4f}")
print(f"Эксцесс:                 E_y = {E_Y:.4f}")
print(f"Коэффициент вариации:    V_y = {V_Y:.2f}%")
```

Полученные характеристики позволяют описать распределение плотности: выборочное среднее задаёт типичный уровень признака, дисперсия и СКО отражают степень разброса значений, коэффициент вариации показывает относительную изменчивость, а асимметрия и эксцесс позволяют судить о форме распределения по сравнению с нормальным законом.

## Задание 3. Двумерный интервальный вариационный ряд и корреляционная таблица

Теперь одновременно рассматриваются два признака $X$ (alcohol) и $Y$ (density). Необходимо:
- разбить каждую ось на интервалы (по формуле Стерджесса);
- по получённой сетке интервалов построить двумерный интервальный вариационный ряд;
- представить результат в виде корреляционной таблицы частот $n_{ij}$.

```python
# Интервальное разбиение по X и Y и построение двумерного ряда

# Параметры разбиения для X
kX = round(1 + 3.31 * np.log10(N))
x_min, x_max = X.min(), X.max()
hX = (x_max - x_min) / kX

edgesX = np.linspace(x_min, x_max, kX + 1)
midX = (edgesX[:-1] + edgesX[1:]) / 2

print(f"Число интервалов для X: k = {kX}")
print(f"Шаг по X: h_X ≈ {hX:.6f}")

# Двумерная гистограмма частот n_ij
n_ij, x_edges, y_edges = np.histogram2d(X, Y, bins=[edgesX, edgesY])

n_ij = n_ij.astype(int)

corr_table = pd.DataFrame(n_ij,
                          index=["X_%d" % (i+1) for i in range(kX)],
                          columns=["Y_%d" % (j+1) for j in range(kY)])

print("Корреляционная таблица (фрагмент):")
print(corr_table.head())
```

## Задание 4. Коэффициент корреляции: стандартная формула и условные варианты

На основе двумерного вариационного ряда необходимо:
- вычислить выборочный коэффициент корреляции напрямую по паре $(X, Y)$;
- пересчитать те же величины через условные варианты по интервалам и убедиться, что результаты практически совпадают.

```python
# Коэффициент корреляции по стандартным формулам (по исходным данным)

x_bar = np.mean(X)
y_bar = np.mean(Y)

Sxx = np.mean((X - x_bar)**2)
Syy = np.mean((Y - y_bar)**2)
Sxy = np.mean((X - x_bar) * (Y - y_bar))

r_std = Sxy / np.sqrt(Sxx * Syy)

print("Коэффициент корреляции по исходным данным:")
print(f"r (standard) = {r_std:.6f}")

# Коэффициент корреляции через интервальный (условный) ряд

# центры интервалов по X и Y
cx = midX
cy = midY

# общая сумма частот в двумерном ряду
N_check = n_ij.sum()

# выборочные средние через интервальный ряд
x_bar_tab = (n_ij.sum(axis=1) * cx).sum() / N_check
y_bar_tab = (n_ij.sum(axis=0) * cy).sum() / N_check

# ковариация и дисперсии через интервальный ряд
Sxx_tab = (n_ij.sum(axis=1) * (cx - x_bar_tab)**2).sum() / N_check
Syy_tab = (n_ij.sum(axis=0) * (cy - y_bar_tab)**2).sum() / N_check

Sxy_tab = 0.0
for i in range(kX):
    for j in range(kY):
        Sxy_tab += n_ij[i, j] * (cx[i] - x_bar_tab) * (cy[j] - y_bar_tab)
Sxy_tab /= N_check

r_tab = Sxy_tab / np.sqrt(Sxx_tab * Syy_tab)

print("\nКоэффициент корреляции через интервальный ряд:")
print(f"x̄ (табличное) = {x_bar_tab:.6f},  ȳ (табличное) = {y_bar_tab:.6f}")
print(f"Sxx (табличное) = {Sxx_tab:.8f},  Syy (табличное) = {Syy_tab:.8f}")
print(f"Sxy (табличное) = {Sxy_tab:.8f}")
print(f"r (table-based) = {r_tab:.6f}")
print(f"Разность |r_std - r_tab| = {abs(r_std - r_tab):.2e}")
```

Практически совпадающие значения коэффициента корреляции, полученные по исходным данным и по интервальному ряду, подтверждают корректность построения корреляционной таблицы и вычислений по условным вариантам.

## Задание 5. Доверительный интервал для коэффициента корреляции

Используя преобразование Фишера, построим доверительные интервалы для истинного коэффициента корреляции $\rho$ при уровнях доверия $\gamma = 0.95$ и $\gamma = 0.99$.

```python
# Доверительные интервалы для коэффициента корреляции (преобразование Фишера)

z = 0.5 * np.log((1 + r_std) / (1 - r_std))
se_z = 1 / np.sqrt(N - 3)

levels = [0.95, 0.99]
# Критические значения стандартного нормального распределения (приближённо)
z_crits = {0.95: 1.96, 0.99: 2.576}

for gamma in levels:
    z_crit = z_crits[gamma]
    z_min = z - z_crit * se_z
    z_max = z + z_crit * se_z
    rho_min = tanh(z_min)
    rho_max = tanh(z_max)
    print(f"Уровень доверия γ = {gamma}:")
    print(f"  Доверительный интервал для ρ: ({rho_min:.4f}; {rho_max:.4f})")
    print()
```

Полученные интервалы показывают диапазон возможных значений истинного коэффициента корреляции $\rho$, совместимых с наблюдаемыми данными на заданных уровнях доверия.

## Задание 6. Проверка гипотезы о равенстве коэффициента корреляции нулю

Необходимо проверить гипотезу

$H_0\!: \rho = 0$ против альтернативы $H_1\!: \rho \neq 0$ при уровне значимости $\alpha = 0.05$.

В качестве критерия используется $t$-статистика:

$\displaystyle t = r \sqrt{\frac{N-2}{1-r^2}}$.

Если $|t| > t_{1-\alpha/2,\,N-2}$, то гипотеза $H_0$ отвергается.

```python
# Проверка гипотезы H0: ρ = 0 при α = 0.05

alpha = 0.05

t_stat = r_std * np.sqrt((N - 2) / (1 - r_std**2))
# Приближённое критическое значение t_{0.975, N-2} для N≈100
t_crit = 1.984  # можно уточнить по таблице распределения Стьюдента

print(f"t-статистика: t = {t_stat:.4f}")
print(f"Критическое значение t_crit ≈ {t_crit:.3f}")

if abs(t_stat) > t_crit:
    print("Так как |t| > t_crit, гипотеза H0: ρ = 0 отвергается.")
    print("Существует статистически значимая линейная связь между признаками X и Y.")
else:
    print("Так как |t| ≤ t_crit, оснований отвергать гипотезу H0 нет.")
    print("Данных недостаточно, чтобы утверждать о наличии линейной связи.")
```
