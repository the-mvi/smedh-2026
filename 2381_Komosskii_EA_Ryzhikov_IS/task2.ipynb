## Основные теоретические положения

**Начальный эмпирический момент**:
$$
\bar M_k=\frac{1}{N}\sum n_j\,x_j^{\,k}
$$

**Центральный эмпирический момент**:
$$
\bar m_k=\frac{1}{N}\sum n_j\,(x_j-\bar x_в)^k,
где\ \bar{x}_в = \bar{M}_1.
$$

**Выборочная дисперсия**:
$$
D_в=\bar m_2=\frac{1}{N}\sum n_j\,(x_j-\bar x_в)^2
$$

**Выборочное СКО**:
$$
\sigma_в = \sqrt{D_в}
$$

**Исправленная выборочная дисперсия**:
$$
s^2=\frac{N}{N-1}\,D_в
$$

**Исправленное СКО**:
$$
s=\sqrt{s^2}
$$

**Статистическая оценка асимметрии**:
$$
\bar A_s=\frac{\bar m_3}{s^3}
$$

**Статистическая оценка эксцесса**:
$$
\bar E=\frac{\bar m_4}{s^4}-3
$$

**Условная варианта**:
$$u_j = \frac{x_j - C}{h},$$
где $C$ − условный ноль, значение которого выбирается равным значению варианты интервального ряда, являющейся средней или близкой к средней по значению в этом ряду.

**Условный момент k-го порядка**:
$$
\bar{M}_k^* = \frac{1}{N}\sum n_j\left( \frac{x_j-C}{h} \right)^k = \frac{1}{N}\sum n_j u_j^k
$$

**Мода вариационного ряда** - варианта, которой соответствует наибольшая частота:
$$
M_o^* = x_{i-1} + h \cdot \frac{\tilde{m}_i - \tilde{m}_{i-1}}{(\tilde{m}_i - \tilde{m}_{i-1})+(\tilde{m}_i - \tilde{m}_{i+1})},
$$
где $h$ − длина интервалов; $\tilde{m}_i$ − относительная частота i-го интервала.

**Медиана** - варианта, соответсвующая середине ранжированного ряда:
$$
M_e^* = x_{i-1} + \frac{h}{\tilde{m}_i} \left( 0.5 - \sum_{j=1}^{i-1} \tilde{m}_j \right)
$$

**Выборочный коэффициент вариации** - характеризует разброс значений вариант относительно выборочного среднего:
$$
V^* = \frac{\sigma_в}{\bar x_в} \cdot 100\%
$$

```python

import numpy as np
import pandas as pd
import math


a = np.array([0.03, 0.053125, 0.07625, 0.099375, 0.1225, 0.145625, 0.16875, 0.191875])    # нижние границы
b = np.array([0.053125, 0.07625, 0.099375, 0.1225, 0.145625, 0.16875, 0.191875, 0.215])   # верхние границы
n_i = np.array([2, 2, 16, 12, 29, 19, 21, 7])                                             # частоты
N = int(n_i.sum())
N
```

## Шаг 1. Середины интервалов и условные варианты
Берём середины $x_i = (a_i+b_i)/2$. Ширина интервала $h$ (если ряд равномерный, то одинаковая).

Условные варианты:
$$u_i=\frac{x_i-C}{h},$$
где $C$ — условное начало отсчёта.

```python

# Середины интервалов
x_i = (a + b) / 2

# Шаг (ширина интервала)
h = float((b - a)[0])

# Условное начало отсчёта A: по умолчанию возьмём середину интервала с максимальной частотой
idx_modal = int(np.argmax(n_i))
C = float(x_i[idx_modal])

u_i = (x_i - C) / h

pd.DataFrame({
    "i": np.arange(1, len(x_i)+1),
    "a": a, "b": b,
    "x_i": x_i,
    "n_i": n_i,
    "u_i": u_i
})
```

## Таблица №1
Считаем суммы $\sum n_i u_i^k$ (для $k=1..4$) и вспомогательные столбцы.

```python

t1 = pd.DataFrame({
    "i": np.arange(1, len(x_i)+1),
    "x_i": x_i,
    "n_i": n_i,
    "u_i": u_i,
})

t1["n_i*u_i"]   = t1["n_i"] * t1["u_i"]
t1["n_i*u_i^2"] = t1["n_i"] * (t1["u_i"]**2)
t1["n_i*u_i^3"] = t1["n_i"] * (t1["u_i"]**3)
t1["n_i*u_i^4"] = t1["n_i"] * (t1["u_i"]**4)
t1["n_i*(u_i+1)^4"] = t1["n_i"] * ((t1["u_i"] + 1)**4)

# Суммы
sum_row = t1[["n_i", "n_i*u_i", "n_i*u_i^2", "n_i*u_i^3", "n_i*u_i^4", "n_i*(u_i+1)^4"]].sum()
sum_row.name = "Σ"

t1_display = t1.copy()
t1_display.loc["Σ", "i"] = ""
t1_display.loc["Σ", "x_i"] = ""
t1_display.loc["Σ", "n_i"] = sum_row["n_i"]
t1_display.loc["Σ", "u_i"] = "-"
t1_display.loc["Σ", "n_i*u_i"] = sum_row["n_i*u_i"]
t1_display.loc["Σ", "n_i*u_i^2"] = sum_row["n_i*u_i^2"]
t1_display.loc["Σ", "n_i*u_i^3"] = sum_row["n_i*u_i^3"]
t1_display.loc["Σ", "n_i*u_i^4"] = sum_row["n_i*u_i^4"]
t1_display.loc["Σ", "n_i*(u_i+1)^4"] = sum_row["n_i*(u_i+1)^4"]

t1_display
```

∑n_j (u_j+1)^4=∑n_j u_j^4 + 4∑n_j u_j^3 + 6 ∑n_j u_j^2 + 4∑n_j u_j + N

## Шаг 2. Условные и центральные эмпирические моменты
Условные моменты:
$$\bar M_k^* = \frac{1}{N}\sum n_i u_i^k.$$

Центральные моменты через $\bar M_k^*$ (для $k=2..4$):
$$\bar m_2 = (\bar M_2^*-(\bar M_1^*)^2)h^2,$$
$$\bar m_3 = (\bar M_3^* - 3\bar M_2^*\bar M_1^* + 2(\bar M_1^*)^3)h^3,$$
$$\bar m_4 = (\bar M_4^* - 4\bar M_3^*\bar M_1^* + 6\bar M_2^*(\bar M_1^*)^2 - 3(\bar M_1^*)^4)h^4.$$

```python

S1 = sum_row["n_i*u_i"]
S2 = sum_row["n_i*u_i^2"]
S3 = sum_row["n_i*u_i^3"]
S4 = sum_row["n_i*u_i^4"]

M1 = S1 / N
M2 = S2 / N
M3 = S3 / N
M4 = S4 / N

m2 = (M2 - M1**2)*h**2
m3 = (M3 - 3*M2*M1 + 2*(M1**3))*h**3
m4 = (M4 - 4*M3*M1 + 6*M2*(M1**2) - 3*(M1**4))*h**4

t2 = pd.DataFrame({
    "k": [1,2,3,4],
    "M*_k": [M1,M2,M3,M4],
    "m_k": [0, m2, m3, m4],
})
t2
```

## Шаг 3–5. Среднее, дисперсия, СКО, асимметрия, эксцесс
$$\bar x = \frac{1}{N}∑x_i n_i = A + h\bar M_1^*,$$
$$D = m_2 = \frac{1}{N}∑n_i(x_i - \bar x)^2, \quad \sigma = \sqrt{\bar m_2}.$$

Исправленные оценки:
$$S^2 = \frac{N}{N-1}D, \quad S=\sqrt{S^2}.$$

```python
x_bar = sum(x_i*n_i)/N
x_bar_prime = C + h * M1

D = sum((x_i - x_bar) ** 2 * n_i ) / N
D_prime = m2
sigma = math.sqrt(m2)

S2_corr = (N/(N-1)) * D_prime if N > 1 else float("nan")
S_corr = math.sqrt(S2_corr) if N > 1 else float("nan")

As = m3 / (S_corr**3) if S_corr > 0 else float("nan")
Ex = (m4 / (S_corr**4) - 3) if S_corr > 0 else float("nan")

pd.DataFrame({
    "Показатель": ["N", "C                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                            ", "h", "x̄", "x̄'", "D", "D'",  "σ", "S² (испр.)", "S (испр.)", "A_s", "E_x"],
    "Значение":   [N, C, h, x_bar, x_bar_prime, D, D_prime, sigma, S2_corr, S_corr, As, Ex]
})
```

Выборочные среднее $x̄_в$ и дисперсия $D_в$, посчитанные по стандартной формуле и методом условных вариант, совпали, так как получились одинаковые значения $x̄ = 0.139201$ и $D = 0.001419$.

Исправленная дисперсия $s² = 0.001433$ и исправленное СКО $s = 0.037851$ немного больше смещённых оценок $D = 0.001419$ и $σ = 0.037675$, потому что при исправлении деление идёт на $N−1$, а не на $N$. Разница небольшая (примерно $+0.93\%$ по дисперсии и $+0.47\%$ по СКО), поэтому при $N = 108$ влияние смещения уже слабое.

Коэффициент асимметрии $A_s = −0.292$ показывает слабую левостороннюю асимметрию распределения. Эксцесс $E = −0.501$ отрицательный, значит распределение более плоское, чем нормальное (вершина ниже и хвосты легче).


## Шаг 6. Мода, медиана, коэффициент вариации

```python
# Вычисление моды
m_tilde = n_i / N

i = int(np.argmax(m_tilde))

# Нижняя граница модального интервала x_{i-1}
x_im1 = float(a[i])

mt_i   = float(m_tilde[i])
mt_im1 = float(m_tilde[i - 1]) if i > 0 else 0.0
mt_ip1 = float(m_tilde[i + 1]) if i < len(m_tilde) - 1 else 0.0

Mo = x_im1 + h * ((mt_i - mt_im1) / ((mt_i - mt_im1) + (mt_i - mt_ip1)))

# Вычисление медианы

m_tilde_cum = np.cumsum(m_tilde) # Накопленные относительные частоты
i = int(np.searchsorted(m_tilde_cum, 0.5, side="left"))
x_im1 = float(a[i])

# Сумма до предыдущего интервала
mt_i = float(m_tilde[i])
sum_prev = float(m_tilde_cum[i - 1]) if i > 0 else 0.0

Me = x_im1 + (h / mt_i) * (0.5 - sum_prev)

# Вычисление коэффициента вариации ---
V = (sigma / x_bar) * 100

pd.DataFrame({
    "Показатель": ["M_o", "M_e", "V"],
    "Значение": [Mo, Me, V]
})
```

Мода меньше медианы, что указывает на правостороннюю асимметрию распределения: наиболее частое значение находится немного левее середины, а более крупные значения формируют «хвост» справа. Что отличается от оценки полученной по коэффициенту асимметрии. Это происходит, потому, как для оценки моды используется приблеженная формула, а расхожение возникает из-за группировки и округления, и выходит, что оценки моды (0.137), медианы (0.140) и среднего (0.139) близки к друг другу.

Коэффициент вариации отражает относительный уровень разброса. Полученное значение (V = 27%) свидетельствует о умеренной вариативности признака и однородности совокупности.
