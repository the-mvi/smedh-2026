## Основные теоретические положения

**Интервальная оценка** — статистическая оценка параметра $\theta$, которая задаётся двумя числами($(\theta^* − \delta; \theta^* + \delta)$), где $\delta$ — точность, $\gamma = P{|\theta^* − \theta| < \delta}$ — надёжность (доверительная вероятность). 

**Уровень значимости** - $q = 1 − \gamma$.

**Доверительный интервал для математического ожидания при неизвестном СКО**:

$$
\left( \bar x_в - \frac{t_\gamma S}{\sqrt{N}}, \bar x_в + \frac{t_\gamma S}{\sqrt{N}} \right)
$$, 
где $t_\gamma$ — коэффициент Стьюдента.

**Доверительный интервал для среднеквадратичного отклонения**:

$$
(S(1 - q), S(1 + q))
$$, 
где $q$ — табличный коэффициент, зависящий от объема выборки $N$ и надежности $\gamma$.

**Статистической гипотезой** называют гипотезу о виде неизвестного распледеления исследуемой случайной величиный или о значении параметров известных распределений.

Выдвинутую гипотезу называют **нулевой или основной гипотезой** и называют $H_0$.

**Конкурирующей или альтернативной** называют гипотезу $H_1$.

**Односторонний критерий** — это критерий проверки статистической гипотезы, при котором альтернативная гипотеза задаёт отклонение параметра только в одну сторону:
$$H_0: \theta \geq \theta_0, \qquad H_1: \theta < \theta_0 $$
или
$$
H_0: \theta \leq \theta_0, \qquad H_1: \theta > \theta_0.
$$

**Двухсторонний критерий** — это критерий проверки статистической гипотезы, при котором альтернативная гипотеза задаёт отклонение параметра в обе стороны:
$$
H_0: \theta = \theta_0, \qquad H_1: \theta \neq \theta_0.
$$о 
**Ошибка I рода** — отвергается верная $H_0$, хотя она верна.

**Ошибка II рода** — принимается неверная $H_0$, хотя она неверна.

**Мощность критерия** - вероятность совершения ошибки первого рода, обознают как $\alpha$.

**Критерий Пирсона**:

$$
\chi^2_{набл} = \sum_{i=1}^{K} \frac{(n_i - n_i')^2}{n_i'}
$$,
где $n_i'$ — теоретические частоты — частоты, с которыми нормально распределенная величина с параметрами $\bar{x}_в$ и $S$ попадала бы в $i$-й интервал интервального ряда при проведении $N$ испытаний.

Теоретические частоты вычисляются по формуле:

$$
n_i' = N \cdot p_i
$$, 

где $i = 1, 2, 3, ..., K$

$p_i$(вероятность попадания случайной величины в каждый $i$-й интервал интервального ряда) вычисляется по формуле:

$$
p_i = \Phi (z_{i+1}) - \Phi (z_i)
$$,
где $i = 1, 2, 3, ..., K$; 

$\Phi$ - функция Лапласа:

$$
\Phi (z) = \frac{1}{\sqrt{2\pi}} \int_0^z e^{-\frac{t^2}{2}} dt
$$

$z_i$(значения границ интервалов интервального ряда) вычисляется по формуле:
$$
z_i = \frac{x_i - \bar x_в}{S}
$$, 
где $i = 2, 3, ..., K$

**Областью принятия нулевой гипотезы** называют совокупность значения критерия, при которых нулевую гипотезу принимают.

**Критической областью** называют совокупность значения критерия, при которых нулевую гипотезу отвергают.

**Гипотеза о нормальном распределении** принимается, если выполняется:

$$
\chi^2_{набл} \leq \chi^2_{крит}
$$

$$

$$исло оценённых параметров нормального распределения ($\mu$ и $\sigma$).

```python
import numpy as np
import pandas as pd
import math
from scipy import stats

# Данные из ПР2
a = np.array([0.03, 0.053125, 0.07625, 0.099375, 0.1225, 0.145625, 0.16875, 0.191875])
b = np.array([0.053125, 0.07625, 0.099375, 0.1225, 0.145625, 0.16875, 0.191875, 0.215])
n_i = np.array([2, 2, 16, 12, 29, 19, 21, 7])
x_bar = 0.139201
s2 = 0.001433
s = 0.037851

N = int(n_i.sum())
x_i = (a + b) / 2

print(f"N = {N}")
```

## Задача 1

Найдем точность и доверительный интервал для математического ожидания при неизвестном среднеквадратичном отклонении для доверительной точности $\gamma \in \{0.95, 0.99\}$.

При неизвестном СКО используем распределение Стьюдента:
$$
\bar x \pm t_{1-\alpha/2,\;n-1}\frac{s}{\sqrt n}.
$$

Найдём интервалы для $\gamma = 0.95$ и $\gamma = 0.99$.

```python
gammas = [0.95, 0.99]
mu_rows = []

for gamma in gammas:
    alpha = 1 - gamma
    t_crit = stats.t.ppf(1 - alpha / 2, df=N - 1)
    eps = t_crit * s / math.sqrt(N)
    left = x_bar - eps
    right = x_bar + eps
    mu_rows.append([gamma, t_crit, eps, left, right])

mu_table = pd.DataFrame(mu_rows, columns=["γ", "t_кр", "ε", "Левая граница", "Правая граница"])
mu_table
```

С ростом доверительной вероятности интервал становится шире. Это видно при переходе от $\gamma = 0.95$ к $\gamma = 0.99$.

## Задача 2

Найдем значение $q$ при заданных $\gamma$ и $n$. Построим доверительные интервалы для среднеквадратичного отклонения.


Используем распределение $\chi^2$:
$$
\sqrt{\frac{(n-1)s^2}{\chi^2_{1-\alpha/2,\;n-1}}}
\le \sigma \le
\sqrt{\frac{(n-1)s^2}{\chi^2_{\alpha/2,\;n-1}}}.
$$

```python
q_rows = []

for gamma in gammas:
    k = N - 1
    alpha = 1 - gamma
    chi2_left = chi2.ppf(alpha / 2, df=k)
    chi2_right = chi2.ppf(1 - alpha / 2, df=k)

    q = (np.sqrt((k) / chi2_left) - np.sqrt((k) / chi2_right)) / 2
    left = s * (1 - q)
    right = s * (1 + q)
    q_rows.append([gamma, chi2_left, chi2_right, q, left, right])

q_table = pd.DataFrame(q_rows, columns=["γ", "χ²_лево", "χ²_право", "q", "Левая граница СКО", "Правая граница СКО"])

q_table
```

Построенные интервалы для среднеквадратичного отклонения при $\gamma = 0.95$ и $\gamma = 0.99$ не пересекаются, что говорит о том, что с ростом доверительной вероятности интервал становится шире.

Полученные значения для $q$ согласуются с табличными данными: полученные значения лежат между значениями для $n=100, 150$: $q(0.95, 100) = 0.143$, $q(0.99, 100) = 0.198$, $q(0.95, 150) = 0.115$, $q(0.99, 150) = 0.160$.

## Задача 3

Для крайних интервалов используем открытые границы:
- первый интервал: $(-\infty, b_1)$;
- последний интервал: $(a_k, +\infty)$.

Далее вычисляем:
$$
p_i = \Phi\left(\frac{b_i-\bar x}{s}\right) - \Phi\left(\frac{a_i-\bar x}{s}\right),
\qquad
n_i' = n p_i.
$$

```python
p_i = []
interval_labels = []

for i in range(len(a)):
    left = -np.inf if i == 0 else a[i]
    right = np.inf if i == len(a) - 1 else b[i]

    p = stats.norm.cdf((right - x_bar) / s) - stats.norm.cdf((left - x_bar) / s)
    p_i.append(p)

    if i == 0:
        interval_labels.append(f"(-∞; {b[i]:.6f})")
    elif i == len(a) - 1:
        interval_labels.append(f"[{a[i]:.6f}; +∞)")
    else:
        interval_labels.append(f"[{a[i]:.6f}; {b[i]:.6f})")

p_i = np.array(p_i)
n_theor = N * p_i

pearson_table = pd.DataFrame({
    "i": np.arange(1, len(n_i) + 1),
    "Интервал": interval_labels,
    "n_i": n_i,
    "p_i": p_i,
    "n_i'": n_theor,
    "(n_i-n_i')^2": (n_i - n_theor) ** 2,
    "(n_i-n_i')^2/n_i'": (n_i - n_theor) ** 2 / n_theor,
    "n_i^2": n_i ** 2,
    "n_i^2/n_i'": n_i ** 2 / n_theor
})

pearson_table
```

```python
sum_pearson_row = pd.DataFrame([{
    "i": "Σ",
    "Интервал": None,
    "n_i": n_i.sum(),
    "p_i": p_i.sum(),
    "n_i\'": n_theor.sum(),
    "(n_i-n_i')^2": None,
    "(n_i-n_i')^2/n_i'": pearson_table["(n_i-n_i')^2/n_i'"].sum(),
    "n_i^2": None,
    "n_i^2/n_i'": pearson_table["n_i^2/n_i'"].sum()
}])

pearson_table_pretty = pd.concat([pearson_table, sum_pearson_row], ignore_index=True)
pearson_table_pretty
```

## Задача 4

Основная формула:
$$
\chi^2_{набл} = \sum \frac{(n_i-n_i')^2}{n_i'}.
$$

Проверочная формула:
$$
\chi^2_{набл} = \sum \frac{n_i^2}{n_i'} - n.
$$

```python
chi2_obs = pearson_table["(n_i-n_i')^2/n_i'"].sum()
chi2_obs_check = pearson_table["n_i^2/n_i'"].sum() - N

print(f"χ²_набл = {chi2_obs:.10f}")
print(f"Проверка по формуле Σ(n_i² / n_i') - n = {chi2_obs_check:.10f}")
```

Оба способа дают одно и то же значение с точностью до округления. Значит, вычисления выполнены корректно.
