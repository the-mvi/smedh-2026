## Основные теоретические положения

Интервал $(\theta^* - \delta;\ \theta^* + \delta)$ называется доверительным интервалом и является интервальной оценкой параметра $\theta$.

Доверительный интервал для математического ожидания при неизвестном СКО:

$\left( \bar{x}_в - \frac{t_{\gamma} S}{\sqrt{N}}, \bar{x}_в + \frac{t_{\gamma} S}{\sqrt{N}} \right)$, где $ t_\gamma $ — коэффициент Стьюдента.

Доверительный интервал для математического ожидания при известном СКО:

$(S(1 - q), S(1 + q))$, где $ q $ — табличный коэффициент, зависящий от объема выборки $ N $ и надежности $ \gamma $.

Статистической гипотезой называют гипотезу о виде неизвестного распледеления исследуемой случайной величиный или о значении параметров известных распределений.

Выдвинутую гипотезу называют нулевой или основной гипотезой и называют $H_0$.

Проверку статистической гипотезы проводят для принятого уровня значимости $\alpha$

При проверке статистических гипотез на основе выборочных данных вычисляются значения этих случайных величин (наблюденные значения), которые сопоставляются с некоторыми, так называемыми, критическими (зависящими от заданного уровня значимости $\alpha$) значениями этих случайных величин.

Критерий Пирсона:

$\chi^2_{набл} = \sum_{i=1}^{K} \frac{(n_i - n_i')^2}{n_i'}$

Здесь $n_i'$ — теоретические частоты — частоты, с которыми нормально распределенная величина с параметрами $\bar{x}_в$ и $S$ попадала бы в $i$-й интервал интервального ряда при проведении $N$ испытаний.

Формула для вычисления теоретических частот:

$n_i' = N \cdot p_i$, где $i = 1, 2, 3, ..., K$

Формула для вычисления вероятности попадания случайной величины в каждый $i$-й интервал интервального ряда:

$p_i = \Phi (z_{i+1}) - \Phi (z_{i})$, где $i = 1, 2, 3, ..., K$

Формула для вычисления значений границ интервалов интервального ряда:

$z_i = \frac{x_i - \bar{x}_в}{S}$, где $i = 2, 3, ..., K$

Функция Лапласа:

$\Phi (z) = \frac{1}{\sqrt{2\pi}} \int_0^z e^{-\frac{t^2}{2}} dt$

Число степеней свободы:

$k = K - 3$

Областью принятия нулевой гипотезы называют совокупность значения критерия, при которых нулевую гипотезу принимают.

Критической областью называют совокупность значения критерия, при которых нулевую гипотезу отвергают.

Гипотеза о нормальном распределении принимается, если выполняется:

$ \chi^2_{набл} \leq \chi^2_{крит} $

Значение коэффициента $q$ было определено по таблице для ближайшего имеющего значения — 100, поскольку табличное значение для нашего объема выборки не было представлено.

Следовательно, увеличение доверительной вероятности делает интервальную оценку менее точной, так как диапазон допустимых значений истинного параметра расширяется.

## Задание 3

```python
# Функция Лапласа Φ(z)
def Phi(z):
    return stats.norm.cdf(z) - 0.5


intervals = [
    {"left": 10.0, "right": 15.2, "n_i": 16},
    {"left": 15.2, "right": 20.5, "n_i": 22},
    {"left": 20.5, "right": 25.7, "n_i": 25},
    {"left": 25.7, "right": 30.9, "n_i": 23},
    {"left": 30.9, "right": 36.1, "n_i": 17},
    {"left": 36.1, "right": 41.4, "n_i": 4},
    {"left": 41.4, "right": 46.6, "n_i": 3}]

rows = []
table_data = []

# Расчет p_i и n_i'
for i, inter in enumerate(intervals):

    left = inter["left"]
    right = inter["right"]
    n_i = inter["n_i"]

    if i == 0:
        z_l = -np.inf
        phi_l = -0.5
    else:
        z_l = (left - x_v) / s
        phi_l = Phi(z_l)

    if i == len(intervals) - 1:
        z_r = np.inf
        phi_r = 0.5
    else:
        z_r = (right - x_v) / s
        phi_r = Phi(z_r)

    p_i = round(phi_r - phi_l, 4)
    n_i_theor = round(N * p_i, 2)

    rows.append({
        "interval": f"[{left}, {right})",
        "n_i": n_i,
        "p_i": p_i,
        "n_i_theor": n_i_theor
    })

    table_data.append([
        i,
        f"{left}",
        f"{right}",
        n_i,
        "-∞" if z_l == -np.inf else f"{z_l:.3f}",
        "+∞" if z_r == np.inf else f"{z_r:.3f}",
        f"{phi_l:.4f}",
        f"{phi_r:.4f}",
        p_i,
        n_i_theor
    ])


headers = ["i", "x_i", "x_i+1", "n_i", "z_i", "z_i+1", "Φ(z_i)", "Φ(z_i+1)", "p_i", "n_i'"]
print(tabulate(table_data, headers=headers, tablefmt="grid"))
```

Так как в 7 интервале теоретическая частота меньше 5, следовательно, нужно объединить 6 и 7 интервалы.

```python
# Объединение интервалов (n_i' < 5)
merged = []
for row in rows:
    if not merged:
        merged.append(row)
        continue
    if row["n_i_theor"] < 5:
        merged[-1]["interval"] += " + " + row["interval"]
        merged[-1]["n_i"] += row["n_i"]
        merged[-1]["p_i"] += row["p_i"]
        merged[-1]["n_i_theor"] += row["n_i_theor"]
    else:
        merged.append(row)

# Таблица №1
chi_table = []
sum_chi = 0
sum_n_sq_div = 0

for i, row in enumerate(merged, 1):

    n_i = row["n_i"]
    p_i = row["p_i"]
    n_i_theor = row["n_i_theor"]

    diff_sq = round((n_i - n_i_theor) ** 2, 2)
    chi = round(diff_sq / n_i_theor, 2)
    n_sq_div = round((n_i ** 2) / n_i_theor, 2)

    sum_chi += chi
    sum_n_sq_div += n_sq_div

    chi_table.append([i, row["interval"], n_i, p_i, n_i_theor, diff_sq, chi, n_i ** 2, n_sq_div])

# Вычисление сумм
chi_table.append([
    "",
    "Σ",
    sum(r[2] for r in chi_table),
    sum(r[3] for r in chi_table),
    sum(r[4] for r in chi_table),
    "-",
    sum_chi,
    "-",
    sum_n_sq_div
])

headers = ["i", "[x_i, x_i+1)", "n_i", "p_i", "n_i'", "(n_i - n_i')^2", "(n_i - n_i')^2 / n_i'", "n_i^2", "n_i^2 / n_i'"]
print(tabulate(chi_table, headers=headers, tablefmt="grid"))
print("\nНаблюдаемое значение из таблицы: χ² =", round(sum_chi, 2))
```

## Задание 4

Докажем, что

$\chi^2_{набл} = \sum_{i=1}^{k} \frac{n_i^2}{n'_i} - n$

Доказательство:

$\chi^2_{набл} = \sum_{i=1}^{k} \frac{(n_i - n'_i)^2}{n'_i} = \sum_{i=1}^{k} \frac{n_i^2 - 2n_i n'_i + (n'_i)^2}{n'_i} = \sum_{i=1}^{k} \frac{n_i^2}{n'_i} - 2\sum_{i=1}^{k} n_i + \sum_{i=1}^{k} n'_i = \sum_{i=1}^{k} \frac{n_i^2}{n'_i} - 2n + n = \sum_{i=1}^{k} \frac{n_i^2}{n'_i} - n$

```python
chi_obs = round(sum_n_sq_div - N, 2)
print("Наблюдаемое значение по формуле: χ² =", chi_obs)
```

Оба вычисленных наблюдаемых значения сошлись.

## Задание 5

```python
K = len(merged)
df = K - 3
alpha = 0.05

chi_crit = round(stats.chi2.ppf(1 - alpha, df), 2)

print("Критическое значение χ² =", chi_crit)

if sum_chi < chi_crit:
    print("Нулевая гипотеза принимается")
else:
    print("Нулевая гипотеза отвергается")
```

Так как наблюдаемое значение критерия $\chi^2$ оказалось меньше критического, нулевая гипотеза принимается на уровне значимости $\alpha = 0,05$. То есть различия между эмпирическими и теоретическими частотами не являются существенными. Следовательно, выборочные данные согласуются с нормальным распределением признака.
