## Основные теоретические положения

В практической работе будут рассмотренны следующие точечные статистические оценки:

- Начальный эмпирический момент $k$-го порядка:

  $\bar{M}_k = \frac{1}{N}\sum n_j x_j^k$;

  $N = \sum n_j - $ размер выборки; $n_j - $ абсолютная частота $j$-й варианты

- Выборочное среднее $\bar{x}_в = \bar{M}_1$

- Центральный эмпирический момент $k$-го порядка:

  $\bar{m}_k=\frac{1}{N} \sum n_j (x_j - \bar{x}_в)^k$

- Дисперсия $D = \bar{m}_2$, эта оценка является смещенной

- Исправленная оценка дисперсии: $s^2 = \frac{N}{N-1}D_в$

- Среднеквадратическое отклонение:
  - Смещенное: $σ_в = \sqrt{D_в}$
  - Исправленное: $s = \sqrt{s^2}$

- Асимметрии: $\bar{A}_s = \frac{\bar{m}_3}{s^3}$

- Эксцесса: $\bar{E} = \frac{\bar{m}_4}{s^4} - 3$

- Условные эмпирические моменты:

  $\bar{M}_k^*= \frac{1}{N} \sum n_j \left( \frac{x_j - C}{h}\right)^k$;

  $C -$ условный ноль, значения которого выбирается равным значению варианты интервального ряда, являющейся средней или близкой к средней по значению в этом ряду; $h-$ длина интервала.

- Мода, $M_o^*$ $-$ варианта с наибольшей частотой.

Для интервального ряда: $M_o^* = x_{i-1} + h\frac{\tilde{m}_i - \tilde{m}_{i-1}}{\tilde{m}_i - \tilde{m}_{i-1} + \tilde{m}_i - \tilde{m}_{i-1}}$, $i -$ номер интервала наибольшей частоты, $\tilde{m}_i - $ относительная частота ряда $i$

- Медиана, $M_e^* -$ варианта, соответствующая середине ранжированного ряда.
Для интервального ряда: $M_e^* = x_{i-1} + \frac{h}{\tilde{m}_i} \left(0.5 - \overset{i-1}{\underset{j=1}{\sum}}\tilde{m}_j  \right)$

#### Задание 3
Вычислить выборочные среднее
$\bar{x}_в$
 и дисперсию
$D_в$
 с помощью стандартной формулы и с помощью условных вариант. Убедиться, что результаты совпадают. Вычислить выборочное СКО $σ_в$

##### Выборочное среднее $\bar{x}_в$

```python
# Через условные варианты
xv1 = df2.iloc[0,0] * h + C

# Через стандартную формулу
xv2 = df1.loc[:7, 'mid_xi'].to_numpy().dot(df1.loc[:7, 'ni'])
print(f"Выборочные среднее через условные варианты: {xv1} \nВыборочные среднее через стандартные формулы: {xv2} \nРазница: {xv1 - xv2}")
```

Видно, что оба числа практически совпадают. Разница могла возникнуть в следствии ошибок округления при рассчетах.

Выборочная дисперсия $D_в$

```python
# Через условные варианты
D1 = df2.loc[2, 'mk']

# Через стандартную формулу
D2 = ((df1.loc[:7, 'mid_xi'] - xv2) ** 2).to_numpy().dot(df1.loc[:7, 'ni'].to_numpy())
print(f"Дисперсия через условные варианты: {D1} \nДисперсия через стандартные формулы: {D2} \nРазница: {D1 - D2}")
```

Оба значения совпали

##### Выборочное СКО $σ_в$

```python
print(np.sqrt(D1))
```

#### Задание 4
Вычислить исправленную выборочную дисперсию $s^2$ и исправленное СКО $s$. Сравнить данные оценки с смещёнными оценками дисперсии и СКО. Сделать выводы.

```python
# Исправленная дисперсия
s2 = N * D1 / (N - 1)
# Исправленное СКО
s = np.sqrt(s2)

print(f"Исправленная дисперсия: {s2} \nИсправленное СКО: {s} \nРазница: \nДисперсий: {s2 - D1} \n СКО: {s - np.sqrt(D1)}")
```

Исправленные оценки оказались немного больше смещенных оценок в виду достаточно большого размера выборки $N=112$.
