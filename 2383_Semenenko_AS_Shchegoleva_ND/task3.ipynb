## Основные теоретические положения

**Интервальная оценка** - статистическая оценка, которая задается двумя числами (начало и конец интервала где может находится истинное значение параметра $\theta$)


**Точность $\theta^*$** - значение $\delta$ статистическая оценка параметра $\theta^*$, которое удовлетворяет неравенству:

$|\theta^*-\theta|<\delta $

**Наддежность (доверительная вероятность) оценки** $\Theta^*$ - вероятность $\gamma$, для которой выполняется неравенство:

$ \gamma = P\{ \delta > |\Theta - \Theta^*| \} = P\{\Theta^* - \delta < \Theta < \Theta^* + \delta \}$


Доверительный интервал бывает односторонним и двусторонним:

$ \mathbb{P}\{ \beta_1(\gamma) < \Theta < \beta_2(\gamma) \} = \gamma$,

$P\{ \beta(\gamma) < \Theta \} = \gamma$   или   $ P\{ \beta(\gamma) > \Theta \} = \gamma$,

$q = 1-\gamma$ - уровень значимости.

**Доверительный интервал для математического ожидания при неизвестном СКВО**
$$(\bar x_в - t_γS / \sqrt N, \bar x_в + t_γS / \sqrt N)$$



**Нулевая гипотиза $H_0$** - выдвинутая (основная) гипотеза.

**Конкурирующая гипотеза $H_1$** - альтарнативная гипотеза, противоречащая $H_0$.

**Ошибка первого рода** - отвергается верная нулевая гипотеза.

**Ошибка второго рода** - принимается неверная нулевая гипотеза.

**Уровень значимости $\alpha$** - вероятность совешить ошибку первого рода.
**$\beta$** - вероятность ошибки второго рода.

**Мощность критерия $1 - \beta$** - мощность критерия.

Для проверки стат. гипотез используются критерии, с помощью которых вычисляются наблюденные и критические значения случайных величин этих критериев. На основании этих значений выделяют критическую область и область принятия нулевой гипотезы.

3. Проверим гипотезу о нормальности заданного распределения с помощью критерия $χ^2$ (Пирсона). Для этого найдем теоретические частоты и вычислим наблюдаемое значение критерия.

```python
# функция для объединения интервалов с абсолютной частотой меньше 5
def merge_intervals(df):
    i = 1
    while i < len(df):
        if df.loc[i, 'Абсолютная частота (m_i)'] < 5:
            prev_interval = df.loc[i-1, 'Интервал']
            curr_interval = df.loc[i, 'Интервал']

            left = min(prev_interval[0], curr_interval[0])
            right = max(prev_interval[-1], curr_interval[-1])

            df.at[i-1, 'Интервал'] = [left, right]
            df.loc[i-1, 'Абсолютная частота (m_i)'] += df.loc[i, 'Абсолютная частота (m_i)']
            df = df.drop(i).reset_index(drop=True)
            i-=1

        else:
            i += 1

    return df
```

```python
def table_hi(df, s, x_b):
  data_rows = []
  z_borders = []
  k = int(len(df))
  p_i = np.zeros(k)
  borders = np.array([inter[0] for inter in df['Интервал']] + [df['Интервал'].iloc[-1][1]])

  for i in range(k + 1):
    if i == 0:
        z_borders.append(-np.inf)
    elif i == k:
        z_borders.append(np.inf)
    else:
        z_borders.append((borders[i] - x_b) / s)

  for i in range(k):
    p_i[i] = (stats.norm.cdf(z_borders[i + 1]) - stats.norm.cdf(z_borders[i]))

  n_s = p_i*117
  for i in range(1, k+1):
    inter = df.loc[i]['Интервал']
    n_i = df.loc[i]['Абсолютная частота (m_i)']
    data_rows.append({
        'i': inter,
        'n_i': n_i,
        'p_i': round(p_i[i-1], 2),
        "n'_i": round(n_s[i-1], 3),
        "(n_i-n'_i)²": round((n_i - n_s[i-1])**2, 3),
        "(n_i-n'_i)²/n'_i": round(((n_i - n_s[i-1])**2)/ n_s[i-1], 3),
        "n_i²": n_i**2,
        "n_i²/n'_i": round((n_i**2)/n_s[i-1], 3)})
  df = pd.DataFrame(data_rows, index=range(1, len(data_rows)+1))
  df.loc[len(df) + 1] = [
      'Σ',
      df['n_i'].sum(),
      sum(p_i),
      df["n'_i"].sum(),
      '—',
      df["(n_i-n'_i)²/n'_i"].sum(),
      '—',
      df["n_i²/n'_i"].sum(),
  ]
  display(df)
```

```python
df_lenth_merged = merge_intervals(df_lenth)
df_lenth_merged.index = range(1, len(df_lenth_merged) + 1)
```

```python
print("Таблица для длины клюва")
table_hi(df_lenth_merged, s_l, x_B_l)
```

```python
print("Таблица для высоты клюва")
table_hi(df_depth, s_d, x_B_d)
```

4. Докажем, что $χ^2_{набл} = \sum_{i = 1}^k \cfrac{n_i^2}{n_i^{'}} - n$

$χ^2_{набл} = \sum_{i = 1}^k \cfrac{(n_i - n_i^{'})^2}{n_i^{'}}  = \sum_{i = 1}^k \cfrac{n_i^2 - 2n_in_i^{'} + n_i^{'^2}}{n_i^{'}} = \sum \cfrac{n_i^2}{n_i^{'}} - 2\sum n_i + \sum n_i^{'} = \sum \cfrac{n_i^2}{n_i^{'}} - 2n +n = \sum \cfrac{n_i^2}{n_i^{'}} - n$

Проверим для характеристики длина клюва:

$χ^2_{набл} = \sum_{i = 1}^k \cfrac{(n_i - n_i^{'})^2}{n_i^{'}}  = 5,79$

$\sum \cfrac{n_i^2}{n_i^{'}} - n = 122,79 - 117 = 5,79$


Результаты совпали.

Проверим для характеристики высота клюва:

$\sum \cfrac{n_i^2}{n_i^{'}} - n = 133,405 - 117 = 16,405$
