## Основные теоретические положения
### Начальный эмпирический момент
Начальным эмпирическим моментом к-го порядка называется среднее значение к-х степеней элементов вариационного ряда:
$$\bar{M}_k = \frac{1}{N} \sum n_jx_j^k$$

### Центральный эмпирический момент
Центральным эмпирическим моментом к-го порядка называется среднее значение к-х степеней разностей $(x_j - \bar{x}_в)$ для вариационного или интервального ряда:
$$\bar{m}_k = \cfrac{1}{N}\sum n_j(x_j - \bar{x}_в)^k$$

### Статистическая оченка дисперсии(смещенная):
$$D_в = \bar{m_2} = \cfrac{1}{N}\sum n_j(x_j - \bar{x}_в)^2$$

### Исправленная оценка дисперсии:
$$s^2 = \cfrac{N}{N-1}D_в$$

### Статистические оценки СКО:
$$\sigma_в = \sqrt{D_в};   s = \sqrt{s^2}$$

### Статистические оценки ассиметрии и эксцесса
$$\bar{A}_s = \cfrac{\bar{m_3}}{s^3}; \bar{E} = \cfrac{\bar{m_4}}{s^4} - 3$$

Если $\bar{E} >0$, то эмпирическое распределение является более высоким (островершинным) – относительно нормального распределения. Если же $\bar{E} < 0 $– то более низким и пологим. И чем больше $\bar{E}$ по модулю, тем аномальнее высота в ту или иную сторону.

### Условные варианты
$$u_j = \cfrac{x_j -C}{h},$$

где С - условный ноль, значение которого выбирается равным значению варинаты интервального ряда, являющейся средней или близкой к средней по значению в этом ряду. В результате все условные варианты оказываются целыми числами.

### Условные моменты к-того порядка:
$$\bar{M^*_k} = \cfrac{1}{N} \sum n_j \bigg(\cfrac{x_j -C}{h}\bigg)^k = \cfrac{1}{N}\sum n_ju_j^k$$

### Мода

Мода $M^*_o$ вариационного ряда - это варианта с наибольшей частотой.

Мода интервального ряда находится для интервала $(x_{i-1}, x_i)$:

$$M_o^* = x_{i-1} + h \cdot \frac{\tilde{m}_i - \tilde{m}_{i-1}}{(\tilde{m}_i - \tilde{m}_{i-1}) + (\tilde{m}_i - \tilde{m}_{i+1})}$$

где:
- $x_{i-1}$ — начало модального интервала;
- $h$ — ширина интервала;
- $\tilde{m}_i$ — относительная частота модального интервала;
- $\tilde{m}_{i-1}$, $\tilde{m}_{i+1}$ — относительные частоты соседних интервалов.

### Медиана
Медиана $M_e^*$ вариационного ряда — это варианта, соответствующая середине ранжированного ряда.

Для ряда с нечетным числом элементов

$$M_e^* = x_{m+1}$$

Для ряда с четным числом элементов

$$M_e^* = \frac{x_m + x_{m+1}}{2}$$

Для интервального ряда выборочное значение медианы может быть вычислено по формуле:

$$M_e^* = x_{i-1} + h \cdot \frac{0.5 - \sum_{j=1}^{i-1} \tilde{m}_j}{\tilde{m}_i}$$

где:
* $x_{i-1}$ — начало медианного интервала;
* $m_i$ — относительная частота медианного интервала;
* $\sum_{j=1}^{i-1} m_j$ — накопленная относительная частота для интервалов, предшествующих медианному.

Условие выбора медианного интервала:

Значение медианы принадлежит интервалу, для которого накопленная относительная частота равна или превышает 0.5, а накопленная относительная частота для предыдущего интервала меньше 0.5.

Прямая $x = M_e^*$ делит площадь гистограммы на две равные части.

### Выборочный коэффициент вариации

Выборочный коэффициент вариации $V^*$ характеризует относительный разброс значений вариант относительно выборочного среднего и выражается в процентах:

$$V^* = \frac{\sigma_x}{\bar{x}} \cdot 100\%$$

1. Для середин интервального ряда, полученного в практической работе №1, вычислим условные варианты

```python
def conditional_options(df, h, c_i):
  data_rows = []
  C = df.loc[c_i]['Середина интервала']
  for i in range(1, k+1):
    x_i = df.loc[i]['Середина интервала']
    n_i = df.loc[i]['Абсолютная частота (m_i)']
    u_i = round((x_i - C) / h)
    data_rows.append({
        'i': i,
        'x_i': x_i,
        'n_i': n_i,
        'u_i': u_i,
        'n_i*u_i': n_i * u_i,
        'n_i*u_i^2': n_i * u_i**2,
        'n_i*u_i^3': n_i * u_i**3,
        'n_i*u_i^4': n_i * u_i**4,
        'n_i(u_i+1)^4': n_i * (u_i+1)**4
        })
  df = pd.DataFrame(data_rows, index=range(1, len(data_rows)+1))
  df.loc[len(df) + 1] = [
      'Σ',
      '-',
      round(df['n_i'].sum()),
      '—',
      df['n_i*u_i'].sum(),
      df['n_i*u_i^2'].sum(),
      df['n_i*u_i^3'].sum(),
      df['n_i*u_i^4'].sum(),
      df['n_i(u_i+1)^4'].sum(),
  ]
  display(df)
  print('∑ n_i(u_i+1)^4 = ', round(df.loc[len(df)]['n_i(u_i+1)^4'], 4))
  print('∑ n_i*u_i^4 + 4∑ n_i*u_i^3 + 6∑ n_i*u_i^2 + 4∑ n_i*u_i + N = ', end='')
  print(round(df.loc[len(df)]['n_i*u_i^4'] + 4*df.loc[len(df)]['n_i*u_i^3'] +
              6*df.loc[len(df)]['n_i*u_i^2'] + 4*df.loc[len(df)]['n_i*u_i']+N, 4))
  return df
```

```python
print('Таблица №1 для признака culmen_length_mm')
co_len = conditional_options(df_lenth, h_lenth, 4)
```

2. Вычислим условные эмпирические моменты $\bar{M}^*_k$ через условные варианты. С помощью условных эмпирических моментов вычислим центральные эмпирические моменты $\bar{m}_k$.

```python
def moments(df, h, c_i):
  data_rows = []
  C = df.loc[c_i]['x_i']
  size = len(df)
  M_1 = round(df.loc[size]['n_i*u_i']/N, 2)
  M_2 =  round(df.loc[size]['n_i*u_i^2']/N, 2)
  M_3 =  round(df.loc[size]['n_i*u_i^3']/N, 2)
  M_4 =  round(df.loc[size]['n_i*u_i^4']/N, 2)
  M = [M_1, M_2, M_3, M_4]
  m_1 = 0
  m_2 = round((M_2 - M_1**2) * h**2, 2)
  m_3 = round((M_3 - 3*M_2*M_1 + 2*M_1**3) * h**3, 2)
  m_4 = round((M_4 - 4*M_3*M_1 + 6*M_2*M_1**2 - 3*M_1**4) * h**4, 2)
  m = [m_1, m_2, m_3, m_4]
  for i in range(1, 5):
    data_rows.append({
        'k': i,
        'M̄*_k': M[i-1],
        'm̄_k': m[i-1]
    })
  df = pd.DataFrame(data_rows, index=range(1, len(data_rows)+1))
  display(df)
  print('Выборочные среднее x_в = ', round(M_1 * h + C, 2))
  print('СКО = ', round(m_2**(1/2), 2))
  return df
```

3. Вычислим выборочные среднее $\bar{x}_в$ и дисперсию $D_в$ с помощью стандартной формулы и с помощью условных вариант. Вычислим выборочное СКО
$σ_в$

```python
def calculate_x_D(df):
  product  = []
  squared_deviations = []
  for idx, row in df.iterrows():
      x_j = row['Середина интервала']
      n_j = row['Относительная частота (m_i/N)']
      product.append(x_j * n_j)

  x_B = round(sum(product), 2)
  for idx, row in df.iterrows():
    x_j = row['Середина интервала']
    n_j = row['Относительная частота (m_i/N)']
    squared_deviations.append((x_j - x_B) ** 2 * n_j)
  D = round(sum(squared_deviations), 2)
  sigma = round(D**(0.5), 2)
  print(x_B, D, sigma)
  return x_B, D, sigma
```

```python
print('Выборочные среднее x¯в, дисперсия Dв и СКО σв\n')
print('C помощью стандартной формулы:')
print('Для признака длина клюва:')
x_B_l, D_l, sigma_l = calculate_x_D(df_lenth)
print('Для признака высота клюва:')
x_B_d, D_d, sigma_d = calculate_x_D(df_depth)
```

5. Найдем статистическую оценку коэффициентов асимметрии
$\bar{A}_s$ и эксцесса $\bar{E}_в$

```python
def calculate_A_E(df, h, c_i, D):
  data_rows = []
  C = df.loc[c_i]['x_i']
  size = len(df)
  M_1 = round(df.loc[size]['n_i*u_i'] / N, 2)
  M_2 =  round(df.loc[size]['n_i*u_i^2'] / N, 2)
  M_3 =  round(df.loc[size]['n_i*u_i^3'] / N, 2)
  M_4 =  round(df.loc[size]['n_i*u_i^4'] / N, 2)
  M = [M_1, M_2, M_3, M_4]
  m_1 = round(M_1 * h + C, 2)
  m_2 = round((M_2 - M_1**2) * h**2, 2)
  m_3 = round((M_3 - 3*M_2*M_1 + 2*M_1**3) * h**3, 2)
  m_4 = round((M_4 - 4*M_3*M_1 + 6*M_2*M_1**2 - 3*M_1**4) * h**4, 2)

  s_3 = ((N/(N-1) * D)**(0.5))**3
  s_4 = (N/(N-1) * D)**2

  A = round(m_3/s_3, 2)
  E = round(m_4/s_4 - 3, 2)
  return A, E
```

```python
A_len, E_len = calculate_A_E(co_len, h_lenth, 4, D_l)
print("Статистические оценки коэффициентов асимметрии 𝐴¯𝑠 и эксцесса  𝐸¯в для длины клюва", "\nA = ", A_len,"\nE = ", E_len)
```

Для длины клюва коэффициент ассиметрии больше 0, что означает, что в распределение умеренно скошено впарво.

Так как коэффициент эксцесса больше 0, то распределение островершинное относительно нормального распределения.

```python
A_d, E_d = calculate_A_E(co_depth, h_depth, 2, D_d)
print("Статистические оценки коэффициентов асимметрии 𝐴¯𝑠 и эксцесса  𝐸¯в для высоты клюва", "\nA = ", A_d,"\nE = ", E_d)
```

Для высоты клюва коэффициент ассиметрии больше 0, что говорит о том, что распределение скошено вправо и, так как значение больше 0.5, то ассиметрия существенная.

Коэффициент эксцесса меньше 0, значит распределение является пологим.

6. Для интервального ряда вычислим моду, медиану и коэффициент вариации $V^∗$

Мода интервального ряда определяется через интервал с наибольшей частотой.

Для признака длина клюва это 4-ый интервал [45.3, 48.7] с относительной частотой 0,2991.

Для признака высота клюва это 2-ой интервал [14.3, 15.3] с относительной частотой 0,2650.

```python
def mode(df, i, h):
  x_i = df.loc[i]['Интервал'][0]
  m_i = df.loc[i]['Относительная частота (m_i/N)']
  m_0 = df.loc[i-1]['Относительная частота (m_i/N)']
  m_1 = df.loc[i+1]['Относительная частота (m_i/N)']
  M_0 = x_i + h * (m_i - m_0) / (2*m_i - m_0 - m_1)
  return round(M_0, 4)

mode_length = mode(df_lenth, 4, h_lenth)
mode_depth =  mode(df_depth, 2, h_depth)
print('Мода для признака длина клюва = ', mode_length)
print('Мода для признака высота клюва = ', mode_depth)
```

Значение медианы принадлежит интервалу, для которого накопленная относительная частота равна или превышает 0.5, а накопленная относительная частота для предыдущего интервала меньше 0,5.

Таким образом для признака длина клюва это 3-ый интервал [42.0, 45.2] с накопленной относительной частотой 0,5385.

А для  признака высота клюва это 3-ой интервал [15.4, 16.5] с накопленной относительной частотой 0,6239.

```python
def median(df, i, h):
  x_i = df.loc[i]['Интервал'][0]
  m_i = df.loc[i]['Относительная частота (m_i/N)']
  sum_m_i_1 = df.loc[i-1]['Накопленная отн. частота']
  M_e = x_i + h / m_i * (0.5 - sum_m_i_1)
  return round(M_e, 4)

median_length = median(df_lenth, 3, h_lenth)
median_depth =  median(df_depth, 3, h_depth)
print('Мода для признака длина клюва = ', median_length)
print('Мода для признака высота клюва = ', median_depth)
```

```python
V_l = round(sigma_l/x_B_l*100, 2)
print('Выборочный коэффициент вариации для признака длина клюва = ', V_l)
V_d = round(sigma_d/x_B_d*100, 2)
print('Выборочный коэффициент вариации для признака высота клюва = ', V_d)
```
