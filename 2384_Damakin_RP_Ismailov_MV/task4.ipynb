## Основные теоретические сведения

**Корреляционная зависимость** отражает наличие статистической связи между двумя случайными величинами $X$ и $Y$.

**Выборочный коэффициент корреляции** служит оценкой генерального коэффициента корреляции и характеризует степень тесноты линейной связи между признаками. Его можно вычислить по формуле:

$\bar r_{xy} = \frac{\sum_{i=1}^{K_y} \sum_{j=1}^{K_x} n_{ij} y_i x_j - N \bar x_в \bar y_в}{N S_x S_y}$

Здесь $n_{ij}$ обозначает частоту попадания наблюдений в соответствующую ячейку корреляционной таблицы, $x_i$ и $y_i$ — значения или середины интервалов вариант, а $N$ — объём выборки.

Для упрощения вычислений выборочный коэффициент корреляции можно находить с помощью **условных вариант**:

$\bar r_{xy} = \frac{\sum_{i=1}^{K_y} \sum_{j=1}^{K_x} n_{ij} u_i v_j - N \bar u_в \bar v_в}{N \bar \sigma_u \bar \sigma_v}$

где

$u_i = \frac{x_i - C_x}{h_x}$,  
$v_j = \frac{y_j - C_y}{h_y}$,  

$C_x, C_y$ — условные начала отсчёта, а $h_x, h_y$ — величины интервалов группировки;

$\bar \sigma_u = \sqrt{\bar{u^2} - \bar u^2}$,  
$\bar \sigma_v = \sqrt{\bar{v^2} - \bar v^2}$

Для построения доверительного интервала применяется **преобразование Фишера**:

$z = 0.5 \ln \frac{1 + \bar r_{xy}}{1 - \bar r_{xy}}$

Среднее квадратическое отклонение величины $z$ определяется формулой:

$\bar \sigma_z = \frac{1}{\sqrt{N - 3}}$

Тогда доверительный интервал для генерального коэффициента корреляции имеет вид:

$( \bar z - \lambda(\gamma) \cdot \bar \sigma_z, \bar z + \lambda(\gamma) \cdot \bar \sigma_z)$,

где значение $\lambda(\gamma)$ выбирается из условия:

$\Phi [\lambda(\gamma)] = \frac{\gamma}{2}$

После этого выполняется **обратное преобразование Фишера**:

$r = \tanh(z) = \frac{e^{2z} - 1}{e^{2z} + 1}$

Для проверки значимости выборочного коэффициента корреляции рассматривают гипотезы:

$H_0: r = 0$,  
$H_1: r \ne 0$

В качестве статистики критерия используют величину:

$T_{набл} = \frac{\bar r_{xy} \sqrt{N - 2}}{\sqrt{1 - \bar r_{xy}^2}}$

Если выполняется условие $|T_{набл}| \leq t_{крит}(\alpha, k)$, то нет оснований отвергать гипотезу $H_0$.  
Если же $|T_{набл}| > t_{крит}(\alpha, k)$, то гипотеза $H_0$ не согласуется с выборочными данными и отвергается.

```python
# Формируем интервальный ряд для признака E
data = raw['E'].dropna().reset_index(drop=True)

# Размер выборочной совокупности
N = len(data)

# Определяем количество интервалов и их ширину
k = int(1 + 3.31 * np.log10(N))
h = (data.max() - data.min()) / k

# Вычисляем границы интервалов
intervals = [round(data.min() + i * h, 1) for i in range(k + 1)]

# Создаём структуру для хранения интервалов
interval_data = []
for i in range(k):
    left, right = intervals[i], intervals[i + 1]
    interval_data.append([i + 1, left, right, []])

# Раскладываем наблюдения по соответствующим интервалам
for val in data:
    for i, left, right, vals in interval_data:
        if (i == k and left <= val <= right) or (i < k and left <= val < right):
            vals.append(val)
            break

# Готовим данные для таблицы интервального распределения
abs_acc_m_i = 0
intervals_mid = []
frequencies_abs = []
accumulated_freq_abs = []
intervals_left = []
intervals_right = []
rows = []

for i, left, right, vals in interval_data:
    abs_m_i = len(vals)
    abs_acc_m_i += abs_m_i
    rel_m_i = round(abs_m_i / N, 4)
    rel_acc_m_i = round(abs_acc_m_i / N, 4)
    mid = round((left + right) / 2, 2)

    intervals_mid.append(mid)
    frequencies_abs.append(abs_m_i)
    accumulated_freq_abs.append(abs_acc_m_i)
    intervals_left.append(left)
    intervals_right.append(right)

    if i == k:
        interval_str = f"[{left:.1f}, {right:.1f}]"
    else:
        interval_str = f"[{left:.1f}, {right:.1f})"

    rows.append({
        'i': i,
        'Интервал': interval_str,
        '~x_i': mid,
        'm_i': abs_m_i,
        '~m_i': rel_m_i,
        'm_нак_i': abs_acc_m_i,
        '~m_нак_i': rel_acc_m_i
    })

rows.append({
    'i': 'Σ',
    'Интервал': '-',
    '~x_i': '-',
    'm_i': abs_acc_m_i,
    '~m_i': round(abs_acc_m_i / N, 4),
    'm_нак_i': '-',
    '~m_нак_i': '-'
})

intervals_df = pd.DataFrame(rows)

md('### Интервальный вариационный ряд')
md_table(intervals_df, digits=4)

# Подготавливаем значения для построения графиков
frequencies_rel = [freq / N for freq in frequencies_abs]
accumulated_freq_rel = [acc / N for acc in accumulated_freq_abs]
x_step = [0] + intervals
y_step_abs = [0] + accumulated_freq_abs + [1.0]
y_step_rel = [0] + accumulated_freq_rel + [1.0]

fig, axes = plt.subplots(2, 3, figsize=(20, 12))

# Полигон распределения абсолютных частот
ax = axes[0, 0]
ax.plot(intervals_mid, frequencies_abs, 'o-', color='teal', linewidth=2, markersize=6, markerfacecolor='turquoise')
ax.set_xticks(intervals_mid)
ax.set_xticklabels([f'{x:.1f}' for x in intervals_mid])
ax.set_title('Полигон абсолютных частот')
ax.set_xlabel('Середины интервалов')
ax.set_ylabel('Абсолютные частоты')
ax.grid()

# Гистограмма распределения абсолютных частот
ax = axes[0, 1]
for i in range(k):
    left = intervals_left[i]
    right = intervals_right[i]
    width = right - left
    height = frequencies_abs[i] / width
    ax.bar(left, height, width, align='edge', color='plum', edgecolor='purple', linewidth=1.5)
ax.set_xticks(intervals_mid)
ax.set_xticklabels([f'{x:.1f}' for x in intervals_mid])
ax.set_title('Гистограмма абсолютных частот')
ax.set_xlabel('Середины интервалов')
ax.set_ylabel('Абсолютные частоты')
ax.grid()

# Эмпирическая функция распределения по абсолютным частотам
ax = axes[0, 2]
ax.step(x_step, y_step_abs, where='post', linewidth=2.5, color='seagreen')
ax.scatter(x_step[:-1], y_step_abs[:-1], color='seagreen', s=30, zorder=5)
ax.set_xticks(x_step)
ax.set_xticklabels([f'{x:.1f}' for x in x_step])
ax.set_title('Эмпирическая функция распределения (абсолютные частоты)')
ax.set_ylabel('Накопленная абсолютная частота')
ax.set_xlabel('Границы интервалов')
ax.grid()

# Полигон распределения относительных частот
ax = axes[1, 0]
ax.plot(intervals_mid, frequencies_rel, 'o-', color='coral', linewidth=2, markersize=6, markerfacecolor='peachpuff')
ax.set_xticks(intervals_mid)
ax.set_xticklabels([f'{x:.1f}' for x in intervals_mid])
ax.set_title('Полигон относительных частот')
ax.set_xlabel('Середины интервалов')
ax.set_ylabel('Относительные частоты')
ax.grid()

# Гистограмма распределения относительных частот
ax = axes[1, 1]
for i in range(k):
    left = intervals_left[i]
    right = intervals_right[i]
    width = right - left
    height = frequencies_rel[i] / width
    ax.bar(left, height, width, align='edge', color='lightsteelblue', edgecolor='steelblue', linewidth=1.5)
ax.set_xticks(intervals_mid)
ax.set_xticklabels([f'{x:.1f}' for x in intervals_mid])
ax.set_title('Гистограмма относительных частот')
ax.set_xlabel('Середины интервалов')
ax.set_ylabel('Относительные частоты')
ax.grid()

# Эмпирическая функция распределения по относительным частотам
ax = axes[1, 2]
ax.step(x_step, y_step_rel, where='post', linewidth=2.5, color='darkorange')
ax.scatter(x_step[:-1], y_step_rel[:-1], color='darkorange', s=30, zorder=5)
ax.set_xticks(x_step)
ax.set_xticklabels([f'{x:.1f}' for x in x_step])
ax.set_title('Эмпирическая функция распределения (относительные частоты)')
ax.set_ylabel('Накопленная относительная частота')
ax.set_xlabel('Границы интервалов')
ax.grid()

plt.tight_layout()
plt.show()
```

```python
# Формируем таблицу интервального ряда
df_int = pd.DataFrame({
    'i': range(1, k + 1),
    'left': intervals_left,
    'right': intervals_right,
    'x_i': intervals_mid,
    'm_i': frequencies_abs
})

# Выбираем условный нуль C
modal_idx = df_int['m_i'].idxmax()
C = df_int.loc[modal_idx, 'x_i']
h = intervals_right[0] - intervals_left[0]

# Рассчитываем условные варианты u_i
df_int['u_i'] = ((df_int['x_i'] - C) / h).round().astype(int)
df_int['n_i*u_i'] = df_int['m_i'] * df_int['u_i']
df_int['n_i*u_i^2'] = df_int['m_i'] * (df_int['u_i'] ** 2)
df_int['n_i*u_i^3'] = df_int['m_i'] * (df_int['u_i'] ** 3)
df_int['n_i*u_i^4'] = df_int['m_i'] * (df_int['u_i'] ** 4)
df_int['n_i*(u_i+1)^4'] = df_int['m_i'] * ((df_int['u_i'] + 1) ** 4)

# Вычисляем итоговые суммы по расчётным столбцам
sum_n = df_int['m_i'].sum()
sum_n_u = df_int['n_i*u_i'].sum()
sum_n_u2 = df_int['n_i*u_i^2'].sum()
sum_n_u3 = df_int['n_i*u_i^3'].sum()
sum_n_u4 = df_int['n_i*u_i^4'].sum()
sum_n_u_plus1_4 = df_int['n_i*(u_i+1)^4'].sum()

sums_row = pd.DataFrame([{
    'i': 'Σ',
    'x_i': '-',
    'm_i': sum_n,
    'u_i': '-',
    'n_i*u_i': sum_n_u,
    'n_i*u_i^2': sum_n_u2,
    'n_i*u_i^3': sum_n_u3,
    'n_i*u_i^4': sum_n_u4,
    'n_i*(u_i+1)^4': sum_n_u_plus1_4
}])

df_with_sums = pd.concat([df_int, sums_row], ignore_index=True)

# Выводим расчётную таблицу
display_cols = [
    'i', 'x_i', 'm_i', 'u_i',
    'n_i*u_i', 'n_i*u_i^2', 'n_i*u_i^3',
    'n_i*u_i^4', 'n_i*(u_i+1)^4'
]

md('### Расчётная таблица')
md_table(df_with_sums[display_cols].rename(columns={'m_i': 'n_i'}), digits=4)

# Проверяем вычисления по контрольной формуле
control = sum_n + 4 * sum_n_u + 6 * sum_n_u2 + 4 * sum_n_u3 + sum_n_u4

control_df = pd.DataFrame({
    'Проверка': [
        'N + 4Σnu + 6Σnu² + 4Σnu³ + Σnu⁴',
        'Σn_i(u_i+1)^4'
    ],
    'Значение': [
        control,
        sum_n_u_plus1_4
    ]
})

md('### Контроль вычислений')
md_table(control_df, digits=4)

# Находим условные эмпирические моменты M_k*
M1 = sum_n_u / N
M2 = sum_n_u2 / N
M3 = sum_n_u3 / N
M4 = sum_n_u4 / N

# Вычисляем центральные эмпирические моменты m_k
m1_central = 0
m2_central = (M2 - M1**2) * (h**2)
m3_central = (M3 - 3 * M1 * M2 + 2 * (M1**3)) * (h**3)
m4_central = (M4 - 4 * M1 * M3 + 6 * (M1**2) * M2 - 3 * (M1**4)) * (h**4)

# Рассчитываем выборочные характеристики:
# среднее, дисперсию, среднее квадратическое отклонение,
# а также исправленные дисперсию и СКО
x_v = np.sum(df_int['x_i'] * df_int['m_i']) / N
D_v = np.sum(df_int['m_i'] * (df_int['x_i'] - x_v) ** 2) / N
sigma_v = np.sqrt(D_v)
s2 = (N / (N - 1)) * D_v
s = np.sqrt(s2)

# Определяем коэффициенты асимметрии и эксцесса
As = m3_central / (s ** 3)
E = m4_central / (s ** 4) - 3

# Находим моду интервального ряда
x_low_modal = df_int.loc[modal_idx, 'left']
f_modal = df_int.loc[modal_idx, 'm_i'] / N
f_prev_modal = df_int.loc[modal_idx - 1, 'm_i'] / N if modal_idx > 0 else 0
f_next_modal = df_int.loc[modal_idx + 1, 'm_i'] / N if modal_idx < k - 1 else 0
Mo = x_low_modal + h * (f_modal - f_prev_modal) / ((f_modal - f_prev_modal) + (f_modal - f_next_modal))

# Находим медиану интервального ряда
median_idx = df_int[df_int['m_i'].cumsum() >= N / 2].index[0]
x_low_med = df_int.loc[median_idx, 'left']
f_med = df_int.loc[median_idx, 'm_i'] / N
acc_prev_med = df_int.loc[:median_idx - 1, 'm_i'].sum() / N if median_idx > 0 else 0
Me = x_low_med + h * (0.5 - acc_prev_med) / f_med

# Вычисляем коэффициент вариации
V = (sigma_v / x_v) * 100

# Формируем итоговую таблицу с результатами расчётов
results = {
    'Параметр': [
        'Выборочное среднее (x_в)',
        'Выборочная дисперсия (D_в)',
        'Выборочное СКО (σ_в)',
        'Исправленная дисперсия (s²)',
        'Исправленное СКО (s)',
        'Коэффициент асимметрии (As)',
        'Коэффициент эксцесса (E)',
        'Мода (Mo)',
        'Медиана (Me)',
        'Коэффициент вариации (V*)'
    ],
    'Значение': [
        round(x_v, 2),
        round(D_v, 2),
        round(sigma_v, 4),
        round(s2, 2),
        round(s, 4),
        round(As, 2),
        round(E, 2),
        round(Mo, 2),
        round(Me, 2),
        round(V, 2)
    ]
}

results_df = pd.DataFrame(results)

md('### Итоговые результаты')
md_table(results_df, digits=4)
```

Коэффициент вариации равен **18,9%**, что говорит об **умеренной вариации признака**. Поскольку коэффициент вариации меньше **33%**, исследуемую совокупность можно считать **достаточно однородной**.


## 2 Задание

```python
# Извлекаем данные по двум признакам из таблицы raw
X_data = raw['nu'].dropna().values
Y_data = raw['E'].dropna().values

# Приводим оба массива к одинаковой длине
N = min(len(X_data), len(Y_data))
X_data = X_data[:N]
Y_data = Y_data[:N]

# Определяем число интервалов для каждого признака
k_X = int(1 + 3.31 * np.log10(N))
k_Y = int(1 + 3.31 * np.log10(N))

# Находим границы интервалов для признака X
X_min, X_max = X_data.min(), X_data.max()
h_X = (X_max - X_min) / k_X
X_intervals = [round(X_min + i * h_X, 2) for i in range(k_X + 1)]

# Находим границы интервалов для признака Y
Y_min, Y_max = Y_data.min(), Y_data.max()
h_Y = (Y_max - Y_min) / k_Y
Y_intervals = [round(Y_min + i * h_Y, 2) for i in range(k_Y + 1)]

# Создаём матрицу двумерного распределения частот
freq_matrix = np.zeros((k_Y, k_X), dtype=int)

for x_val, y_val in zip(X_data, Y_data):
    x_idx = k_X - 1
    for i in range(k_X):
        if i < k_X - 1:
            if X_intervals[i] <= x_val < X_intervals[i + 1]:
                x_idx = i
                break
        else:
            if X_intervals[i] <= x_val <= X_intervals[i + 1]:
                x_idx = i
                break

    y_idx = k_Y - 1
    for j in range(k_Y):
        if j < k_Y - 1:
            if Y_intervals[j] <= y_val < Y_intervals[j + 1]:
                y_idx = j
                break
        else:
            if Y_intervals[j] <= y_val <= Y_intervals[j + 1]:
                y_idx = j
                break

    freq_matrix[y_idx, x_idx] += 1

# Формируем подписи интервалов по оси X
x_headers = []
for i in range(k_X):
    if i < k_X - 1:
        x_headers.append(f"[{X_intervals[i]:.2f}, {X_intervals[i + 1]:.2f})")
    else:
        x_headers.append(f"[{X_intervals[i]:.2f}, {X_intervals[i + 1]:.2f}]")

# Формируем подписи интервалов по оси Y
y_headers = []
for j in range(k_Y):
    if j < k_Y - 1:
        y_headers.append(f"[{Y_intervals[j]:.2f}, {Y_intervals[j + 1]:.2f})")
    else:
        y_headers.append(f"[{Y_intervals[j]:.2f}, {Y_intervals[j + 1]:.2f}]")

# Подсчитываем суммы по строкам, столбцам и общее количество наблюдений
row_sums = np.sum(freq_matrix, axis=1)
col_sums = np.sum(freq_matrix, axis=0)
total_sum = np.sum(freq_matrix)

# Собираем двумерную таблицу частот
df_2d = pd.DataFrame(freq_matrix, index=y_headers, columns=x_headers)
df_2d.insert(0, 'Y \\ X', y_headers)
df_2d['Σ'] = row_sums

total_row = pd.DataFrame([['Σ'] + list(col_sums) + [total_sum]], columns=df_2d.columns)
df_2d_full = pd.concat([df_2d, total_row], ignore_index=True)

md('### Двумерная таблица распределения частот')
md_table(df_2d_full, digits=4)
```

## 3 Задание

```python
# Находим середины интервалов по X и Y
x_mid = [(X_intervals[i] + X_intervals[i + 1]) / 2 for i in range(k_X)]
y_mid = [(Y_intervals[i] + Y_intervals[i + 1]) / 2 for i in range(k_Y)]

# Собираем данные для корреляционной таблицы
rows = []

for j in range(k_Y):
    current_row = {'Y \\ X': round(y_mid[j], 2)}
    for i in range(k_X):
        value = freq_matrix[j, i]
        current_row[f'{x_mid[i]:.2f}'] = value if value > 0 else '-'
    current_row['Σ_y'] = row_sums[j]
    rows.append(current_row)

corr_df = pd.DataFrame(rows)

# Формируем строку с суммами по столбцам
total_row = {'Y \\ X': 'Σ_x'}
for i in range(k_X):
    total_row[f'{x_mid[i]:.2f}'] = col_sums[i]
total_row['Σ_y'] = total_sum

corr_df = pd.concat([corr_df, pd.DataFrame([total_row])], ignore_index=True)

md('### Корреляционная таблица')
md_table(corr_df, digits=4)
```

По корреляционной таблице можно заметить, что основная часть наблюдений сосредоточена в центральных интервалах обоих признаков: по $X$ наибольшие частоты приходятся на значения с серединами **417,50**, **456,50** и **495,50**, а по $Y$ — на интервалы с серединами **108,39**, **125,95** и **143,51**. Крайние сочетания значений встречаются существенно реже, что свидетельствует о концентрации данных вблизи средних уровней обоих признаков.

Распределение частот в таблице имеет выраженную диагональную направленность: при переходе к более высоким значениям $X$ в целом возрастают и значения $Y$. Наибольшие частоты сосредоточены в ячейках, расположенных вблизи главной диагонали, например $(417{,}50;,108{,}39)$, $(456{,}50;,125{,}95)$ и $(495{,}50;,143{,}51)$. Это указывает на наличие **положительной корреляционной зависимости** между рассматриваемыми признаками.

При этом зависимость не является строго функциональной, так как частоты распределены не только вдоль главной диагонали, но и в соседних ячейках. Следовательно, между признаками наблюдается **умеренная прямая связь**: с увеличением признака $X$ признак $Y$ также, как правило, возрастает, однако зависимость сопровождается заметным разбросом наблюдений.

## 4 Задание

```python
# Находим коэффициент корреляции по стандартной формуле

x_mid = np.array(x_mid)
y_mid = np.array(y_mid)
col_sums = np.array(col_sums)
row_sums = np.array(row_sums)

x_v = np.sum(x_mid * col_sums) / N
y_v = np.sum(y_mid * row_sums) / N

Sx = np.sqrt(np.sum(col_sums * (x_mid - x_v) ** 2) / N)
Sy = np.sqrt(np.sum(row_sums * (y_mid - y_v) ** 2) / N)

sum_nxy = 0
for i in range(k_X):
    for j in range(k_Y):
        sum_nxy += freq_matrix[j, i] * x_mid[i] * y_mid[j]

r_xy_standard = round((sum_nxy - N * x_v * y_v) / (N * Sx * Sy), 4)

corr_result_df = pd.DataFrame({
    'Показатель': [
        'Линейный коэффициент корреляции r_xy'
    ],
    'Значение': [
        r_xy_standard
    ]
})

md('### Результаты вычисления коэффициента корреляции')
md_table(corr_result_df, digits=4)
```

```python
# Находим коэффициент корреляции через условные варианты

Cx = x_mid[np.argmax(col_sums)]
Cy = y_mid[np.argmax(row_sums)]

u = (np.array(x_mid) - Cx) / h_X
v = (np.array(y_mid) - Cy) / h_Y

sum_nu = np.sum(col_sums * u)
sum_nv = np.sum(row_sums * v)

sum_nu2 = np.sum(col_sums * u**2)
sum_nv2 = np.sum(row_sums * v**2)

sum_nuv = 0
for i in range(k_X):
    for j in range(k_Y):
        sum_nuv += freq_matrix[j, i] * u[i] * v[j]

u_mean = sum_nu / N
v_mean = sum_nv / N

sigma_u = np.sqrt(sum_nu2 / N - u_mean**2)
sigma_v = np.sqrt(sum_nv2 / N - v_mean**2)

r_xy_uv = round((sum_nuv - N * u_mean * v_mean) / (N * sigma_u * sigma_v), 4)

corr_uv_df = pd.DataFrame({
    'Показатель': [
        'Линейный коэффициент корреляции r_xy'
    ],
    'Значение': [
        r_xy_uv
    ]
})

md('### Вычисление коэффициента корреляции через условные варианты')
md_table(corr_uv_df, digits=4)
```

Оба способа вычисления дали практически одинаковые результаты: по стандартной формуле получено $r_{xy} = 0{,}8833$, а методом условных вариант — $r_{xy} = 0{,}883$. Близость этих значений подтверждает правильность выполненных расчётов и свидетельствует о корректности проведённого анализа.

Полученное значение коэффициента корреляции указывает на **сильную положительную связь** между признаками. Это означает, что с увеличением признака $X$ значения признака $Y$ в большинстве случаев также возрастают.

Следовательно, результаты численного расчёта согласуются с анализом корреляционной таблицы: между признаками наблюдается выраженная **прямая корреляционная зависимость**.



## 5 Задание

```python
# Применяем преобразование Фишера к коэффициенту корреляции
z = 0.5 * np.log((1 + r_xy_standard) / (1 - r_xy_standard))

# Вычисляем СКО преобразованной величины z
sigma_z = 1 / np.sqrt(N - 3)

# Задаём значения доверительной вероятности
gamma_values = [0.95, 0.99]

# Формируем таблицу доверительных интервалов
rows = []
for gamma in gamma_values:
    lambda_val = stats.norm.ppf((1 + gamma) / 2)

    z_lower = z - lambda_val * sigma_z
    z_upper = z + lambda_val * sigma_z

    r_lower = np.tanh(z_lower)
    r_upper = np.tanh(z_upper)

    rows.append({
        'γ': gamma,
        'z': z,
        'σ_z': sigma_z,
        'λ': lambda_val,
        'Нижняя граница r_xy': r_lower,
        'Верхняя граница r_xy': r_upper
    })

fisher_df = pd.DataFrame(rows)

md('### Расчёт доверительных интервалов для коэффициента корреляции')
md_table(fisher_df, digits=4)
```

Оба полученных доверительных интервала полностью находятся в положительной области и включают достаточно высокие значения коэффициента корреляции. Это даёт основание утверждать, что связь между признаками является **устойчиво положительной** и достаточно тесной.

При уровне надёжности $\gamma = 0{,}95$ доверительный интервал имеет вид:

$$
0{,}8353 \leq r_{xy} \leq 0{,}9179
$$

а при уровне надёжности $\gamma = 0{,}99$:

$$
0{,}8168 \leq r_{xy} \leq 0{,}9266
$$

С увеличением уровня надёжности от **0,95** до **0,99** доверительный интервал расширяется, поскольку более высокая надёжность оценки требует охвата большего диапазона возможных значений коэффициента корреляции.

Поскольку оба интервала расположены далеко от нуля и полностью лежат в положительной области, можно сделать вывод о наличии **сильной положительной корреляционной связи** между признаками.



## 6 Задание

```python
# Устанавливаем уровень значимости
alpha = 0.05

# Вычисляем наблюдаемое значение t-критерия
T_obs = r_xy_standard * np.sqrt(N - 2) / np.sqrt(1 - r_xy_standard**2)

# Находим критическое значение t-распределения
t_crit = stats.t.ppf(1 - alpha / 2, df=N - 2)

# Составляем итоговую таблицу проверки гипотезы
hypothesis_df = pd.DataFrame({
    'Показатель': [
        'Уровень значимости α',
        'Наблюдаемое значение T_набл',
        'Критическое значение t_крит',
        'Вывод'
    ],
    'Значение': [
        alpha,
        T_obs,
        t_crit,
        'Гипотеза H0 отвергается, коэффициент корреляции значим'
        if abs(T_obs) > t_crit
        else 'Нет оснований отвергать H0, коэффициент корреляции незначим'
    ]
})

md('### Проверка статистической значимости коэффициента корреляции')
md_table(hypothesis_df, digits=4)
```

Поскольку наблюдаемое значение критерия $T_{\text{набл}}$ оказалось больше критического значения $t_{\text{крит}}$, нулевая гипотеза отвергается на заданном уровне значимости $\alpha$. Это означает, что полученный выборочный коэффициент корреляции является **статистически значимым**, а выявленная связь между признаками не может считаться случайной.

Следовательно, между исследуемыми признаками существует **линейная корреляционная зависимость**. При этом полученные результаты позволяют сделать вывод о наличии **умеренной прямой связи**: с увеличением одного признака другой в среднем также имеет тенденцию к возрастанию, хотя зависимость не является строго функциональной и сопровождается определённым разбросом наблюдений.
