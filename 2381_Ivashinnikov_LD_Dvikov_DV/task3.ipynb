## Основные теоретические положения

**Интервальная оценка** — оценка параметра Θ в виде интервала (Θ* − δ; Θ* + δ), где δ — точность, γ = P{|Θ* − Θ| < δ} — надёжность (доверительная вероятность). Уровень значимости q = 1 − γ.

Доверительный интервал для мат. ожидания (σ неизвестно)

При неизвестном σ величина $t = \frac{\bar{x} - a}{S / \sqrt{N}}$ распределена по закону Стьюдента с $k = N - 1$ степенями свободы. Доверительный интервал:

$$\left(\bar{x} - \frac{t_\gamma S}{\sqrt{N}};\ \bar{x} + \frac{t_\gamma S}{\sqrt{N}}\right)$$

где $t_\gamma$ определяется по таблице распределения Стьюдента (Приложение 6).

Доверительный интервал для СКО

Представив $\delta = qS$, доверительный интервал для σ:

$$(S(1 - q);\ S(1 + q))$$

где $q$ определяется по таблице (Приложение 7) для заданных γ и N.

Критерий Пирсона χ²

Гипотеза $H_0$: выборочные данные — значения нормально распределённой СВ. Наблюдаемое значение (правосторонний критерий):

$$\chi^2_{\text{набл}} = \sum_{i=1}^{K} \frac{(n_i - n'_i)^2}{n'_i}$$

Вычисление теоретических частот: $z_i = (x_i - \bar{x}) / S$ ($z_1 = -\infty$, $z_{K+1} = +\infty$), $p_i = \Phi(z_{i+1}) - \Phi(z_i)$ (Приложение 4), $n'_i = N \cdot p_i$.

Число степеней свободы: $k = K - 3$. Критическое значение $\chi^2_{\text{крит}} = \chi^2(\alpha, k)$ (Приложение 5).

Область принятия $H_0$: $\chi^2_{\text{набл}} \leq \chi^2_{\text{крит}}$.

Ошибки при проверке гипотез

Ошибка I рода (α) — отвергнуть верную $H_0$. Ошибка II рода (β) — принять неверную $H_0$. Мощность критерия — $1 - \beta$.

```python
import pandas as pd
import numpy as np
from scipy import stats

df = pd.read_csv('dataset.csv', sep=';', decimal='.')
print(f'Столбцы: {df.columns.tolist()}')
df.describe()
```

1. Вычислить точность и доверительный интервал для математического ожидания при неизвестном среднеквадратичном отклонении при заданном объёме выборки для доверительной точности $\gamma \in \{0.95, 0.99\}$. Сделать выводы.

```python
nu = df['nu (kg/cm^3)']
E = df['E (kg/cm^2)']

N = len(df)
k = N - 1    # число степеней свободы для распределения Стьюдента

# Выборочное среднее
x_mean_nu = nu.mean()
x_mean_E = E.mean()

# Исправленное выборочное СКО
S_nu = nu.std(ddof=1)
S_E = E.std(ddof=1)

print(f'N = {N}')
print(f'k = N - 1 = {k}')
print(f'nu (кг/см³)')
print(f'Выборочное среднее x̄ = {x_mean_nu:.4f}')
print(f'Исправленное СКО   S = {S_nu:.4f}')
print(f'E (кг/см²)')
print(f'Выборочное среднее x̄ = {x_mean_E:.4f}')
print(f'Исправленное СКО   S = {S_E:.4f}')
```

Нахождение квантилей распределения Стьюдента. Для двустороннего доверительного интервала с надёжностью γ квантиль $t_\gamma$ находится из условия:

$$P\left(|t| < t_\gamma\right) = \gamma \quad \Rightarrow \quad t_\gamma = t_{(1+\gamma)/2,\, k}$$

```python
gammas = [0.95, 0.99]

print(f'Число степеней свободы: k = {k}')
print()

for gamma in gammas:
    # Квантиль распределения Стьюдента
    t_gamma = stats.t.ppf((1 + gamma) / 2, df=k)
    print(f'γ = {gamma}: t_γ = t({(1+gamma)/2}, {k}) = {t_gamma:.4f}')
```

Вычисление точности δ и доверительных интервалов

Точность оценки:
$$\delta = \frac{t_\gamma \cdot S}{\sqrt{N}}$$

Доверительный интервал:
$$\left(\bar{x} - \delta;\; \bar{x} + \delta\right)$$

```python
# Функция для вычисления доверительного интервала при неизвестном СКО. 
def confidence_interval(x_mean, S, N, gamma):
    k = N - 1
    t_gamma = stats.t.ppf((1 + gamma) / 2, df=k)
    delta = t_gamma * S / np.sqrt(N)
    left = x_mean - delta
    right = x_mean + delta
    return t_gamma, delta, left, right
```

```python
print(f'Параметр: nu (кг/см³)')
print(f'x̄ = {x_mean_nu:.4f}, S = {S_nu:.4f}, N = {N}, k = {k}')
print('=' * 65)

for gamma in gammas:
    t_g, delta, left, right = confidence_interval(x_mean_nu, S_nu, N, gamma)
    
    print(f'\nγ = {gamma}:')
    print(f'  t_γ = {t_g:.4f}')
    print(f'  δ = t_γ · S / √N = {t_g:.4f} · {S_nu:.4f} / √{N} = {delta:.4f}')
    print(f'  Доверительный интервал: ({left:.4f}; {right:.4f})')

print(f'Параметр: E (кг/см²)')
print(f'x̄ = {x_mean_E:.4f}, S = {S_E:.4f}, N = {N}, k = {k}')
print('=' * 65)

for gamma in gammas:
    t_g, delta, left, right = confidence_interval(x_mean_E, S_E, N, gamma)
    
    print(f'\nγ = {gamma}:')
    print(f'  t_γ = {t_g:.4f}')
    print(f'  δ = t_γ · S / √N = {t_g:.4f} · {S_E:.4f} / √{N} = {delta:.4f}')
    print(f'  Доверительный интервал: ({left:.4f}; {right:.4f})')
```

Вывод: Доверительные интервалы для математического ожидания построены с использованием распределения Стьюдента ($k = N - 1 = 108$). Полученные значения $t_\gamma$ согласуются с табличными (Приложение 6).

При увеличении надёжности с $\gamma = 0.95$ до $\gamma = 0.99$ точность оценки снижается ($\delta$ возрастает), что приводит к расширению доверительного интервала. Это соответствует теоретическому положению: повышение доверительной вероятности возможно только за счёт увеличения ширины интервала при фиксированном объёме выборки.

3. Проверить гипотезу о нормальности заданного распределения с помощью критерия $\chi^2$ (Пирсона). Для этого необходимо найти теоретические частоты и вычислить наблюдаемое значение критерия. Для удобства вычисления необходимо заполнить табл. 1.

```python
from scipy.stats import norm

num_intervals = round(1 + 3.322 * np.log10(N))
print(f'Число интервалов (Стёрджес): K = {num_intervals}')


def laplace_func(val):
    if val == -np.inf:
        return -0.5
    if val == np.inf:
        return 0.5
    return norm.cdf(val) - 0.5


def pearson_test(sample, mean_val, std_val, size, num_bins, feature):
    edges = np.linspace(sample.min(), sample.max(), num_bins + 1)

    # стандартизация: z_i = (x_i - x_mean) / S, z_1 = -inf, z_{K+1} = +inf
    z_bounds = np.array([(edges[i] - mean_val) / std_val for i in range(num_bins + 1)])
    z_bounds[0], z_bounds[-1] = -np.inf, np.inf

    # вероятности через функцию Лапласа
    probs = np.array([laplace_func(z_bounds[i + 1]) - laplace_func(z_bounds[i])
                       for i in range(num_bins)])

    # теоретические и наблюденные частоты
    expected = size * probs
    observed = np.histogram(sample, bins=edges)[0].astype(float)

    # хи-квадрат
    residuals = (observed - expected) ** 2 / expected
    chi2_result = residuals.sum()

    # подписи интервалов
    labels = []
    for i in range(num_bins):
        lo = '(-inf' if i == 0 else f'[{edges[i]:.4f}'
        hi = '+inf)' if i == num_bins - 1 else f'{edges[i + 1]:.4f})'
        labels.append(f'{lo}, {hi}')

    # датафрейм
    rows = {
        'i': list(range(1, num_bins + 1)) + ['sum'],
        '[xi, xi+1)': labels + [''],
        'ni': list(observed.astype(int)) + [int(observed.sum())],
        'pi': list(np.round(probs, 4)) + [round(probs.sum(), 4)],
        "n'i": list(np.round(expected, 4)) + [round(expected.sum(), 4)],
        "(ni - n'i)^2": list(np.round((observed - expected) ** 2, 4)) + ['-'],
        "(ni-n'i)^2/n'i": list(np.round(residuals, 4)) + [round(chi2_result, 4)],
        'ni^2': list((observed ** 2).astype(int)) + ['-'],
        "ni^2/n'i": list(np.round(observed ** 2 / expected, 4)) + [round((observed ** 2 / expected).sum(), 4)],
    }
    result_table = pd.DataFrame(rows)

    print(f'Признак: {feature}')
    print(f'x_mean = {mean_val:.4f}, S = {std_val:.4f}, N = {size}, K = {num_bins}')
    print(f'Хи-квадрат наблюдаемое = {chi2_result:.4f}')

    # критическое значение
    deg_free = num_bins - 3
    chi2_crit = stats.chi2.ppf(1 - 0.05, df=deg_free)

    print(f'df = K - 3 = {num_bins} - 3 = {deg_free}')
    print(f'Хи-квадрат критическое = {chi2_crit:.4f}')

    if chi2_result <= chi2_crit:
        print(f'{chi2_result:.4f} <= {chi2_crit:.4f} -> H0 принимается\n')
    else:
        print(f'{chi2_result:.4f} > {chi2_crit:.4f} -> H0 отвергается\n')

    return observed, expected, result_table
```

```python
obs_nu, exp_nu, tbl_nu = pearson_test(nu.values, x_mean_nu, S_nu, N, num_intervals, 'nu (kg/cm^3)')
tbl_nu
```

```python
obs_e, exp_e, tbl_e = pearson_test(E.values, x_mean_E, S_E, N, num_intervals, 'E (kg/cm^2)')
tbl_e
```

4. Доказать, что
$$\chi^2_{\text{набл}} = \sum_{i=1}^{k} \frac{n_i^2}{n'_i} - n.$$
Проконтролировать корректность вычисления $\chi^2_{\text{набл}}$.

Раскроем скобки в формуле $\chi^2_{\text{набл}}$:

$$\chi^2_{\text{набл}} = \sum_{i=1}^{K} \frac{(n_i - n'_i)^2}{n'_i} = \sum_{i=1}^{K} \frac{n_i^2 - 2n_i n'_i + n'^2_i}{n'_i} = \sum_{i=1}^{K}\left(\frac{n_i^2}{n'_i} - 2n_i + n'_i\right)$$

$$= \sum_{i=1}^{K}\frac{n_i^2}{n'_i} - 2\sum_{i=1}^{K}n_i + \sum_{i=1}^{K}n'_i$$

Так как $\sum_{i=1}^{K} n_i = N$ и $\sum_{i=1}^{K} n'_i = N$, получаем:

$$\chi^2_{\text{набл}} = \sum_{i=1}^{K}\frac{n_i^2}{n'_i} - 2N + N = \sum_{i=1}^{K}\frac{n_i^2}{n'_i} - N$$

Что и требовалось доказать.

```python
for tag, obs, exp in [('nu', obs_nu, exp_nu), ('E', obs_e, exp_e)]:
    formula_1 = np.sum((obs - exp) ** 2 / exp)
    formula_2 = np.sum(obs ** 2 / exp) - N

    print(f'Признак {tag}:')
    print(f'  Основная формула:        {formula_1:.4f}')
    print(f'  Преобразованная формула:  {formula_2:.4f}')
    print(f'  Совпадение: {np.isclose(formula_1, formula_2)}')
    print()
```

Для обоих признаков результаты вычисления $\chi^2_{\text{набл}}$ по основной и преобразованной формулам совпадают: для $\nu$ — $5.0916 = 5.0916$, для $E$ — $7.6029 = 7.6029$. Это подтверждает корректность вычислений и справедливость доказанного тождества $\chi^2_{\text{набл}} = \sum \frac{n_i^2}{n'_i} - N$.

5. По заданному уровню значимости $\alpha = 0.05$ и числу степеней свободы $df$ найти критическую точку $\chi^2_{\text{крит}}$ и сравнить с наблюдаемым значением. Сделать выводы.

По заданному уровню значимости $\alpha = 0.05$ и числу степеней свободы $df = K - 3$ находим критическую точку $\chi^2_{\text{крит}}$ и сравниваем с наблюдаемым значением. Если $\chi^2_{\text{набл}} \leq \chi^2_{\text{крит}}$ — гипотеза $H_0$ принимается, иначе — отвергается.

```python
alpha = 0.05
for tag, obs, exp in [('nu', obs_nu, exp_nu), ('E', obs_e, exp_e)]:
    chi2_obs = np.sum((obs - exp) ** 2 / exp)
    deg_free = len(obs) - 3
    chi2_crit = stats.chi2.ppf(1 - alpha, df=deg_free)

    print(f'Признак {tag}:')
    print(f'  alpha = {alpha}, df = K - 3 = {len(obs)} - 3 = {deg_free}')
    print(f'  Хи-квадрат наблюдаемое  = {chi2_obs:.4f}')
    print(f'  Хи-квадрат критическое  = {chi2_crit:.4f}')

    if chi2_obs <= chi2_crit:
        print(f'  {chi2_obs:.4f} <= {chi2_crit:.4f} -> H0 принимается')
    else:
        print(f'  {chi2_obs:.4f} > {chi2_crit:.4f} -> H0 отвергается')
    print()
```

Для обоих признаков при уровне значимости $\alpha = 0.05$ и числе степеней свободы $df = K - 3 = 5$ наблюдаемые значения не превышают критическое: для $\nu$ — $5.09 \leq 11.07$, для $E$ — $7.60 \leq 11.07$. Гипотеза $H_0$ о нормальном распределении принимается для обоих признаков. Расхождения между эмпирическими и теоретическими частотами статистически незначимы.
