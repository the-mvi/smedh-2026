## Основные теоретические положения

Начальный эмпирический момент k-го порядка:
$$\bar M_k = \frac{1}{N}\sum n_jx^k_j$$

Центральный эмпирический момент k-го порядка:
$$\bar m_k = \frac{1}{N}n_j(x_j-\bar x_в)^k$$, где $\bar x_в = \bar M_1$.

Выборочная дисперсия:
$$D_в=\bar m_2=\frac{1}{N}\sum n_j(x_j-\bar x_в)^2$$

Исправленная дисперсия:
$$s^2=\frac{N}{N-1}D_в$$

Выборочное СКО:
$$\sigma_в=\sqrt{D_в}$$

Исправленное СКО:
$$s=\sqrt{s^2}$$

Коэффициент асимметрии:
$$\bar A_s = \frac{\bar m_3}{s^3}$$

Коэффициент эксцесса:
$$\bar E = \frac{\bar m_4}{s^4}-3$$

Мода – это значение, которое встречается в наборе данных чаще всего (самое популярное значение или категория).

Медиана — это серединное значение в упорядоченном наборе данных (выборке), которое делит его ровно на две части: 50% значений меньше неё, 50% — больше. Она устойчива к выбросам (резким отклонениям) поэтому лучше описывает «типичный» показатель чем среднее арифметическое.

Коэффициент вариации – статистический показатель, определяющий относительный разброс (дисперсию) данных вокруг среднего значения. Он рассчитывается как отношение среднеквадратичного отклонения (𝜎) к среднему арифметическому (x̄), выраженное в процентах. Позволяет сравнивать неоднородность наборов данных с разными единицами измерения или масштабами.

## Выполнение работы

```python
import pandas as pd


NU_HEADER, E_HEADER = 'nu (kg/cm^3)', 'E (kg/cm^2)'

df = pd.read_csv('/content/dataset.csv', sep=';')
df.describe()
```

```python
import numpy as np


INTERVAL = r'$[x_i,x_{i+1})$'
MIDRANGE = r'$\tilde x_i$'
FREQUENCY = r'$m_i$'
RELATIVE_FREQUENCY = r'$\tilde m_i$'
FREQUENCY_CUM_SUM = r'$m_i^{нак}$'
RELATIVE_FREQUENCY_CUM_SUM = r'$\tilde m_i^{нак}$'
SUM = r'$\sum$'


def sturges_interval_series(s: pd.Series) -> pd.DataFrame:
  n = len(s)
  k = int(1 + 3.322 * np.log10(n))

  intervals = pd.cut(s, bins=k, right=False)

  table = intervals.value_counts().sort_index().reset_index()

  table.columns = [INTERVAL, FREQUENCY]
  table[MIDRANGE] = table[INTERVAL].apply(lambda x: (x.left + x.right) / 2)
  table[RELATIVE_FREQUENCY] = table[FREQUENCY] / n
  table[FREQUENCY_CUM_SUM] = table[FREQUENCY].cumsum()
  table[RELATIVE_FREQUENCY_CUM_SUM] = table[RELATIVE_FREQUENCY].cumsum()

  table.index = range(1, len(table) + 1)

  sum_row = pd.DataFrame({
      INTERVAL: [''],
      MIDRANGE: [''],
      FREQUENCY: [table[FREQUENCY].sum()],
      RELATIVE_FREQUENCY: [table[RELATIVE_FREQUENCY].sum()],
      FREQUENCY_CUM_SUM: [''],
      RELATIVE_FREQUENCY_CUM_SUM: ['']
  }, index=[SUM])

  table = pd.concat([table, sum_row])

  return table[
      [
          INTERVAL,
          MIDRANGE,
          FREQUENCY,
          RELATIVE_FREQUENCY,
          FREQUENCY_CUM_SUM,
          RELATIVE_FREQUENCY_CUM_SUM
      ]
  ]

# Интервальный ряд
interval_nu, interval_e = sturges_interval_series(df[NU_HEADER]), sturges_interval_series(df[E_HEADER])
```

1. Для середин интервального ряда, полученного в практической работе №1, вычислить условные варианты. Заполнить табл. 1 (в последней строке Σ необходимо заполнить суммы столбцов; ячейки отмеченные прочерком заполнять не надо). Провести контроль вычислений.

```python
import numpy as np
import pandas as pd


X = r'$x_i$'
N = r'$n_i$'
U = r'$u_i$'
NU = r'$n_iu_i$'
NU2 = r'$n_iu_i^2$'
NU3 = r'$n_iu_i^3$'
NU4 = r'$n_iu_i^4$'
NUPLUS14 = r'$n_i(u_i+1)^4$'
mid = 3


def build_variant(table: pd.DataFrame) -> pd.DataFrame:
  midranges = table[MIDRANGE][:-1].astype(float)

  idx_modal = table[FREQUENCY][:-1].astype(int).values.argmax()
  c = midranges.iloc[idx_modal]
  h = table[INTERVAL].iloc[0].length
  u = np.round((midranges - c) / h).astype(int)
  mid = idx_modal

  nu = table[FREQUENCY][:-1] * u
  nu2 = table[FREQUENCY][:-1] * (u ** 2)
  nu3 = table[FREQUENCY][:-1] * (u ** 3)
  nu4 = table[FREQUENCY][:-1] * (u ** 4)
  nuplus14 = table[FREQUENCY][:-1] * ((u+1) ** 4)

  result = pd.DataFrame({
      X: midranges,
      N: table[FREQUENCY][:-1],
      U: u,
      NU: nu,
      NU2: nu2,
      NU3: nu3,
      NU4: nu4,
      NUPLUS14: nuplus14
  })

  sum_row = pd.DataFrame({
      X: [''],
      N: [result[N].sum()],
      U: [''],
      NU: [result[NU].sum()],
      NU2: [result[NU2].sum()],
      NU3: [result[NU3].sum()],
      NU4: [result[NU4].sum()],
      NUPLUS14: [result[NUPLUS14].sum()],
  }, index=[SUM])

  return pd.concat([result, sum_row]).rename_axis('i')


variant_nu, variant_e = build_variant(interval_nu), build_variant(interval_e)
```

2. Вычислить условные эмпирические моменты $\bar M^*_k$ через условные варианты. С помощью условных эмпирических моментов вычислить центральные эмпирические моменты $\bar m_k$. Полученные результаты занести в табл. 2.

```python
MOMENT = r'$\bar M_k^*$'
CENTRAL_MOMENT = r'$\bar m_k$'


def build_moments(table: pd.DataFrame, h: float) -> pd.DataFrame:
  m = table[[NU, NU2, NU3, NU4]].loc[SUM] / table[N].loc[SUM]
  central_m = [
    0,
    (m[NU2]-m[NU]**2) * (h**2),
    (m[NU3]-3*m[NU2]*m[NU]+2*(m[NU]**3)) * (h**3),
    (m[NU4]-4*m[NU3]*m[NU]+6*m[NU2]*(m[NU]**2)-3*(m[NU]**4)) * (h**4)
  ]

  result = pd.DataFrame({
      MOMENT: m.to_list(),
      CENTRAL_MOMENT: central_m
  }, index=list(range(1, len(m)+1)))

  return result.rename_axis('k')


moment_nu = build_moments(variant_nu, interval_nu[INTERVAL][1].length)
moment_e = build_moments(variant_e, interval_e[INTERVAL][1].length)
```

3. Вычислить выборочные среднее $\bar x_в$ и дисперсию $D_в$ с помощью стандартной формулы и с помощью условных вариант. Убедиться, что результаты совпадают. Вычислить выборочное СКО $\sigma_в$.

```python
nu_mean = (variant_nu[X][:-1] * variant_nu[N][:-1]).sum() / variant_nu[N].loc[SUM]
e_mean = (variant_e[X][:-1] * variant_e[N][:-1]).sum() / variant_e[N].loc[SUM]
print(f'nu_mean={np.round(nu_mean, 4)}, e_mean={np.round(e_mean, 4)}')

nu_variance = (variant_nu[N][:-1]*(variant_nu[X][:-1]-nu_mean)**2).sum() / variant_nu[N].loc[SUM]
e_variance = (variant_e[N][:-1]*(variant_e[X][:-1]-e_mean)**2).sum() / variant_e[N].loc[SUM]
print(f'nu_variance={np.round(nu_variance, 4)}, e_variance={np.round(e_variance, 4)}')
```

```python
nu_mean_cond = interval_nu[INTERVAL].loc[1].length * moment_nu[MOMENT].loc[1] + interval_nu[MIDRANGE].loc[mid]
e_mean_cond = interval_e[INTERVAL].loc[1].length * moment_e[MOMENT].loc[1] + interval_e[MIDRANGE].loc[mid]
print(f'Условные варианты: nu_mean={np.round(nu_mean_cond, 4)}, e_mean={np.round(e_mean_cond, 4)}')

nu_variance_cond = (interval_nu[INTERVAL].loc[1].length ** 2) * (moment_nu[MOMENT].loc[2] - moment_nu[MOMENT].loc[1] ** 2)
e_variance_cond = (interval_e[INTERVAL].loc[1].length ** 2) * (moment_e[MOMENT].loc[2] - moment_e[MOMENT].loc[1] ** 2)
print(f'Условная дисперсия: nu_variance={np.round(nu_variance_cond, 4)}, e_variance={np.round(e_variance_cond, 4)}')
```

Значения почти одинаковые. Ошибки возникают при округлении во время вычислений.

```python
sigma_v_nu = np.sqrt(nu_variance)
sigma_v_e = np.sqrt(e_variance)
print(f'Выборочное СКО: nu={np.round(sigma_v_nu, 4)}, e={np.round(sigma_v_e, 4)}')
```

4. Вычислить исправленную выборочную дисперсию $s^2$ и исправленное СКО $s$. Сравнить данные оценки со смещёнными оценками дисперсии и СКО. Сделать выводы.

```python
nu_variance_fixed = (variant_nu[N].loc[SUM] / (variant_nu[N].loc[SUM] - 1)) * nu_variance
e_variance_fixed = (variant_e[N].loc[SUM] / (variant_e[N].loc[SUM] - 1)) * e_variance

print(f'Исправленная дисперсия: nu={np.round(nu_variance_fixed, 4)}, e={np.round(e_variance_fixed, 4)}')
```

```python
sigma_v_fixed_nu = np.sqrt(nu_variance_fixed)
sigma_v_fixed_e = np.sqrt(e_variance_fixed)
print(f'Исправленное СКО: nu={np.round(sigma_v_fixed_nu, 4)}, e={np.round(sigma_v_fixed_e, 4)}')
```

Исправленная дисперсия получилась больше смещённой (в случае nu это не сильно заметно). Это связано с тем, что при вычислении смещённой дисперсии деление выполняется на $N$, следовательно, разброс данных немного уменьшается. Деление на $N-1$ позволяет компенсировать возникающее занижение оценки, поэтому исправленная дисперсия получается больше и точнее отражает разброс данных в выборке.

СКО также получилось немного больше, что логично, если знать как он считается.

5. Найти статистическую оценку коэффициентов асимметрии $\bar A_s$ и эксцесса $\bar E$. Сделать выводы.

```python
asymmetry_nu = moment_nu[CENTRAL_MOMENT].loc[3] / (sigma_v_fixed_nu ** 3)
asymmetry_e = moment_e[CENTRAL_MOMENT].loc[3] / (sigma_v_fixed_e ** 3)
print(f'Коэффициент асимметрии: nu={asymmetry_nu:.2f}, e={asymmetry_e:.2f}')
```

```python
excess_nu = moment_nu[CENTRAL_MOMENT].loc[4] / (sigma_v_fixed_nu ** 4) - 3
excess_e = moment_e[CENTRAL_MOMENT].loc[4] / (sigma_v_fixed_e ** 4) - 3
print(f'Коэффициент эксцесса: nu={excess_nu:.2f}, e={excess_e:.2f}')
```

Отрицательный коэффициент асимметрии указывает на левостороннюю асимметрию (длинный «хвост» слева).

Отрицательное значение коэффициента эксцесса указывает на плосковершинность распределения по сравнению с нормальным. Вершина выражена менее явно, а данные распределены более равномерно вокруг среднего.

6. Для интервального ряда вычислить моду $M_о^*$, медиану $M_e^*$ и коэффициент вариации $V^*$ заданного распределения. Сделать выводы.

```python
modal_idx_nu = interval_nu[FREQUENCY][:-1].idxmax()
modal_interval_nu = interval_nu.loc[modal_idx_nu]

modal_idx_e = interval_e[FREQUENCY][:-1].idxmax()
modal_interval_e = interval_e.loc[modal_idx_e]

x_low_nu = modal_interval_nu[INTERVAL].left
f_mod_int_nu = modal_interval_nu[RELATIVE_FREQUENCY]
f_prev_mod_int_nu = interval_nu.loc[modal_idx_nu - 1, RELATIVE_FREQUENCY] if modal_idx_nu > 0 else 0
f_next_mod_int_nu = interval_nu.loc[modal_idx_nu + 1, RELATIVE_FREQUENCY] if modal_idx_nu < len(interval_nu) - 1 - 1 else 0

x_low_e = modal_interval_e[INTERVAL].left
f_mod_int_e = modal_interval_e[RELATIVE_FREQUENCY]
f_prev_mod_int_e = interval_e.loc[modal_idx_e - 1, RELATIVE_FREQUENCY] if modal_idx_e > 0 else 0
f_next_mod_int_e = interval_e.loc[modal_idx_e + 1, RELATIVE_FREQUENCY] if modal_idx_e < len(interval_e) - 1 - 1 else 0

Mo_nu = x_low_nu + interval_nu[INTERVAL].loc[1].length * (f_mod_int_nu - f_prev_mod_int_nu) / ((f_mod_int_nu - f_prev_mod_int_nu) + (f_mod_int_nu - f_next_mod_int_nu))
Mo_e = x_low_e + interval_e[INTERVAL].loc[1].length * (f_mod_int_e - f_prev_mod_int_e) / ((f_mod_int_e - f_prev_mod_int_e) + (f_mod_int_e - f_next_mod_int_e))
print(f'Мода: nu={Mo_nu:.4f}, e={Mo_e:.4f}')
```

```python
median_idx_nu = interval_nu[:-1][interval_nu[FREQUENCY_CUM_SUM][:-1] >= interval_nu[FREQUENCY].loc[SUM] / 2].index[0]
median_interval_nu = interval_nu.loc[median_idx_nu]

median_idx_e = interval_e[:-1][interval_e[FREQUENCY_CUM_SUM][:-1] >= interval_e[FREQUENCY].loc[SUM] / 2].index[0]
median_interval_e = interval_e.loc[median_idx_e]

x_low_med_nu = median_interval_nu[INTERVAL].left
f_med_int_nu = median_interval_nu[RELATIVE_FREQUENCY]
acc_prev_med_int_nu = interval_nu.loc[median_idx_nu - 1, RELATIVE_FREQUENCY_CUM_SUM] if median_idx_nu > 0 else 0
Me_nu = x_low_med_nu + interval_nu[INTERVAL].loc[1].length * (0.5 - acc_prev_med_int_nu) / f_med_int_nu

x_low_med_e = median_interval_e[INTERVAL].left
f_med_int_e = median_interval_e[RELATIVE_FREQUENCY]
acc_prev_med_int_e = interval_e.loc[median_idx_e - 1, RELATIVE_FREQUENCY_CUM_SUM] if median_idx_e > 0 else 0
Me_e = x_low_med_e + interval_e[INTERVAL].loc[1].length * (0.5 - acc_prev_med_int_e) / f_med_int_e

print(f'Медиана: nu={Me_nu:.4f}, e={Me_e:.4f}')
```

```python
variation_nu = (sigma_v_nu / nu_mean) * 100
variation_e = (sigma_v_e / e_mean) * 100

print(f'Коэффициент вариации: nu={variation_nu:.4f}, e={variation_e:.4f}')
```

Значение моды больше, чем значение медианы. Это показывает, что данные имеют левостороннюю асимметрию, что было показано также в п. 5.

Коэффициент вариации показывает относительный разброс данных вокруг среднего значения. Для обоих признаков показывает среднюю (умеренную) вариацию.
