## Основные теоретические положения

**Интервальная оценка параметра** – оценка, определяемая двумя числами: границами интервала.

Пусть $\theta^*$ – оценка параметра $\theta$. Чем меньше $|\theta - \theta^*|$, тем точнее оценка $\theta^*$. Тогда некоторая величина $\delta > |\theta - \theta^*|$ характеризует **точность** оценки.

**Надёжность (доверительная вероятность)** оценки $\theta$ по $\theta^*$ – вероятность $\gamma$, при которой $|\theta - \theta^*| < \delta$. Иными словами, $\gamma = P(|\theta - \theta^*| < \delta)$, т. е.
$$
\gamma = P(\theta^* - \delta < \theta < \theta^* + \delta)
$$

**Доверительный интервал** $(\theta^* - \delta;\; \theta^* + \delta)$ – интервал, покрывающий параметр $\theta$ с заданной надёжностью $\gamma$.



**Статистическая гипотеза** – предположение о генеральной совокупности, проверяемое по выборке.

**Статистический критерий** – критерий, по которому принимается или опровергается гипотеза.

Пусть $H_0$ – нулевая (изначальная) гипотеза, а $H_1$ – альтернативная (конкурирующая) гипотеза. Тогда $\alpha$ (**уровень значимости критерия**) – вероятность принять альтернативную гипотезу (или отвергнуть нулевую):
$$
\alpha = P(H_1/H_0)
$$

**Эмпирические частоты** – абсолютные частоты $(m_i)$, наблюдаемые в эксперименте.

**Точечная вероятность** – вероятность того, что $X \in (x_{i-1};\, x_i]$, которая в случае нмального распределения рассчитывается следующим образом:
$$
p_i = P(x_{i-1} < X < x_i) = \Phi\!\left(\frac{x_i - a}{\sigma}\right) - \Phi\!\left(\frac{x_{i-1} - a}{\sigma}\right)
$$
где $a$ – выборочное среднее, $\sigma$ – выборочное СКО (или исправленное СКО), а $\Phi(t)$ – значение функции Лапласа.

**Выравнивающая (теоретическая) частота** – величина, которая находится по формуле:
$$
m'_i = n p_i
$$
где $n$ – объём выборки, а $p_i$ – точечная вероятность.

**Статистика $\chi^2_{\text{набл.}}$** – величина, которая вычисляется следующим образом:
$$
\chi^2_{\text{набл.}} = \sum_{i=1}^{k} \frac{(m_i - m'_i)^2}{m'_i}
$$
где $k$ – количество интервалов.

**Число степеней свободы** – величина, которая вычисляется так:
$$
df = k - r - 1
$$
где $r$ – число параметров дифференциальной функции распределения. Для нормального распределения $r = 2$, поэтому:
$$
df = k - 3
$$

**Критерий $\chi^2$ (Пирсона)** – условие принятия гипотезы о том, что данная случайная величина $X$ подчиняется нормальному распределению. Этот критерий выражается следующим неравенством:
$$
\chi^2_{\text{набл.}} < \chi^2_{\text{крит.}}
$$
где $\chi^2_{\text{набл.}}$ – статистика, а $\chi^2_{\text{крит.}}$ – величина, зависящая от уровня значимости $\alpha$ и числа степеней свободы $df$. Значения $\chi^2_{\text{крит.}}$ табулированы.

Если данное неравенство выполняется, то закон теоретического распределения (в нашем случае – нормального) не противоречит опытным данным и нет оснований отвергать гипотезу $H_0$. В противном случае нулевая гипотеза отвергается.

## Постановка задачи
Для заданной надежности определить (на основании выборочных данных и результатов выполнения практической работы №2) границы доверительных интервалов для математического ожидания и среднеквадратичного отклонения случайной величины. Проверить гипотезу о нормальном распределении исследуемой случайной величины с помощью критерия Пирсона $\chi^2$. Дать содержательную интерпретацию полученным результатам.


## Выполнение работы

```python
import numpy as np
import pandas as pd
import scipy.stats as st
from IPython.display import display


df = pd.read_csv('sample.csv', comment='#')
data = df['nu'].values
n = len(data)

k = 8
counts_int, bins = np.histogram(data, bins=k)
h = bins[1] - bins[0]
midpoints = (bins[:-1] + bins[1:]) / 2


x_bar = np.sum(midpoints * counts_int) / n
D_v = np.sum(((midpoints - x_bar)**2) * counts_int) / n
s = np.sqrt((n / (n - 1)) * D_v)

print(f"Объем выборки: {n}")
print(f"Количество интервалов: {k}")
print(f"Выборочное среднее: {x_bar:.4f}")
print(f"Исправленное СКО:  {s:.4f}")
```

### Задание 1


Доверительный интервал для математического ожидания с надёжностью $\gamma = 1-\alpha$ строится c помощью распределения Стьюдента:
$$
\left( \bar{X} - t_{\alpha/2}(n-1)\,\frac{S}{\sqrt{n}},\; \bar{X} + t_{\alpha/2}(n-1)\,\frac{S}{\sqrt{n}} \right),
$$
где $t_{\alpha/2}(n-1)$ — квантиль уровня $1-\alpha/2$ распределения Стьюдента с $n-1$ степенью свободы.


```python
print("--- 1. ДОВЕРИТЕЛЬНЫЕ ИНТЕРВАЛЫ ДЛЯ МАТЕМАТИЧЕСКОГО ОЖИДАНИЯ ---\n")
for gamma in [0.95, 0.99]:
    alpha = 1 - gamma

    t_crit = st.t.ppf(1 - alpha/2, df=n-1)


    margin = t_crit * (s / np.sqrt(n))

    print(f"Надежность γ = {gamma}:")
    print(f"  Точность оценки (погрешность) = {margin:.4f}")
    print(f"  Доверительный интервал: ({x_bar - margin:.4f} < a < {x_bar + margin:.4f})\n")
```

Как можно заметить, при увеличении доверительной вероятности, доверительный интервал также увеличивается, что соответствует интуитивной модели: в больший интервал вероятность попадания выше.

### Задание 2


Доверительный интервал для дисперсии $\sigma^2$ с надёжностью $\gamma = 1-\alpha$ строится по формуле

$$\left( \frac{(n-1)S^2}{\chi^2_{1-\alpha/2}(n-1)},\; \frac{(n-1)S^2}{\chi^2_{\alpha/2}(n-1)} \right),$$

где $\chi^2_{p}(k)$ — квантиль уровня $p$ распределения $\chi^2$ с $k$ степенями свободы.

Интервал для самого среднеквадратического отклонения $\sigma$ получается извлечением квадратного корня из границ интервала для дисперсии:
$$
\left( \sqrt{\frac{(n-1)S^2}{\chi^2_{1-\alpha/2}(n-1)}},\; \sqrt{\frac{(n-1)S^2}{\chi^2_{\alpha/2}(n-1)}} \right).
$$

```python
print("--- 2. ДОВЕРИТЕЛЬНЫЕ ИНТЕРВАЛЫ ДЛЯ СКО ---\n")
for gamma in [0.95, 0.99]:
    alpha = 1 - gamma

    chi2_left = st.chi2.ppf(alpha/2, df=n-1)
    chi2_right = st.chi2.ppf(1 - alpha/2, df=n-1)


    lower_bound = s * np.sqrt((n - 1) / chi2_right)
    upper_bound = s * np.sqrt((n - 1) / chi2_left)


    q_approx = upper_bound / s - 1

    print(f"Надежность γ = {gamma}:")
    print(f"  Табличное значение q ≈ {q_approx:.4f}")
    print(f"  Доверительный интервал: ({lower_bound:.4f} < σ < {upper_bound:.4f})\n")
```

Здесь также наблюдается расширение доверительного интервала при увеличении надёжности.

### Задание 3


```python
print("--- 3. РАСЧЕТНАЯ ТАБЛИЦА (КРИТЕРИЙ ПИРСОНА) ---\n")

z_bounds = (bins - x_bar) / s
z_bounds[0] = -np.inf
z_bounds[-1] = np.inf

P_i = st.norm.cdf(z_bounds[1:]) - st.norm.cdf(z_bounds[:-1])
n_prime = n * P_i

table3 = pd.DataFrame({
    'i': range(1, k + 1),
    '[x_i, x_{i+1})': [f"[{bins[i]:.4f}, {bins[i+1]:.4f})" for i in range(k)],
    'n_i': counts_int,
    'P_i': P_i,
    "n'_i": n_prime,
    "(n_i - n'_i)^2 / n'_i": ((counts_int - n_prime)**2) / n_prime,
    "n_i^2 / n'_i": (counts_int**2) / n_prime
})

sums = table3.sum(numeric_only=True)
sums['i'] = 'Σ'
sums['[x_i, x_{i+1})'] = '-'
table3.loc['Σ'] = sums

display(table3)

```

### Задание 4


Пусть имеется $k$ интервалов, $n_i$ — наблюдаемые частоты, $n_i'$ — ожидаемые частоты, причём выполнено условие
$$
\sum_{i=1}^{k} n_i = \sum_{i=1}^{k} n_i' = n,
$$
где $n$ — общий объём выборки.

Статистика $\chi^2$ Пирсона обычно записывается в виде

$$\chi^2 = \sum_{i=1}^{k} \frac{(n_i - n_i')^2}{n_i'}. \tag{1}$$

Покажем, что эта формула тождественно равна

$$\chi^2 = \sum_{i=1}^{k} \frac{n_i^2}{n_i'} - n. \tag{2}$$

Раскроем квадрат разности в числителе дроби из (1):

$$(n_i - n_i')^2 = n_i^2 - 2n_i n_i' + (n_i')^2.$$

Подставим это выражение под знак суммы:

$$\chi^2 = \sum_{i=1}^{k} \frac{n_i^2 - 2n_i n_i' + (n_i')^2}{n_i'}
      = \sum_{i=1}^{k} \left( \frac{n_i^2}{n_i'} - 2n_i + n_i' \right).$$

Разобьём сумму на три независимые суммы:

$$\chi^2 = \sum_{i=1}^{k} \frac{n_i^2}{n_i'} - 2\sum_{i=1}^{k} n_i + \sum_{i=1}^{k} n_i'.$$

Воспользуемся тем, что $\sum_{i=1}^{k} n_i = n$ и $\sum_{i=1}^{k} n_i' = n$:
$$
\chi^2 = \sum_{i=1}^{k} \frac{n_i^2}{n_i'} - 2n + n
      = \sum_{i=1}^{k} \frac{n_i^2}{n_i'} - n.
$$

Таким образом, мы получили формулу (2). Равенство доказано.

```python
print("\n--- 4. КОНТРОЛЬ ВЫЧИСЛЕНИЙ ---")
chi2_formula1 = table3.loc['Σ', "(n_i - n'_i)^2 / n'_i"]
chi2_formula2 = table3.loc['Σ', "n_i^2 / n'_i"] - n

print(f"Расчет по основной формуле: χ^2_набл = {chi2_formula1:.6f}")
print(f"Расчет по упрощенной формуле (Σ(n_i^2/n'_i) - n): χ^2_набл = {chi2_formula2:.6f}")
if np.isclose(chi2_formula1, chi2_formula2):
    print("Численные значения совпадают.\n")
```

### Задание 5


Для проверки гипотезы согласия с нормальным распределением достаточно проерить попадание статистики $\chi^2$ в облать принятия.


```python
print("--- 5. СТАТИСТИЧЕСКОЕ РЕШЕНИЕ ---")
alpha_significance = 0.05
df_chi2 = k - 3

chi2_crit = st.chi2.ppf(1 - alpha_significance, df_chi2)
print(f"Уровень значимости α = {alpha_significance}, Степени свободы df = {df_chi2}")
print(f"χ^2_крит = {chi2_crit:.4f}")
print(f"χ^2_набл = {chi2_formula1:.4f}\n")

if chi2_formula1 < chi2_crit:
    print("ВЫВОД: χ^2_набл < χ^2_крит.")
    print("Нет оснований отвергать нулевую гипотезу. Распределение исследуемой величины можно считать нормальным.")
else:
    print("ВЫВОД: χ^2_набл >= χ^2_крит.")
    print("Нулевая гипотеза отвергается. Распределение исследуемой величины значимо отличается от нормального.")
```
