## Основные теоретические положения

Пусть $X = (X_1, \dots, X_N)$ — выборка из параметрического распределения с параметром $\theta \in \Theta$. Точечная оценка — это измеримая функция $\hat{\theta}: X \to \Theta$ . В основном к ним предъявляются требования:


* Состоятельность: $\forall \epsilon>0 \ \lim\limits_{n\to \infty}\mathbb{P}_{\theta}(|\hat{\theta}_n-\theta|>\epsilon) = 0$
* Несмещенность: $\mathbb{E}_{\theta}(\hat{\theta}) = \theta$
* Эффективность: несмещённость и минимальная дисперсия.

Эмпирическим g-моментом называется следующее выражение:
$\frac{1}{N}\sum g(X_i)$ , или, для интервального ряда, $\frac{1}{N}\sum n_i g(x_i)$. Например, выборочный момент порядка $k$: $M_k = \frac{1}{N}\sum n_i x_i^k$

Условные варианты $u_i$ вводятся для упрощения расчетов моментов в слуае интервального ряда и вычисляются по формуле:

$u_i = \frac{x_i - C}{h}$,

где $x_i$ — середины интервалов, $h$ — ширина интервала, а $C$ — «ложный ноль» — середина интервала, имеющего наибольшую частоту.

В таких обозначениях условный эмпирический момент $k$-го порядка вычисляется по формуле:

$M_k^* = \frac{1}{N} \sum n_i u_i^k$.

Центральный эмпирический момент $k$-го порядка ($m_k$) вычисляется как момент порядка k цкетрированной выборки ($X-\bar{X}$).

Метод моментов позволяет оценить параметры распределения путем приравнивания теоретических моментов распределения к соответствующим эмпирическим мометнам.

## Выполнение работы

### Задание 1


```python
import pandas as pd
import numpy as np
from IPython.display import display

df = pd.read_csv('sample.csv', comment='#')
sample = df['nu'].values
n = len(sample)

k = int(np.floor(1 + 3.322 * np.log10(n)))
counts_int, bins = np.histogram(sample, bins=k)
h = bins[1] - bins[0]
midpoints = (bins[:-1] + bins[1:]) / 2

mode_idx = np.argmax(counts_int)
C = midpoints[mode_idx]

u_i = np.round((midpoints - C) / h).astype(int)

table1 = pd.DataFrame({
    'i': range(1, k + 1),
    'x_i': midpoints,
    'n_i': counts_int,
    'u_i': u_i,
    'n_i*u_i': counts_int * u_i,
    'n_i*u_i^2': counts_int * (u_i**2),
    'n_i*u_i^3': counts_int * (u_i**3),
    'n_i*u_i^4': counts_int * (u_i**4),
    'n_i*(u_i+1)^4': counts_int * ((u_i + 1)**4)
})

sums = table1.sum(numeric_only=True)
sums['i'] = 'Σ'
sums['x_i'] = '-'
sums['u_i'] = '-'
table1.loc['Σ'] = sums

print("Таблица 1. Расчет условных вариант")
display(table1)

sum_nu_plus_1_4 = table1.loc['Σ', 'n_i*(u_i+1)^4']
control_sum = (table1.loc['Σ', 'n_i*u_i^4'] +
               4 * table1.loc['Σ', 'n_i*u_i^3'] +
               6 * table1.loc['Σ', 'n_i*u_i^2'] +
               4 * table1.loc['Σ', 'n_i*u_i'] + n)

print("\n--- Контроль вычислений ---")
print(f"Левая часть (из таблицы): {sum_nu_plus_1_4}")
print(f"Правая часть (по формуле): {control_sum}")
if sum_nu_plus_1_4 == control_sum:
    print("Суммы совпадают.")
else:
    print("Суммы не совпадают.")
```

### Задание 2


```python
M = {
    1: table1.loc['Σ', 'n_i*u_i'] / n,
    2: table1.loc['Σ', 'n_i*u_i^2'] / n,
    3: table1.loc['Σ', 'n_i*u_i^3'] / n,
    4: table1.loc['Σ', 'n_i*u_i^4'] / n
}

m = {
    1: 0,
    2: (M[2] - M[1]**2) * (h**2),
    3: (M[3] - 3*M[1]*M[2] + 2*(M[1]**3)) * (h**3),
    4: (M[4] - 4*M[1]*M[3] + 6*M[2]*(M[1]**2) - 3*(M[1]**4)) * (h**4)
}

table2 = pd.DataFrame({
    'k': [1, 2, 3, 4],
    'M*_k': [np.round(M[1], 4), np.round(M[2], 4), np.round(M[3], 4), np.round(M[4], 4)],
    'm_k': [np.round(m[1], 4), np.round(m[2], 4), np.round(m[3], 4), np.round(m[4], 4)]
})

print("Таблица 2. Условные и центральные эмпирические моменты")
display(table2)
```

### Задание 3


```python
x_bar_cond = C + M[1] * h
D_cond = m[2]
sigma_cond = np.sqrt(D_cond)

x_bar_std = np.sum(midpoints * counts_int) / n
D_std = np.sum(((midpoints - x_bar_std)**2) * counts_int) / n
sigma_std = np.sqrt(D_std)

print("--- Сравнение методов расчета ---")
print(f"Среднее (условные варианты): {x_bar_cond:.6f} | Среднее (интервальный ряд): {x_bar_std:.6f}")
print(f"Дисперсия (условные варианты): {D_cond:.6f} | Дисперсия (интервальный ряд): {D_std:.6f}")
print(f"Выборочное СКО (σ_в): {sigma_cond:.6f}")

diff_mean = abs(x_bar_cond - x_bar_std)
diff_var = abs(D_cond - D_std)
print(f"\nРазница в расчетах среднего: {diff_mean:.10f}")
print(f"Разница в расчетах дисперсии: {diff_var:.10f}")
```

### Задание 4


```python
s2 = (n / (n - 1)) * D_cond
s = np.sqrt(s2)

print(f"Смещенная дисперсия (D_в): {D_cond:.6f} | Исправленная дисперсия (s^2): {s2:.6f}")
print(f"Смещенное СКО (σ_в): {sigma_cond:.6f} | Исправленное СКО (s): {s:.6f}")
```

Расчеты выборочного среднего и дисперсии через условные варианты совпадают с прямыми вычислениями по исходному массиву данных. Исправленные оценки $s^2$ и $s$ закономерно превышают смещенные $D_в$ и $\sigma_в$, так как умножаются на поправочный коэффициент $\frac{n}{n-1}$.

### Задание 5


Статистические оценки коэффициентов ассиметрии и эксцесса вычисляются по формулам $\bar{A} = \frac{\bar{m}_3}{^3}, \ \bar{E} = \frac{\bar{m}_4}{s^4}-3$.

```python
As = m[3] / (s**3)
E = (m[4] / (s**4)) - 3

print(f"Коэффициент асимметрии (A_s): {As:.4f}")
print(f"Коэффициент эксцесса (E): {E:.4f}")
```

Коэффициент асимметрии отрицательный, что указывает на левостороннюю асимметрию.

Также отрицательное значение коэффициента эксцесса свидетельствует о том, что распределение является плосковершинным.

### Задание 6


```python
f_m = counts_int[mode_idx]
f_m_prev = counts_int[mode_idx - 1] if mode_idx > 0 else 0
f_m_next = counts_int[mode_idx + 1] if mode_idx < k - 1 else 0
Mo = bins[mode_idx] + h * ((f_m - f_m_prev) / ((f_m - f_m_prev) + (f_m - f_m_next)))

cum_freqs = np.cumsum(counts_int)
med_idx = np.argmax(cum_freqs >= n / 2)
F_prev = cum_freqs[med_idx - 1] if med_idx > 0 else 0
Me = bins[med_idx] + h * ((n / 2 - F_prev) / counts_int[med_idx])

CV = (s / x_bar_cond) * 100

print(f"Мода (M_o*): {Mo:.4f}")
print(f"Медиана (M_e*): {Me:.4f}")
print(f"Коэффициент вариации (V*): {CV:.2f}%")
```
